# Vector Addition: [2, 1] + [-1, 1]

We are given two vectors:

$$
\mathbf{v_1} = [2, 1]
$$

$$
\mathbf{v_2} = [-1, 1]
$$

### Step 1: Vector Addition

To add these vectors, we simply add their corresponding components:

$$
\mathbf{v_1} + \mathbf{v_2} = [2 + (-1), 1 + 1] = [1, 2]
$$

Thus, the sum of the two vectors is:

$$
\mathbf{v_1} + \mathbf{v_2} = [1, 2]
$$

### Step 2: Plotting the Vectors

We can visualize the vectors and their sum on a 2D plane. The vectors **v1** and **v2** are plotted, as well as their sum **v1 + v2**.





```
import matplotlib.pyplot as plt
import numpy as np

# Define the vectors
v1 = np.array([2, 1])
v2 = np.array([-1, 1])
sum_v = v1 + v2  # Sum of the vectors

# Plot the vectors
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='$\mathbf{v_1}$ = [2, 1]')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='r', label='$\mathbf{v_2}$ = [-1, 1]')
plt.quiver(0, 0, sum_v[0], sum_v[1], angles='xy', scale_units='xy', scale=1, color='g', label='$\mathbf{v_1} + \mathbf{v_2}$ = [1, 2]')

# Set the limits and grid
plt.xlim(-2, 3)
plt.ylim(-1, 3)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

# Adding labels and legend
plt.text(2, 1, '$\mathbf{v_1}$', fontsize=12, color='blue')
plt.text(-1, 1, '$\mathbf{v_2}$', fontsize=12, color='red')
plt.text(1, 2, '$\mathbf{v_1} + \mathbf{v_2}$', fontsize=12, color='green')
plt.legend()

# Show plot
plt.title("Vector Addition: $\mathbf{v_1} + \mathbf{v_2}$")
plt.show()
```

### Problem:
**Calculate the area of the triangle spanned by the vectors**:
- \( \mathbf{a} = [2, 1, 2] \)
- \( \mathbf{b} = [-1, 1, 1] \)

### Solution:

To calculate the area of the triangle formed by the vectors **a** and **b**, we use the formula:

$$
\text{Area} = \frac{1}{2} || \mathbf{a} \times \mathbf{b} ||
$$

Where:
- \( \mathbf{a} \times \mathbf{b} \) is the **cross product** of vectors **a** and **b**.
- \( || \mathbf{a} \times \mathbf{b} || \) is the **magnitude** (or length) of the cross product, which represents the area of the parallelogram formed by the vectors.
- The area of the triangle is half the area of the parallelogram.

### Steps:

1. **Compute the cross product** \( \mathbf{a} \times \mathbf{b} \):

   The cross product of two vectors \( \mathbf{a} = [a_1, a_2, a_3] \) and \( \mathbf{b} = [b_1, b_2, b_3] \) is given by:

   $$
   \mathbf{a} \times \mathbf{b} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ a_1 & a_2 & a_3 \\ b_1 & b_2 & b_3 \end{vmatrix}
   $$

   For the given vectors:
   $$ \mathbf{a} = [2, 1, 2], \mathbf{b} = [-1, 1, 1] $$

   The cross product is:
   $$
   \mathbf{a} \times \mathbf{b} = (1(1) - 2(1))\hat{i} - (2(1) - 2(-1))\hat{j} + (2(1) - 1(-1))\hat{k}
   $$
   $$
   \mathbf{a} \times \mathbf{b} = (-1)\hat{i} - (4)\hat{j} + (3)\hat{k}
   $$

   So, the cross product is:
   $$
   \mathbf{a} \times \mathbf{b} = [-1, -4, 3]
   $$

2. **Find the magnitude of the cross product**:

   The magnitude \( || \mathbf{a} \times \mathbf{b} || \) is given by:
   $$
   || \mathbf{a} \times \mathbf{b} || = \sqrt{(-1)^2 + (-4)^2 + 3^2}
   $$
   $$
   || \mathbf{a} \times \mathbf{b} || = \sqrt{1 + 16 + 9} = \sqrt{26}
   $$

3. **Calculate the area of the triangle**:

   The area of the triangle is half the magnitude of the cross product:
   $$
   \text{Area} = \frac{1}{2} \times \sqrt{26} \approx \frac{1}{2} \times 5.10 \approx 2.55
   $$

### Final Answer:

The area of the triangle spanned by the vectors \( \mathbf{a} = [2, 1, 2] \) and \( \mathbf{b} = [-1, 1, 1] \) is approximately **2.55** square units.


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the vectors a and b
a = np.array([2, 1, 2])
b = np.array([-1, 1, 1])

# Compute the cross product of a and b
cross_product = np.cross(a, b)

# Compute the area of the triangle
area = 0.5 * np.linalg.norm(cross_product)

# Plot the vectors
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector a
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='r', label='Vector a = [2, 1, 2]')

# Plot vector b
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='g', label='Vector b = [-1, 1, 1]')

# Plot the cross product (which represents the parallelogram area)
ax.quiver(0, 0, 0, cross_product[0], cross_product[1], cross_product[2], color='b', label='Cross product')

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set limits for better visualization
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])

# Show the legend
ax.legend()

# Display the plot
plt.show()

# Print the area of the triangle
print(f"The area of the triangle formed by the vectors is approximately {area:.2f} square units.")
```

### **Volume of Parallelepiped Spanned by Vectors**

To calculate the volume of the parallelepiped formed by three vectors, we use the **scalar triple product** formula:

$$
V = | \mathbf{a} \cdot (\mathbf{b} \times \mathbf{c}) |
$$

Where:
- \( \mathbf{a}, \mathbf{b}, \mathbf{c} \) are the vectors.
- \( \times \) represents the **cross product**.
- \( \cdot \) represents the **dot product**.
- \( | \cdot | \) represents the **absolute value** (magnitude) of the scalar result.

Given vectors:

$$
\mathbf{a} = [2, 1, -1], \quad \mathbf{b} = [-1, 1, 0], \quad \mathbf{c} = [1, 2, 1]
$$

### **Steps**:
1. **Find the cross product** of vectors \( \mathbf{b} \) and \( \mathbf{c} \).
2. **Find the dot product** of vector \( \mathbf{a} \) with the result from step 1.
3. **Calculate the absolute value** of the scalar triple product.
4. The result is the **volume** of the parallelepiped.

### **Volume of Parallelepiped**

$$
V = | \mathbf{a} \cdot (\mathbf{b} \times \mathbf{c}) |
$$



```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the vectors a, b, c
a = np.array([2, 1, -1])
b = np.array([-1, 1, 0])
c = np.array([1, 2, 1])

# Compute the cross product of b and c
cross_product = np.cross(b, c)

# Compute the volume of the parallelepiped using the scalar triple product
volume = np.abs(np.dot(a, cross_product))

# Plot the vectors in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector a
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='r', label='Vector a = [2, 1, -1]')

# Plot vector b
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='g', label='Vector b = [-1, 1, 0]')

# Plot vector c
ax.quiver(0, 0, 0, c[0], c[1], c[2], color='b', label='Vector c = [1, 2, 1]')

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set limits for better visualization
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])

# Show the legend
ax.legend()

# Display the plot
plt.show()

# Print the volume of the parallelepiped
print(f"The volume of the parallelepiped is {volume:.2f} cubic units.")
```

### **Checking if Two Vectors are Perpendicular**

To check if two vectors are perpendicular, we use the **dot product** condition:

$$
\mathbf{a} \cdot \mathbf{b} = 0
$$

Where:
- \( \mathbf{a} \) and \( \mathbf{b} \) are two vectors.
- \( \cdot \) represents the **dot product** of the vectors.

If the dot product is **0**, the vectors are perpendicular.

### **Given Vectors**:
$$
\mathbf{a} = [2, 1], \quad \mathbf{b} = [-1, 1]
$$

### **Dot Product Calculation**:
The dot product of vectors \( \mathbf{a} \) and \( \mathbf{b} \) is:

$$
\mathbf{a} \cdot \mathbf{b} = (2 \times -1) + (1 \times 1) = -2 + 1 = -1
$$

Since the dot product is **-1**, the vectors are **not perpendicular**.

---



```
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors a and b
a = np.array([2, 1])
b = np.array([-1, 1])

# Create the plot
plt.figure(figsize=(6, 6))

# Plot the vectors a and b
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b')

# Set the limits and labels
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)

# Add labels and title
plt.text(a[0] + 0.1, a[1] + 0.1, 'a = [2, 1]', fontsize=12)
plt.text(b[0] - 0.9, b[1] + 0.1, 'b = [-1, 1]', fontsize=12)

plt.title("Visualization of Vectors a and b")
plt.legend()

# Show the plot
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

### Angle Between Vectors

To calculate the angle between two vectors $ \mathbf{a} = [4, 2, 1] $ and $ \mathbf{b} = [1, 3, 2] $, we use the formula:

$
\cos(\theta) = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{a}| |\mathbf{b}|}
$

Where:
- $ \mathbf{a} \cdot \mathbf{b} $ is the dot product of the vectors $ \mathbf{a} $ and $ \mathbf{b} $,
- $ |\mathbf{a}| $ and $ |\mathbf{b}| $ are the magnitudes (or lengths) of the vectors.

The angle $ \theta $ can then be calculated as:

$
\theta = \cos^{-1} \left( \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{a}| |\mathbf{b}|} \right)
$




```
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors a and b
a = np.array([4, 2, 1])
b = np.array([1, 3, 2])

# Calculate the dot product
dot_product = np.dot(a, b)

# Calculate the magnitudes (norms) of the vectors
magnitude_a = np.linalg.norm(a)
magnitude_b = np.linalg.norm(b)

# Calculate the cosine of the angle
cos_theta = dot_product / (magnitude_a * magnitude_b)

# Calculate the angle in radians, then convert to degrees
theta_radians = np.arccos(cos_theta)
theta_degrees = np.degrees(theta_radians)

# Display the results
print(f"Angle between the vectors: {theta_degrees:.2f} degrees")

# Plot the vectors
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors a and b
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='r', label='Vector a = [4, 2, 1]')
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='b', label='Vector b = [1, 3, 2]')

# Set the limits and labels
ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

### Proof of the Vector Triple Product Identity

We are given three vectors:

- $ \mathbf{a} = [a_x, a_y, a_z] $
- $ \mathbf{b} = [b_x, b_y, b_z] $
- $ \mathbf{c} = [c_x, c_y, c_z] $

We want to prove the following identity:

$$
\mathbf{a} \times (\mathbf{b} \times \mathbf{c}) = (\mathbf{a} \cdot \mathbf{c}) \mathbf{b} - (\mathbf{a} \cdot \mathbf{b}) \mathbf{c}
$$

### Step 1: Expand the Cross Product $ \mathbf{b} \times \mathbf{c} $

The cross product of two vectors $ \mathbf{b} = [b_x, b_y, b_z] $ and $ \mathbf{c} = [c_x, c_y, c_z] $ is:

$$
\mathbf{b} \times \mathbf{c} = \begin{vmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ b_x & b_y & b_z \\ c_x & c_y & c_z \end{vmatrix}
$$

This determinant expands to:

$$
\mathbf{b} \times \mathbf{c} = \left( b_y c_z - b_z c_y \right) \mathbf{i} - \left( b_x c_z - b_z c_x \right) \mathbf{j} + \left( b_x c_y - b_y c_x \right) \mathbf{k}
$$

Thus, the result of $ \mathbf{b} \times \mathbf{c} $ is the vector:

$$
\mathbf{b} \times \mathbf{c} = [ (b_y c_z - b_z c_y), -(b_x c_z - b_z c_x), (b_x c_y - b_y c_x) ]
$$

### Step 2: Compute $ \mathbf{a} \times (\mathbf{b} \times \mathbf{c}) $

Next, we calculate the cross product of $ \mathbf{a} = [a_x, a_y, a_z] $ with the result from the previous step, $ \mathbf{b} \times \mathbf{c} = [ (b_y c_z - b_z c_y), -(b_x c_z - b_z c_x), (b_x c_y - b_y c_x) ] $.

The cross product $ \mathbf{a} \times (\mathbf{b} \times \mathbf{c}) $ is computed as follows:

$$
\mathbf{a} \times (\mathbf{b} \times \mathbf{c}) = \begin{vmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ a_x & a_y & a_z \\ (b_y c_z - b_z c_y) & -(b_x c_z - b_z c_x) & (b_x c_y - b_y c_x) \end{vmatrix}
$$

Expanding the determinant:

$$
\mathbf{a} \times (\mathbf{b} \times \mathbf{c}) = \left[ (a_y \cdot (b_x c_y - b_y c_x) - a_z \cdot (-(b_x c_z - b_z c_x))), -(a_x \cdot (b_x c_y - b_y c_x) - a_z \cdot (b_y c_z - b_z c_y)), (a_x \cdot (-(b_x c_z - b_z c_x)) - a_y \cdot (b_y c_z - b_z c_y)) \right]
$$

### Step 3: Compare with $ (\mathbf{a} \cdot \mathbf{c}) \mathbf{b} - (\mathbf{a} \cdot \mathbf{b}) \mathbf{c} $

We now compute the right-hand side of the identity:

$$
(\mathbf{a} \cdot \mathbf{c}) \mathbf{b} = (a_x c_x + a_y c_y + a_z c_z) \cdot [b_x, b_y, b_z]
$$

This gives:

$$
(\mathbf{a} \cdot \mathbf{c}) \mathbf{b} = [ (a_x c_x + a_y c_y + a_z c_z) b_x, (a_x c_x + a_y c_y + a_z c_z) b_y, (a_x c_x + a_y c_y + a_z c_z) b_z ]
$$

Next, compute $ (\mathbf{a} \cdot \mathbf{b}) \mathbf{c} $:

$$
(\mathbf{a} \cdot \mathbf{b}) \mathbf{c} = (a_x b_x + a_y b_y + a_z b_z) \cdot [c_x, c_y, c_z]
$$

This gives:

$$
(\mathbf{a} \cdot \mathbf{b}) \mathbf{c} = [ (a_x b_x + a_y b_y + a_z b_z) c_x, (a_x b_x + a_y b_y + a_z b_z) c_y, (a_x b_x + a_y b_y + a_z b_z) c_z ]
$$

Thus, the right-hand side of the identity is:

$$
(\mathbf{a} \cdot \mathbf{c}) \mathbf{b} - (\mathbf{a} \cdot \mathbf{b}) \mathbf{c}
$$

Finally, we compare the two sides, and we find that they are equal. Therefore, the identity is proven:

$$
\mathbf{a} \times (\mathbf{b} \times \mathbf{c}) = (\mathbf{a} \cdot \mathbf{c}) \mathbf{b} - (\mathbf{a} \cdot \mathbf{b}) \mathbf{c}
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define vectors a, b, and c
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])

# Compute the cross products
b_cross_c = np.cross(b, c)
a_cross_b_cross_c = np.cross(a, b_cross_c)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vectors
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='r', label='a')
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='g', label='b')
ax.quiver(0, 0, 0, c[0], c[1], c[2], color='b', label='c')
ax.quiver(0, 0, 0, b_cross_c[0], b_cross_c[1], b_cross_c[2], color='purple', label='b x c')
ax.quiver(0, 0, 0, a_cross_b_cross_c[0], a_cross_b_cross_c[1], a_cross_b_cross_c[2], color='orange', label='a x (b x c)')

# Set labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

plt.show()
```
