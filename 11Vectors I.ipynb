# Vectors: Markdown Notes

## 1. Definition of a Vector
A vector is a quantity that has both magnitude and direction. It is typically represented as an ordered list of numbers, which are the vector's components.

### Examples:
- In two dimensions:  
  $$ \vec{v} = \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} $$  
- In three dimensions:  
  $$ \vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix} $$

---

## 2. Vector Operations

### 2.1 Addition
If  
$$ \vec{a} = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix} $$  
and  
$$ \vec{b} = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}, $$  
then  
$$ \vec{a} + \vec{b} = \begin{bmatrix} a_1 + b_1 \\ a_2 + b_2 \\ a_3 + b_3 \end{bmatrix}. $$

### 2.2 Scalar Multiplication
For a scalar $k$,  
$$ k \vec{a} = \begin{bmatrix} k a_1 \\ k a_2 \\ k a_3 \end{bmatrix}. $$

---

## 3. Dot Product
The dot product of two vectors  
$$ \vec{a} = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix} $$  
and  
$$ \vec{b} = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix} $$  
is  
$$ \vec{a} \cdot \vec{b} = a_1 b_1 + a_2 b_2 + a_3 b_3. $$

---

## 4. Cross Product (in 3D)
The cross product of  
$$ \vec{a} = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix} $$  
and  
$$ \vec{b} = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix} $$  
is  
$$ \vec{a} \times \vec{b} = \begin{bmatrix} a_2 b_3 - a_3 b_2 \\ a_3 b_1 - a_1 b_3 \\ a_1 b_2 - a_2 b_1 \end{bmatrix}. $$

---

## 5. Magnitude of a Vector
The magnitude (length) of vector  
$$ \vec{a} = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix} $$  
is  
$$ |\vec{a}| = \sqrt{a_1^2 + a_2^2 + a_3^2}. $$

---

## 6. Unit Vector
A unit vector in the direction of  
$$ \vec{a} $$  
is  
$$ \hat{a} = \frac{\vec{a}}{|\vec{a}|}. $$

---

## 7. Applications of Vectors
- Physics: Force, velocity, and acceleration
- Graphics: Representing points and transformations
- Engineering: Structural analysis and mechanics

These notes summarize key concepts and operations involving vectors. They are fundamental tools in linear algebra and many applied fields.


# Scaling a Vector to Length 1

Given vector:  
$$ \vec{a} = \begin{bmatrix} 3 \\ 4 \end{bmatrix} $$

The length of the vector is:  
$$ |\vec{a}| = \sqrt{3^2 + 4^2} = 5 $$

Scaling factor to make the length 1:  
$$ k = \frac{1}{|\vec{a}|} = \frac{1}{5} $$

Unit vector:  
$$ \hat{a} = k \vec{a} = \frac{1}{5} \begin{bmatrix} 3 \\ 4 \end{bmatrix} = \begin{bmatrix} 0.6 \\ 0.8 \end{bmatrix} $$


```
import matplotlib.pyplot as plt
import numpy as np

# Define the vector
a = np.array([3, 4])

# Compute the magnitude
magnitude = np.linalg.norm(a)

# Scaling factor and unit vector
scaling_factor = 1 / magnitude
unit_vector = a * scaling_factor

# Display results
print(f"Original vector: {a}")
print(f"Magnitude of vector: {magnitude}")
print(f"Scaling factor: {scaling_factor}")
print(f"Unit vector: {unit_vector}")

# Visualization
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector (3, 4)')
plt.quiver(0, 0, unit_vector[0], unit_vector[1], angles='xy', scale_units='xy', scale=1, color='red', label='Unit Vector (0.6, 0.8)')

plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Original and Scaled Unit Vector")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Calculating the Length and Unit Vector

Given vector:  
$$ \vec{b} = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $$

The length of the vector is:  
$$ |\vec{b}| = \sqrt{1^2 + 1^2} = \sqrt{2} \approx 1.41 $$

Scaling factor to make the length 1:  
$$ k = \frac{1}{|\vec{b}|} = \frac{1}{\sqrt{2}} $$

Unit vector:  
$$ \hat{b} = k \vec{b} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ 1 \end{bmatrix} \approx \begin{bmatrix} 0.707 \\ 0.707 \end{bmatrix} $$


```
import matplotlib.pyplot as plt
import numpy as np

# Define the vector
b = np.array([1, 1])

# Compute the magnitude
magnitude = np.linalg.norm(b)

# Scaling factor and unit vector
scaling_factor = 1 / magnitude
unit_vector = b * scaling_factor

# Display results
print(f"Original vector: {b}")
print(f"Magnitude of vector: {magnitude}")
print(f"Scaling factor: {scaling_factor}")
print(f"Unit vector: {unit_vector}")

# Visualization
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector (1, 1)')
plt.quiver(0, 0, unit_vector[0], unit_vector[1], angles='xy', scale_units='xy', scale=1, color='red', label='Unit Vector (0.707, 0.707)')

plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Original and Scaled Unit Vector")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Plotting the Vector and Unit Vector

Given vector:  
$$ \vec{b} = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $$

Unit vector:  
$$ \hat{b} \approx \begin{bmatrix} 0.707 \\ 0.707 \end{bmatrix} $$

```
import matplotlib.pyplot as plt
import numpy as np

# Define the original vector
b = np.array([1, 1])

# Compute the magnitude and unit vector
magnitude = np.linalg.norm(b)
unit_vector = b / magnitude

# Plot original vector and unit vector
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f'Original Vector: {b.tolist()}')  # .tolist() for cleaner output
plt.quiver(0, 0, unit_vector[0], unit_vector[1], angles='xy', scale_units='xy', scale=1, color='red', label=f'Unit Vector: [{unit_vector[0]:.3f}, {unit_vector[1]:.3f}]')

# Formatting the plot
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Original Vector and Unit Vector")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Finding Cartesian Coordinates in a New Basis

Given a vector **v** in terms of a new basis **b₁**, **b₂**, and **b₃**, the Cartesian coordinates of **v** can be found by solving the following system:

We have:

$$
v = c₁ b₁ + c₂ b₂ + c₃ b₃
$$

Where:

- \( v = [2, 3, 4] \)
- \( b₁ = [1, 0, 1] \)
- \( b₂ = [0, 1, 0] \)
- \( b₃ = [1, 0, -1] \)

We need to solve for the coefficients \( c₁, c₂, c₃ \) in terms of the given basis. This gives us the Cartesian coordinates of **v** in the new basis.

#### Steps:
1. Write the system as a matrix equation:

$$
\begin{bmatrix}
1 & 0 & 1 \\
0 & 1 & 0 \\
1 & 0 & -1
\end{bmatrix}
\begin{bmatrix}
c₁ \\
c₂ \\
c₃
\end{bmatrix}
=
\begin{bmatrix}
2 \\
3 \\
4
\end{bmatrix}
$$

2. Solve for the coefficients \( c₁, c₂, c₃ \).


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the basis vectors and the vector v
b1 = np.array([1, 0, 1])
b2 = np.array([0, 1, 0])
b3 = np.array([1, 0, -1])
v = np.array([2, 3, 4])

# Create the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the basis vectors
ax.quiver(0, 0, 0, b1[0], b1[1], b1[2], color='r', label='b1 (1, 0, 1)')
ax.quiver(0, 0, 0, b2[0], b2[1], b2[2], color='g', label='b2 (0, 1, 0)')
ax.quiver(0, 0, 0, b3[0], b3[1], b3[2], color='b', label='b3 (1, 0, -1)')

# Plot the vector v
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='purple', label='v (2, 3, 4)')

# Customize the plot
ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
ax.set_zlim([0, 4])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
