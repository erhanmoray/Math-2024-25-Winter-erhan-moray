### Dividing a Line Segment in a Given Ratio

We are given two points \( A(-1, 2) \) and \( B(3, -2) \), and we need to divide the line segment connecting these points in the ratio \( 1:3 \).

To find the point \( P(x, y) \) that divides the segment \( AB \) in the ratio \( 1:3 \), we use the section formula:

#### Section Formula
If a line segment connects two points \( A(x_1, y_1) \) and \( B(x_2, y_2) \), and divides the segment in the ratio \( m:n \), then the coordinates of the point dividing the segment are given by:

$$
x = \frac{n x_1 + m x_2}{m + n}
$$

$$
y = \frac{n y_1 + m y_2}{m + n}
$$

### Step 1: Applying the Formula

Here, \( A(-1, 2) \) and \( B(3, -2) \) and the ratio is \( 1:3 \), so we have \( m = 1 \) and \( n = 3 \).

#### Calculating the coordinates of the dividing point:
$$
x = \frac{3 \cdot (-1) + 1 \cdot 3}{1 + 3} = \frac{-3 + 3}{4} = \frac{0}{4} = 0
$$

$$
y = \frac{3 \cdot 2 + 1 \cdot (-2)}{1 + 3} = \frac{6 - 2}{4} = \frac{4}{4} = 1
$$

So, the coordinates of the point \( P \) are \( P(0, 1) \).

### Step 2: Visualization on the Graph

We will now plot the points \( A(-1, 2) \), \( B(3, -2) \), and the dividing point \( P(0, 1) \) on the graph and connect them.



```
import matplotlib.pyplot as plt

# Coordinates of points A, B, and P
A = (-1, 2)
B = (3, -2)
P = (0, 1)

# Create a plot
plt.figure(figsize=(6,6))

# Plot points A, B, and P
plt.scatter([A[0], B[0], P[0]], [A[1], B[1], P[1]], color='red')

# Annotate points A, B, and P
plt.text(A[0], A[1], 'A(-1, 2)', fontsize=12, ha='right')
plt.text(B[0], B[1], 'B(3, -2)', fontsize=12, ha='right')
plt.text(P[0], P[1], 'P(0, 1)', fontsize=12, ha='right')

# Plot the line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], color='blue', linestyle='--', label="Line segment AB")

# Plot the point P that divides the segment in ratio 1:3
plt.plot(P[0], P[1], 'go')  # Point P

# Set axis limits and labels
plt.xlim(-2, 4)
plt.ylim(-3, 3)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)

# Title and display the plot
plt.title("Dividing the Line Segment AB in the Ratio 1:3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Projecting Vector onto OX and OY Axes

Given the vector \( \mathbf{a} = (3, 4) \), we will project it onto the OX and OY axes.

#### Projection of a Vector onto the OX Axis:

The projection of vector \( \mathbf{a} \) onto the OX axis is given by:

$$
\text{Proj}_{OX}(\mathbf{a}) = (a_x, 0)
$$

For \( \mathbf{a} = (3, 4) \), the projection onto the OX axis is:

$$
\text{Proj}_{OX}(\mathbf{a}) = (3, 0)
$$

#### Projection of a Vector onto the OY Axis:

The projection of vector \( \mathbf{a} \) onto the OY axis is given by:

$$
\text{Proj}_{OY}(\mathbf{a}) = (0, a_y)
$$

For \( \mathbf{a} = (3, 4) \), the projection onto the OY axis is:

$$
\text{Proj}_{OY}(\mathbf{a}) = (0, 4)
$$

### Step 1: Visualizing the Projections

We will now plot the vector \( \mathbf{a} = (3, 4) \), its projection onto the OX axis, and its projection onto the OY axis on a graph.



```
import matplotlib.pyplot as plt

# Coordinates of the vector a and its projections
a = (3, 4)
proj_ox = (3, 0)
proj_oy = (0, 4)

# Create a plot
plt.figure(figsize=(6,6))

# Plot the vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label="Vector a(3, 4)")

# Plot the projection onto OX axis
plt.quiver(0, 0, proj_ox[0], proj_ox[1], angles='xy', scale_units='xy', scale=1, color='red', label="Projection onto OX (3, 0)")

# Plot the projection onto OY axis
plt.quiver(0, 0, proj_oy[0], proj_oy[1], angles='xy', scale_units='xy', scale=1, color='green', label="Projection onto OY (0, 4)")

# Set axis limits and labels
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)

# Title and display the plot
plt.title("Projection of Vector a = (3, 4) onto OX and OY Axes")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Projecting Vector \( \mathbf{a} = (2, 3) \) onto Vector \( \mathbf{b} = (1, 1) \)

The projection of vector \( \mathbf{a} \) onto vector \( \mathbf{b} \) is given by the formula:

$$
\text{Proj}_{\mathbf{b}}(\mathbf{a}) = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} \mathbf{b}
$$

#### Step 1: Calculate the Dot Product

First, calculate the dot product of vectors \( \mathbf{a} \) and \( \mathbf{b} \):

$$
\mathbf{a} \cdot \mathbf{b} = (2)(1) + (3)(1) = 5
$$

Next, calculate the dot product of vector \( \mathbf{b} \) with itself:

$$
\mathbf{b} \cdot \mathbf{b} = (1)(1) + (1)(1) = 2
$$

#### Step 2: Calculate the Projection

Now, calculate the projection of vector \( \mathbf{a} \) onto vector \( \mathbf{b} \):

$$
\text{Proj}_{\mathbf{b}}(\mathbf{a}) = \frac{5}{2} \mathbf{b} = \frac{5}{2} (1, 1) = \left(\frac{5}{2}, \frac{5}{2}\right)
$$

Thus, the projection of vector \( \mathbf{a} = (2, 3) \) onto vector \( \mathbf{b} = (1, 1) \) is:

$$
\text{Proj}_{\mathbf{b}}(\mathbf{a}) = \left(\frac{5}{2}, \frac{5}{2}\right)
$$

### Step 3: Visualizing the Projection

We will now plot the vector \( \mathbf{a} = (2, 3) \), the vector \( \mathbf{b} = (1, 1) \), and the projection of \( \mathbf{a} \) onto \( \mathbf{b} \) on a graph.



```
import matplotlib.pyplot as plt
import numpy as np

# Vectors a and b
a = np.array([2, 3])
b = np.array([1, 1])

# Calculate the projection of a onto b
projection = (np.dot(a, b) / np.dot(b, b)) * b

# Create a plot
plt.figure(figsize=(6,6))

# Plot the vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label="Vector a (2, 3)")

# Plot the vector b
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label="Vector b (1, 1)")

# Plot the projection of a onto b
plt.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='red', label="Projection of a onto b")

# Set axis limits and labels
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)

# Title and display the plot
plt.title("Projection of Vector a = (2, 3) onto Vector b = (1, 1)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Projecting Vector \( \mathbf{b} = (1, 1) \) onto Vector \( \mathbf{a} = (2, 3) \)

```
import matplotlib.pyplot as plt
import numpy as np

# Vectors b and a
b = np.array([1, 1])
a = np.array([2, 3])

# Calculate the projection of b onto a
projection = (np.dot(b, a) / np.dot(a, a)) * a

# Create a plot
plt.figure(figsize=(6,6))

# Plot the vector b
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='blue', label="Vector b (1, 1)")

# Plot the vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='green', label="Vector a (2, 3)")

# Plot the projection of b onto a
plt.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='red', label="Projection of b onto a")

# Set axis limits and labels
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.grid(True)

# Title and display the plot
plt.title("Projection of Vector b = (1, 1) onto Vector a = (2, 3)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
