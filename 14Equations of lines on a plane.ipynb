# Equations of Lines on a Plane

The equation of a line on a plane can be represented in different forms depending on the information available. Below are common forms of line equations:

### 1. **Slope-Intercept Form**

If the slope \(m\) and y-intercept \(b\) of a line are known, the equation is:

$$
y = mx + b
$$

### 2. **Point-Slope Form**

If a point \((x_1, y_1)\) on the line and the slope \(m\) are known, the equation is:

$$
y - y_1 = m(x - x_1)
$$

### 3. **Two-Point Form**

If two points \((x_1, y_1)\) and \((x_2, y_2)\) are known, the equation of the line can be found using:

$$
y - y_1 = \frac{y_2 - y_1}{x_2 - x_1}(x - x_1)
$$

### 4. **General Form**

The general form of the equation of a line is:

$$
Ax + By + C = 0
$$

where \(A\), \(B\), and \(C\) are constants.

### 5. **Horizontal and Vertical Lines**

- A horizontal line has an equation:  
  $$ y = k $$  
  where \(k\) is a constant.
  
- A vertical line has an equation:  
  $$ x = h $$  
  where \(h\) is a constant.

### Example: Find the Equation of a Line

Find the equation of a line passing through points \((1, 2)\) and \((3, 4)\).

#### Step 1: Calculate the Slope

The slope \(m\) is:

$$
m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{4 - 2}{3 - 1} = 1
$$

#### Step 2: Use the Point-Slope Form

Using point \((1, 2)\), the equation is:

$$
y - 2 = 1(x - 1)
$$

Simplifying:

$$
y = x + 1
$$

This is the slope-intercept form of the equation.


# Finding the Equation of a Line

We are given two points $A(1, 2)$ and $B(3, 4)$. The goal is to find the equation of the line passing through these points and visualize it.

### Step 1: Find the Slope

The formula for the slope $m$ between two points $(x_1, y_1)$ and $(x_2, y_2)$ is:

$m = \frac{y_2 - y_1}{x_2 - x_1}$

Substituting the values of points $A(1, 2)$ and $B(3, 4)$:

$m = \frac{4 - 2}{3 - 1} = \frac{2}{2} = 1$

### Step 2: Use the Point-Slope Form

The point-slope form is:

$y - y_1 = m(x - x_1)$

Using point $A(1, 2)$:

$y - 2 = 1(x - 1)$

Simplifying:

$y = x + 1$

### Final Equation of the Line

$y = x + 1$


```
import numpy as np
import matplotlib.pyplot as plt

# Define points
A = (1, 2)
B = (3, 4)

# Define the line
x_values = np.linspace(0, 5, 100)
y_values = x_values + 1  # Line equation: y = x + 1

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = x + 1', color='blue')
plt.scatter(*A, color='red', label='Point A(1, 2)')
plt.scatter(*B, color='green', label='Point B(3, 4)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line through Points A(1, 2) and B(3, 4)')
plt.legend()
plt.show()
```

# Finding the Equation of a Line

We are given point $A(1, 2)$, and the line is parallel to $y = 2x + 3$. We need to find the equation of a line passing through point $A$.

### Step 1: Identify the Slope

Since the given line is $y = 2x + 3$, its slope is $m = 2$. The line we are finding is parallel, so it has the same slope $m = 2$.

### Step 2: Use the Point-Slope Form

The point-slope form of a line is:

$y - y_1 = m(x - x_1)$

Substituting point $A(1, 2)$ and slope $m = 2$:

$y - 2 = 2(x - 1)$

Simplifying:

$y - 2 = 2x - 2$

$y = 2x$

### Final Equation of the Line

$y = 2x$


```
import numpy as np
import matplotlib.pyplot as plt

# Define point A
A = (1, 2)

# Define the line
x_values = np.linspace(-2, 5, 100)
y_parallel_line = 2 * x_values + 3  # Original line: y = 2x + 3
y_new_line = 2 * x_values  # New line: y = 2x

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_parallel_line, '--', label='y = 2x + 3 (Given Line)', color='gray')
plt.plot(x_values, y_new_line, label='y = 2x (New Line)', color='blue')
plt.scatter(*A, color='red', label='Point A(1, 2)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Through A(1, 2) Parallel to y = 2x + 3')
plt.legend()
plt.show()
```

# Finding the Equation of a Line

We are given point $A(1, 2)$, and the line is perpendicular to $y = 2x + 3$. We need to find the equation of a line passing through point $A$ and perpendicular to the given line.

### Step 1: Identify the Slope of the Given Line

The slope of the given line $y = 2x + 3$ is $m_1 = 2$.

### Step 2: Find the Slope of the Perpendicular Line

The slope of a line perpendicular to a line with slope $m_1$ is:

$m_2 = -\frac{1}{m_1} = -\frac{1}{2}$

### Step 3: Use the Point-Slope Form

The point-slope form of a line is:

$y - y_1 = m(x - x_1)$

Substituting point $A(1, 2)$ and slope $m_2 = -\frac{1}{2}$:

$y - 2 = -\frac{1}{2}(x - 1)$

Simplifying:

$y - 2 = -\frac{1}{2}x + \frac{1}{2}$

$y = -\frac{1}{2}x + \frac{5}{2}$

### Final Equation of the Line

$y = -\frac{1}{2}x + \frac{5}{2}$


```
import numpy as np
import matplotlib.pyplot as plt

# Define point A
A = (1, 2)

# Define the lines
x_values = np.linspace(-2, 5, 100)
y_original_line = 2 * x_values + 3  # Original line: y = 2x + 3
y_perpendicular_line = -0.5 * x_values + 2.5  # New line: y = -0.5x + 2.5

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_original_line, '--', label='y = 2x + 3 (Original Line)', color='gray')
plt.plot(x_values, y_perpendicular_line, label='y = -0.5x + 2.5 (Perpendicular Line)', color='blue')
plt.scatter(*A, color='red', label='Point A(1, 2)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Through A(1, 2) Perpendicular to y = 2x + 3')
plt.legend()
plt.show()
```

### Intersection Point and Angle Between Two Lines

We are given two lines:

1. $ y = 2x + 3 $
2. $ y = 3x + 2 $

#### 1. Finding the Intersection Point

To find the intersection point of these two lines, set the equations equal to each other:

$
2x + 3 = 3x + 2
$

Solving for $ x $:

$
2x - 3x = 2 - 3 \quad \Rightarrow \quad -x = -1 \quad \Rightarrow \quad x = 1
$

Substitute $ x = 1 $ back into either of the equations to find $ y $. Using the first equation:

$
y = 2(1) + 3 = 5
$

Thus, the intersection point is $ (1, 5) $.

#### 2. Calculating the Angle Between the Two Lines

The formula for the angle $ \theta $ between two lines with slopes $ m_1 $ and $ m_2 $ is:

$
\tan(\theta) = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|
$

For the two lines:

- The slope of the first line is $ m_1 = 2 $
- The slope of the second line is $ m_2 = 3 $

Now substitute these values into the formula:

$
\tan(\theta) = \left| \frac{2 - 3}{1 + 2 \times 3} \right| = \left| \frac{-1}{7} \right| = \frac{1}{7}
$

Thus, the angle $ \theta $ is:

$
\theta = \tan^{-1} \left( \frac{1}{7} \right)
$

Using a calculator:

$
\theta \approx 8.13^\circ
$




```
import numpy as np
import matplotlib.pyplot as plt

# Calculate the angle between the lines
# The formula for the angle between two lines is:
# tan(theta) = |(m1 - m2) / (1 + m1 * m2)|
# where m1 and m2 are the slopes of the two lines.

m1 = 2  # Slope of the first line y = 2x + 3
m2 = 3  # Slope of the second line y = 3x + 2

# Calculate tan(theta)
tan_theta = abs((m1 - m2) / (1 + m1 * m2))

# Find theta in radians and then convert to degrees
theta = np.arctan(tan_theta)
theta_deg = np.degrees(theta)  # Angle in degrees

# Set up the x values for plotting
x = np.linspace(-2, 3, 400)  # Range of x values for a smooth plot

# Define the y values for each line using the equations
y1 = 2*x + 3  # Line 1: y = 2x + 3
y2 = 3*x + 2  # Line 2: y = 3x + 2

# Create the plot
plt.figure(figsize=(8, 6))

# Plot both lines
plt.plot(x, y1, label='y = 2x + 3', color='blue', linewidth=2)  # Line 1 (blue)
plt.plot(x, y2, label='y = 3x + 2', color='red', linewidth=2)  # Line 2 (red)

# Plot the intersection point (1, 5)
plt.scatter(1, 5, color='black', zorder=5)  # Intersection point at (1,5)
plt.text(1, 5, '  (1, 5)', fontsize=12, verticalalignment='bottom', color='black')  # Label the intersection

# Adding axis lines for better visual understanding
plt.axhline(0, color='black',linewidth=0.5)  # x-axis (horizontal line)
plt.axvline(0, color='black',linewidth=0.5)  # y-axis (vertical line)

# Adding grid for better reference
plt.grid(True)

# Adding title and labels
plt.title(f'Intersection at (1, 5)\nAngle between lines: {theta_deg:.2f}', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')

# Show the legend
plt.legend()

# Display the plot
plt.show()

# Return the calculated angle in degrees
theta_deg
```

### Equation of the Line Passing Through Point A(1, 2) and Parallel to the Vector v = [2, 3]

We are given:
- Point $A(1, 2)$
- Vector $\mathbf{v} = [2, 3]$

#### 1. Parametric Form of the Line

The equation of a line passing through a point $A(x_1, y_1)$ and parallel to a vector $\mathbf{v} = [v_x, v_y]$ can be written in parametric form as:

$$
x = x_1 + v_x t
$$
$$
y = y_1 + v_y t
$$

Where:
- $(x_1, y_1) = (1, 2)$ is the point the line passes through,
- $(v_x, v_y) = (2, 3)$ is the direction vector,
- $t$ is the parameter.

Substituting the given values, the equations become:

$$
x = 1 + 2t
$$
$$
y = 2 + 3t
$$

Thus, the parametric equations of the line are:

$$
x = 1 + 2t
$$
$$
y = 2 + 3t
$$

#### 2. Converting to a Standard Line Equation

To convert this to a standard form, we solve for $t$ in terms of $x$:

$$
x = 1 + 2t \quad \Rightarrow \quad t = \frac{x - 1}{2}
$$

Substitute this into the equation for $y$:

$$
y = 2 + 3\left(\frac{x - 1}{2}\right)
$$
$$
y = 2 + \frac{3(x - 1)}{2}
$$
$$
y = 2 + \frac{3x}{2} - \frac{3}{2}
$$
$$
y = \frac{4}{2} - \frac{3}{2} + \frac{3x}{2}
$$
$$
y = \frac{1}{2} + \frac{3x}{2}
$$

Thus, the equation of the line is:

$$
y = \frac{3x}{2} + \frac{1}{2}
$$



```
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values
x = np.linspace(-2, 4, 400)

# Define the line equation y = (3/2)x + 1/2
y = (3/2) * x + 1/2

# Point A(1, 2)
A = (1, 2)

# Plot the line
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = \frac{3x}{2} + \frac{1}{2}$', color='blue', linewidth=2)

# Plot point A(1, 2)
plt.scatter(A[0], A[1], color='red', zorder=5)  # Point A
plt.text(A[0], A[1], '  A(1, 2)', fontsize=12, verticalalignment='bottom', color='red')  # Label point A

# Adding axis lines for better visual understanding
plt.axhline(0, color='black',linewidth=0.5)  # x-axis (horizontal line)
plt.axvline(0, color='black',linewidth=0.5)  # y-axis (vertical line)

# Adding grid for better reference
plt.grid(True)

# Adding title and labels
plt.title('Line passing through A(1, 2) and parallel to vector v = [2, 3]', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

### Lines Parallel and Perpendicular to the Line $y = 2x + 3$

We are given the line equation:

$$
y = 2x + 3
$$

This is in slope-intercept form, where the slope $m = 2$.

#### 1. Line Parallel to $y = 2x + 3$

A line parallel to another line has the same slope. Therefore, a line parallel to $y = 2x + 3$ will also have a slope of $2$. Let's choose a point, say $(1, 1)$, and use the point-slope form to write the equation of the parallel line.

The point-slope form of a line is:

$$
y - y_1 = m(x - x_1)
$$

Where $m$ is the slope and $(x_1, y_1)$ is the point. Substituting the slope $m = 2$ and point $(1, 1)$, we get:

$$
y - 1 = 2(x - 1)
$$

Simplifying:

$$
y = 2x - 1
$$

So, the equation of the line parallel to $y = 2x + 3$ is:

$$
y = 2x - 1
$$

#### 2. Line Perpendicular to $y = 2x + 3$

Two lines are perpendicular if the product of their slopes is $-1$. The slope of the given line is $m = 2$. For a line to be perpendicular, the product of its slope $m_{\perp}$ with $2$ must be $-1$:

$$
m \cdot m_{\perp} = -1
$$

Substituting $m = 2$:

$$
2 \cdot m_{\perp} = -1
$$

Solving for $m_{\perp}$:

$$
m_{\perp} = -\frac{1}{2}
$$

So, the slope of the perpendicular line is $-\frac{1}{2}$. Now, let's choose the point $(1, 1)$ again and use the point-slope form to find the equation of the perpendicular line.

Substituting the slope $m_{\perp} = -\frac{1}{2}$ and the point $(1, 1)$ into the point-slope form:

$$
y - 1 = -\frac{1}{2}(x - 1)
$$

Simplifying:

$$
y = -\frac{1}{2}x + \frac{3}{2}
$$

So, the equation of the line perpendicular to $y = 2x + 3$ is:

$$
y = -\frac{1}{2}x + \frac{3}{2}
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values
x = np.linspace(-2, 4, 400)

# Define the given line y = 2x + 3
y_given = 2 * x + 3

# Define the parallel line y = 2x - 1
y_parallel = 2 * x - 1

# Define the perpendicular line y = -1/2 * x + 3/2
y_perpendicular = (-1/2) * x + 3/2

# Plot the lines
plt.figure(figsize=(8, 6))
plt.plot(x, y_given, label=r'$y = 2x + 3$', color='blue', linewidth=2)
plt.plot(x, y_parallel, label=r'$y = 2x - 1$', color='green', linewidth=2)
plt.plot(x, y_perpendicular, label=r'$y = -\frac{1}{2}x + \frac{3}{2}$', color='red', linewidth=2)

# Plot points of intersection for reference (arbitrary points on the lines)
plt.scatter(1, 3, color='blue', zorder=5)  # Point on the given line
plt.scatter(1, 1, color='green', zorder=5)  # Point on the parallel line
plt.scatter(1, 1.5, color='red', zorder=5)  # Point on the perpendicular line

# Adding axis lines for better visual understanding
plt.axhline(0, color='black',linewidth=0.5)  # x-axis (horizontal line)
plt.axvline(0, color='black',linewidth=0.5)  # y-axis (vertical line)

# Adding grid for better reference
plt.grid(True)

# Adding title and labels
plt.title('Parallel and Perpendicular Lines to $y = 2x + 3$', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

### Distance from Point A(1, 2) to the Line $y = 2x + 3$

We are given:
- Line equation: $y = 2x + 3$, which we can rewrite as $2x - y + 3 = 0$.
- Point $A(1, 2)$.

To find the distance from a point $(x_1, y_1)$ to a line $Ax + By + C = 0$, we use the formula:

$$
d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

From the line equation $2x - y + 3 = 0$, we have:
- $A = 2$
- $B = -1$
- $C = 3$

Substitute the values into the formula:

$$
d = \frac{|2(1) + (-1)(2) + 3|}{\sqrt{2^2 + (-1)^2}} = \frac{|2 - 2 + 3|}{\sqrt{4 + 1}} = \frac{|3|}{\sqrt{5}} = \frac{3}{\sqrt{5}} \approx 1.34
$$

Thus, the distance from point $A(1, 2)$ to the line $y = 2x + 3$ is approximately $1.34$ units.


```
import numpy as np
import matplotlib.pyplot as plt

# Given line y = 2x + 3, rewritten as 2x - y + 3 = 0
def line_eq(x):
    return 2 * x + 3

# Point A(1, 2)
A = (1, 2)

# Create an array of x values for the line plot
x = np.linspace(-2, 4, 400)
y = line_eq(x)

# Plot the line y = 2x + 3
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = 2x + 3$', color='blue', linewidth=2)

# Plot point A(1, 2)
plt.scatter(A[0], A[1], color='red', zorder=5)  # Point A
plt.text(A[0], A[1], '  A(1, 2)', fontsize=12, verticalalignment='bottom', color='red')  # Label point A

# Plot the perpendicular line from point A to the given line
# Calculate slope of the given line (m = 2), perpendicular slope will be -1/2
slope_perpendicular = -1 / 2
# Using the point-slope form of a line: y - y1 = m(x - x1)
x_perpendicular = np.linspace(-2, 4, 400)
y_perpendicular = slope_perpendicular * (x_perpendicular - A[0]) + A[1]
plt.plot(x_perpendicular, y_perpendicular, label=r'Perpendicular from A', color='green', linestyle='--')

# Adding axis lines for better visual understanding
plt.axhline(0, color='black', linewidth=0.5)  # x-axis (horizontal line)
plt.axvline(0, color='black', linewidth=0.5)  # y-axis (vertical line)

# Adding grid for better reference
plt.grid(True)

# Adding title and labels
plt.title('Distance from Point A(1, 2) to the Line $y = 2x + 3$', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

### Equation of the Line Passing Through Points $A(2, 0)$ and $B(0, 3)$

To find the equation of the line passing through two points $A(x_1, y_1)$ and $B(x_2, y_2)$, we use the **two-point formula**:

$$
y - y_1 = m(x - x_1)
$$

Where $m$ is the slope of the line, calculated as:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

#### Step-by-Step Calculation:

Given:
- Point $A(2, 0)$
- Point $B(0, 3)$

1. **Calculate the slope:**

$$
m = \frac{3 - 0}{0 - 2} = \frac{3}{-2} = -\frac{3}{2}
$$

$$
y - 0 = -\frac{3}{2}(x - 2)

$$
y = -\frac{3}{2}x + 3
$$

Thus, the equation of the line passing through points $A(2, 0)$ and $B(0, 3)$ is:

$$
y = -\frac{3}{2}x + 3
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Points A(2, 0) and B(0, 3)
A = (2, 0)
B = (0, 3)

# Create an array of x values
x = np.linspace(-1, 3, 400)

# Define the line equation y = -3/2 * x + 3
y = -3/2 * x + 3

# Plot the line
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = -\frac{3}{2}x + 3$', color='blue', linewidth=2)

# Plot points A and B
plt.scatter(A[0], A[1], color='red', zorder=5)  # Point A
plt.scatter(B[0], B[1], color='green', zorder=5)  # Point B

# Annotating the points
plt.text(A[0], A[1], '  A(2, 0)', fontsize=12, verticalalignment='bottom', color='red')  # Label point A
plt.text(B[0], B[1], '  B(0, 3)', fontsize=12, verticalalignment='bottom', color='green')  # Label point B

# Adding axis lines for better visual understanding
plt.axhline(0, color='black', linewidth=0.5)  # x-axis (horizontal line)
plt.axvline(0, color='black', linewidth=0.5)  # y-axis (vertical line)

# Adding grid for better reference
plt.grid(True)

# Adding title and labels
plt.title('Line Passing Through Points A(2, 0) and B(0, 3)', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

### Calculate the Angle Between the Line $y = x + 3$ and the $O_x$ Axis

To calculate the angle between a line and the $O_x$-axis, we use the following formula for the slope $m$ of the line:

$$
\theta = \tan^{-1}(m)
$$

Where:
- $\theta$ is the angle the line makes with the $O_x$-axis.
- $m$ is the slope of the line, which is the coefficient of $x$ in the equation of the line.

#### Step-by-Step Calculation:

Given the line equation:

$$
y = x + 3
$$

The slope $m$ is $1$ because the coefficient of $x$ is $1$.

Now, calculate the angle:

$$
\theta = \tan^{-1}(1)
$$

$$
\theta = 45^\circ
$$

Thus, the angle between the line $y = x + 3$ and the $O_x$-axis is $45^\circ$.


```
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the line equation y = x + 3
def line_eq(x):
    return x + 3

# Step 2: Create an array of x values that will be used to plot the line
# We are generating values from -2 to 4 to get a good range for visualizing the line
x = np.linspace(-2, 4, 400)

# Step 3: Calculate the corresponding y values for each x value using the line equation
y = line_eq(x)

# Step 4: Create the plot
plt.figure(figsize=(8, 6))

# Step 5: Plot the line y = x + 3 on the graph
# The line is plotted with blue color and label for the legend
plt.plot(x, y, label=r'$y = x + 3$', color='blue', linewidth=2)

# Step 6: Plot the x-axis (O_x) and y-axis
# We are adding horizontal and vertical lines at y = 0 (x-axis) and x = 0 (y-axis)
plt.axhline(0, color='black', linewidth=0.5)  # x-axis (O_x)
plt.axvline(0, color='black', linewidth=0.5)  # y-axis (O_y)

# Step 7: Add an annotation to indicate the angle
# We are adding the angle  = 45 as text at the location (1, 4) on the plot
plt.text(1, 4, r'$\theta = 45^\circ$', fontsize=14, color='red')

# Step 8: Enable grid for better visual understanding of the plot
plt.grid(True)

# Step 9: Add title and axis labels
plt.title('Line $y = x + 3$ and the $O_x$ Axis', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')

# Step 10: Show the legend (to display the equation of the line)
plt.legend()

# Step 11: Display the plot
plt.show()
```

### Find a Vector Perpendicular to the Line $x + y + 1 = 0$

To find a vector perpendicular to a line, we can use the fact that the direction vector of the line is normal to any vector perpendicular to it. The general form of the equation of a line is:

$$
Ax + By + C = 0
$$

Where $A$ and $B$ are the coefficients of $x$ and $y$, respectively. A vector perpendicular to the line is given by the vector $\mathbf{v} = \langle A, B \rangle$.

#### Given:
The equation of the line is:

$$
x + y + 1 = 0
$$

From this, we can directly read off the values of $A$ and $B$:
- $A = 1$
- $B = 1$

Thus, a vector perpendicular to this line is:

$$
\mathbf{v} = \langle 1, 1 \rangle
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Define the line equation x + y + 1 = 0, or y = -x - 1
def line_eq(x):
    return -x - 1

# Create an array of x values for plotting the line
x_vals = np.linspace(-5, 5, 400)
y_vals = line_eq(x_vals)

# Define the perpendicular vector v = (1, 1)
perp_vector = np.array([1, 1])

# Plot the line
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$x + y + 1 = 0$', color='blue', linewidth=2)

# Plot the perpendicular vector
origin = np.array([0, -1])  # Starting point of the vector (for example, using the point (0, -1))
plt.quiver(*origin, *perp_vector, angles='xy', scale_units='xy', scale=1, color='red', label=r'Perpendicular Vector $\mathbf{v} = (1, 1)$')

# Adding grid, axis labels, and title
plt.axhline(0, color='black',linewidth=0.5)  # x-axis
plt.axvline(0, color='black',linewidth=0.5)  # y-axis
plt.grid(True)
plt.title('Line $x + y + 1 = 0$ and Perpendicular Vector $\mathbf{v} = (1, 1)$')
plt.xlabel('x')
plt.ylabel('y')

# Show legend
plt.legend()

# Display the plot
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```
