### Equations of Second-Order Curves (Conic Sections)

Conic sections are curves that can be defined by second-order polynomial equations in two variables. These curves are obtained by intersecting a plane with a cone. The general form of the equation for conic sections is:

$$
Ax^2 + Bxy + Cy^2 + Dx + Ey + F = 0
$$

Where:
- $A$, $B$, $C$, $D$, $E$, and $F$ are constants.
- The type of conic section depends on the values of these constants.

### Types of Conic Sections:

1. **Circle**
   - A circle is a special case of an ellipse where the two axes have the same length.
   - The general equation of a circle is:

   $$
   x^2 + y^2 + 2Dx + 2Ey + F = 0
   $$

   Where $D$ and $E$ determine the center of the circle, and $F$ determines the radius.

2. **Ellipse**
   - An ellipse is a generalization of a circle where the axes are unequal.
   - The general equation of an ellipse is:

   $$
   Ax^2 + Cy^2 + Dx + Ey + F = 0
   $$

   Where $A$ and $C$ are the coefficients of $x^2$ and $y^2$, and $A \neq C$.

3. **Hyperbola**
   - A hyperbola is a curve where the difference of distances from any point on the curve to two fixed points (foci) is constant.
   - The general equation of a hyperbola is:

   $$
   Ax^2 - By^2 + Dx + Ey + F = 0
   $$

   Where $A$ and $B$ have opposite signs, and the curve consists of two branches.

4. **Parabola**
   - A parabola is a curve where any point on the curve is equidistant from a fixed point (focus) and a fixed line (directrix).
   - The general equation of a parabola is:

   $$
   Ax^2 + Dx + Ey + F = 0
   $$

   or

   $$
   By^2 + Dx + Ey + F = 0
   $$

   Where one of the second-order coefficients is zero.

### Discriminant and Type of Conic:

The discriminant of the general conic equation is given by:

$$
\Delta = B^2 - 4AC
$$

- **Circle**: If \( A = C \) and \( B = 0 \), the conic is a circle.
- **Ellipse**: If \( A \neq C \) and \( B^2 - 4AC < 0 \), the conic is an ellipse.
- **Hyperbola**: If \( B^2 - 4AC > 0 \), the conic is a hyperbola.
- **Parabola**: If \( B^2 - 4AC = 0 \), the conic is a parabola.


### Equation of a Circle with Center at $A(1, 2)$ and Radius $r = 3$

A **circle** is the set of all points in a plane that are at a fixed distance (the radius) from a given point (the center).

The **general equation** of a circle with center at $(h, k)$ and radius $r$ is:

$$
(x - h)^2 + (y - k)^2 = r^2
$$

Where:
- $(h, k)$ is the **center** of the circle.
- $r$ is the **radius** of the circle.

This is derived from the **distance formula** in geometry, which calculates the distance between a point $(x, y)$ and the center of the circle $(h, k)$.

#### Given:
- **Center**: $A(1, 2)$
- **Radius**: $r = 3$

#### Step-by-Step Derivation:
1. The **center** of the circle is given as $A(1, 2)$, meaning $h = 1$ and $k = 2$.
2. The **radius** is given as $r = 3$.

Now, substitute these values into the general equation of the circle:

$$
(x - 1)^2 + (y - 2)^2 = 3^2
$$

3. Simplify the equation:

$$
(x - 1)^2 + (y - 2)^2 = 9
$$

Thus, the equation of the circle with center at $A(1, 2)$ and radius $r = 3$ is:

$$
(x - 1)^2 + (y - 2)^2 = 9
$$

This is the required equation of the circle.


```
import numpy as np
import matplotlib.pyplot as plt

# Center and radius of the circle
center = (1, 2)
radius = 3

# Create theta values for plotting the circle
theta = np.linspace(0, 2*np.pi, 400)

# Parametric equations for the circle
x_circle = center[0] + radius * np.cos(theta)
y_circle = center[1] + radius * np.sin(theta)

# Create the plot
plt.figure(figsize=(6,6))

# Plot the circle
plt.plot(x_circle, y_circle, label=r'$(x - 1)^2 + (y - 2)^2 = 9$', color='blue')

# Plot the center of the circle
plt.scatter(center[0], center[1], color='red', zorder=5, label='Center A(1, 2)')

# Plot the x-axis and y-axis
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Set equal scaling and limits for the axes
plt.axis('equal')
plt.xlim(-2, 6)
plt.ylim(-2, 6)

# Add title, labels, and grid
plt.title('Circle with Center A(1, 2) and Radius 3', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

### Equation of a Parabola Intersecting the $x$-Axis at $x = 2$ and $x = 4$, and Passing Through Point $(3, 1)$

The general equation of a parabola is given by:

$$
y = ax^2 + bx + c
$$

However, since the parabola intersects the $x$-axis at points $x = 2$ and $x = 4$, we know that these are the roots of the parabola. The factored form of the parabola is:

$$
y = a(x - 2)(x - 4)
$$

Where $a$ is a constant that we need to determine.

#### Step 1: Use the point $(3, 1)$ to find the value of $a$
We are given that the parabola passes through the point $(3, 1)$. Substituting $x = 3$ and $y = 1$ into the equation:

$$
1 = a(3 - 2)(3 - 4)
$$

Simplifying:

$$
1 = a(1)(-1)
$$
$$
1 = -a
$$

Thus, solving for $a$:

$$
a = -1
$$

#### Step 2: Write the final equation
Now that we know $a = -1$, we substitute this value into the factored equation:

$$
y = -(x - 2)(x - 4)
$$

Expanding the equation:

$$
y = -(x^2 - 6x + 8)
$$
$$
y = -x^2 + 6x - 8
$$

Thus, the equation of the parabola is:

$$
y = -x^2 + 6x - 8
$$


```
import numpy as np
import matplotlib.pyplot as plt

# Define the equation of the parabola: y = -x^2 + 6x - 8
def parabola(x):
    return -x**2 + 6*x - 8

# Create x values for plotting the parabola
x_values = np.linspace(0, 6, 400)
y_values = parabola(x_values)

# Create the plot
plt.figure(figsize=(6,6))

# Plot the parabola
plt.plot(x_values, y_values, label=r'$y = -x^2 + 6x - 8$', color='blue')

# Mark the roots (x = 2, x = 4) where the parabola intersects the x-axis
plt.scatter([2, 4], [0, 0], color='red', zorder=5, label='Roots (2, 0) and (4, 0)')

# Mark the point (3, 1) where the parabola passes through
plt.scatter(3, 1, color='green', zorder=5, label='Point (3, 1)')

# Plot the x-axis and y-axis
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Set equal scaling and limits for the axes
plt.axis('equal')
plt.xlim(0, 6)
plt.ylim(-10, 10)

# Add title, labels, and grid
plt.title('Parabola: $y = -x^2 + 6x - 8$', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

### Finding the Center of the Ellipse

The general equation of an ellipse is:

$$
Ax^2 + By^2 + Dx + Ey + F = 0
$$

We are given the equation:

$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$

#### Step 1: Group the terms with $x$ and $y$
We start by rearranging the terms in the equation:

$$
x^2 - 4x + 4y^2 - 16y = -16
$$

#### Step 2: Complete the square for $x$ and $y$
To complete the square, we need to add and subtract the appropriate constants inside the equation.

- For the $x$ terms: $x^2 - 4x$, the coefficient of $x$ is $-4$. Half of $-4$ is $-2$, and squaring $-2$ gives $4$. So, add and subtract $4$.
- For the $y$ terms: $4y^2 - 16y$, factor out the $4$ first: $4(y^2 - 4y)$. The coefficient of $y$ inside the parentheses is $-4$. Half of $-4$ is $-2$, and squaring $-2$ gives $4$. So, add and subtract $4$ inside the parentheses.

Thus, we have:

$$
(x^2 - 4x + 4) + 4(y^2 - 4y + 4) = -16 + 4 + 16
$$

#### Step 3: Simplify the equation
Simplifying the equation:

$$
(x - 2)^2 + 4(y - 2)^2 = 4
$$

#### Step 4: Write the equation in standard form
Now, the equation of the ellipse is:

$$
\frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1
$$

This is the standard form of an ellipse equation, where the center is given by \( (h, k) \).

#### Conclusion:
The center of the ellipse is at \( (2, 2) \).


```
import numpy as np
import matplotlib.pyplot as plt

# Ellipse parameters
h, k = 2, 2  # Center of the ellipse
a = 2  # Semi-major axis length
b = 1  # Semi-minor axis length

# Generate the angle values from 0 to 2 (for one full ellipse)
theta = np.linspace(0, 2 * np.pi, 400)

# Parametric equations for the ellipse
x = h + a * np.cos(theta)
y = k + b * np.sin(theta)

# Create the plot
plt.figure(figsize=(8, 8))

# Plot the ellipse using the parametric equations
plt.plot(x, y, label=r'$\frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1$', color='blue')

# Mark the center of the ellipse with a red dot
plt.scatter(h, k, color='red', zorder=5, label='Center (2, 2)', s=100)

# Plot the axes
plt.axhline(0, color='black', linewidth=0.5)  # x-axis
plt.axvline(0, color='black', linewidth=0.5)  # y-axis

# Adding dashed lines for the semi-major and semi-minor axes
plt.plot([h - a, h + a], [k, k], color='green', linestyle='--', label='Semi-major axis (length 2)')
plt.plot([h, h], [k - b, k + b], color='orange', linestyle='--', label='Semi-minor axis (length 1)')

# Set equal scaling for both axes to avoid distortion of the ellipse
plt.axis('equal')

# Set x and y limits to ensure the ellipse is properly displayed
plt.xlim(h - 3, h + 3)  # x-axis limit
plt.ylim(k - 2, k + 2)  # y-axis limit

# Title and labels
plt.title(r'Ellipse: $\frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1$', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

### Problem:

We are given the line equation:

$y = mx - 5$

and the equation of the circle:

$x^2 + y^2 = 1$

We are asked to find the slope $m > 0$ of the line that is tangent to the circle.

### Solution:

To solve this, we need to find the slope of the tangent line to the circle.

#### Steps:

1. **Substitute the equation of the line** $y = mx - 5$ into the circle equation $x^2 + y^2 = 1$.

$x^2 + (mx - 5)^2 = 1$

2. **Expand** the equation:

$x^2 + (m^2x^2 - 10mx + 25) = 1$

Simplifying:

$x^2 + m^2x^2 - 10mx + 25 = 1$

3. **Combine like terms**:

$(1 + m^2)x^2 - 10mx + 24 = 0$

4. For the line to be tangent to the circle, the quadratic equation must have exactly one solution. This means the discriminant must be zero:

$\Delta = b^2 - 4ac = 0$

For the equation $Ax^2 + Bx + C = 0$, where $A = 1 + m^2$, $B = -10m$, and $C = 24$, we have:

$\Delta = (-10m)^2 - 4(1 + m^2)(24) = 0$

5. **Solve for $m$**:

$100m^2 - 4(1 + m^2)(24) = 0$

$100m^2 - 96 - 96m^2 = 0$

$4m^2 = 96$

$m^2 = 24$

$m = \sqrt{24} = 2\sqrt{6}$

Thus, the slope of the tangent line is:

$m = 2\sqrt{6}$


```
import numpy as np
import matplotlib.pyplot as plt

# Circle parameters
circle_radius = 1
center = (0, 0)

# Tangent line parameters
m = 2 * np.sqrt(6)
b = -5  # y-intercept of the line

# Define the circle
theta = np.linspace(0, 2 * np.pi, 400)
x_circle = circle_radius * np.cos(theta)
y_circle = circle_radius * np.sin(theta)

# Define the tangent line
x_line = np.linspace(-2, 2, 400)
y_line = m * x_line - 5

# Plot the circle and the tangent line
plt.figure(figsize=(8, 8))

# Plot the circle
plt.plot(x_circle, y_circle, label=r'Circle: $x^2 + y^2 = 1$', color='blue')

# Plot the tangent line
plt.plot(x_line, y_line, label=r'Tangent line: $y = 2\sqrt{6}x - 5$', color='red', linestyle='--')

# Mark the point of tangency (approximately)
x_tangent = 1 / np.sqrt(7)  # x-coordinate of the tangency point
y_tangent = m * x_tangent - 5  # corresponding y-coordinate

plt.scatter(x_tangent, y_tangent, color='green', zorder=5, label='Point of Tangency')

# Mark the center of the circle
plt.scatter(center[0], center[1], color='purple', zorder=5, label='Center (0, 0)', s=100)

# Add axes
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add title, labels, and grid
plt.title('Circle and Tangent Line', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)

# Set equal aspect ratio
plt.axis('equal')

# Add the legend
plt.legend()

# Display the plot
plt.show()
```

### Problem:

We are given the following equations:

1. Hyperbola:

$x^2 - y^2 = 1$

2. Ellipse's equation:

$x^2 + 4y^2 = 6$

We are tasked with finding the intersection points of these curves.

### Solution:

To find the intersection points, we need to solve the system of equations.

#### Step 1: Solve for $x^2$ and $y^2$ from both equations.

From the hyperbola equation:

$x^2 = y^2 + 1$

#### Step 2: Substitute $x^2 = y^2 + 1$ into the ellipse equation:

Substitute this into the ellipse's equation $x^2 + 4y^2 = 6$:

$(y^2 + 1) + 4y^2 = 6$

#### Step 3: Simplify the equation:

$y^2 + 1 + 4y^2 = 6$

$5y^2 + 1 = 6$

$5y^2 = 5$

$y^2 = 1$

So, we have two possible values for $y$:

$y = 1$ or $y = -1$

#### Step 4: Solve for $x$.

Now that we have the values of $y$, we substitute these into the equation $x^2 = y^2 + 1$.

- When $y = 1$, we get:

$x^2 = 1^2 + 1 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}$

- When $y = -1$, we get:

$x^2 = (-1)^2 + 1 = 2 \quad \Rightarrow \quad x = \pm \sqrt{2}$

### Conclusion:

The intersection points of the hyperbola and the ellipse are:

$(\sqrt{2}, 1), \quad (-\sqrt{2}, 1), \quad (\sqrt{2}, -1), \quad (-\sqrt{2}, -1)$


```
import numpy as np
import matplotlib.pyplot as plt

# Define the equations of the hyperbola and ellipse
theta = np.linspace(0, 2 * np.pi, 400)

# Hyperbola equation: x^2 - y^2 = 1
# Parametric equations for the hyperbola
x_hyperbola_pos = np.cosh(theta)
y_hyperbola_pos = np.sinh(theta)

x_hyperbola_neg = -np.cosh(theta)
y_hyperbola_neg = np.sinh(theta)

# Ellipse equation: x^2 + 4y^2 = 6
# Parametric equations for the ellipse
x_ellipse = np.sqrt(6) * np.cos(theta)
y_ellipse = np.sqrt(6) * np.sin(theta) / 2

# Intersection points (calculated earlier)
intersection_points = [(-np.sqrt(2), 1), (np.sqrt(2), 1), (-np.sqrt(2), -1), (np.sqrt(2), -1)]

# Plot the curves and intersection points
plt.figure(figsize=(8, 8))

# Plot the hyperbola (positive and negative branches)
plt.plot(x_hyperbola_pos, y_hyperbola_pos, label=r'Hyperbola: $x^2 - y^2 = 1$', color='blue')
plt.plot(x_hyperbola_neg, y_hyperbola_neg, color='blue')

# Plot the ellipse
plt.plot(x_ellipse, y_ellipse, label=r'Ellipse: $x^2 + 4y^2 = 6$', color='red')

# Plot the intersection points
for point in intersection_points:
    plt.scatter(point[0], point[1], color='green', zorder=5)

# Add labels for the points
for point in intersection_points:
    plt.text(point[0] + 0.1, point[1] + 0.1, f'({point[0]:.2f}, {point[1]:.2f})', fontsize=12)

# Add title, labels, and grid
plt.title('Intersection of Hyperbola and Ellipse', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)

# Set equal aspect ratio
plt.axis('equal')

# Add the legend
plt.legend()

# Display the plot
plt.show()
```

### Problem:

We are given the equation of a hyperbola:

$x^2 - y^2 = 1$

We are tasked with finding the distance between the branches of the hyperbola.

### Solution:

The equation of the hyperbola $x^2 - y^2 = 1$ is a standard form of the hyperbola with its transverse axis along the x-axis. This can be rewritten as:

$\frac{x^2}{1^2} - \frac{y^2}{1^2} = 1$

where:
- The **center** of the hyperbola is at the origin $(0, 0)$,
- The **semi-major axis** (along the x-axis) has length $a = 1$,
- The **semi-minor axis** (along the y-axis) has length $b = 1$.

#### Step 1: The distance between the branches of a hyperbola along the transverse axis is given by:

The distance between the two branches at a specific value of $y$ can be found by considering the x-coordinates of the points where the hyperbola intersects the horizontal line at a given value of $y$. For a hyperbola $x^2 - y^2 = 1$, at a specific value of $y$, the x-coordinates of the points on the hyperbola are given by:

$x = \pm \sqrt{y^2 + 1}$

Thus, the distance between the branches at any value of $y$ is:

$d = 2 \sqrt{y^2 + 1}$

#### Step 2: Finding the minimum distance:

The minimum distance between the branches occurs when $y = 0$ because this is when the branches are closest. Substituting $y = 0$ into the formula:

$d = 2 \sqrt{0^2 + 1} = 2 \sqrt{1} = 2$

### Conclusion:

The distance between the branches of the hyperbola $x^2 - y^2 = 1$ is **2** units at $y = 0$, which is the minimum distance between the two branches.


```
import numpy as np
import matplotlib.pyplot as plt

# Define the parametric equations for the hyperbola: x^2 - y^2 = 1
theta = np.linspace(0, 2 * np.pi, 400)

# Positive branch of the hyperbola
x_positive = np.cosh(theta)
y_positive = np.sinh(theta)

# Negative branch of the hyperbola
x_negative = -np.cosh(theta)
y_negative = np.sinh(theta)

# Points for the minimum distance (y = 0, x = +-sqrt(2))
x_intersection = np.array([np.sqrt(2), -np.sqrt(2)])
y_intersection = np.array([0, 0])

# Plot the hyperbola
plt.figure(figsize=(8, 8))

# Plot positive and negative branches of the hyperbola
plt.plot(x_positive, y_positive, label=r'Hyperbola: $x^2 - y^2 = 1$', color='blue')
plt.plot(x_negative, y_negative, color='blue')

# Plot the minimum distance points at y = 0
plt.scatter(x_intersection, y_intersection, color='green', zorder=5)
plt.text(np.sqrt(2) + 0.1, 0.1, f'Minimum distance: {2:.2f}', fontsize=12)

# Plot a horizontal line at y = 0
plt.axhline(y=0, color='gray', linestyle='--')

# Add title, labels, and grid
plt.title('Hyperbola and Distance Between Branches', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)

# Set equal aspect ratio
plt.axis('equal')

# Add the legend
plt.legend()

# Display the plot
plt.show()
```
