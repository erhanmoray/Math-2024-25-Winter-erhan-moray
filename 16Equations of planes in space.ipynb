### Equations of Planes in Space

In three-dimensional space, the general equation of a plane can be represented in the following forms:

#### 1. **General Form**:
The general equation of a plane is:

$Ax + By + Cz + D = 0$

Where:
- $A$, $B$, and $C$ are the components of the normal vector to the plane, denoted by $\mathbf{n} = (A, B, C)$,
- $D$ is a constant that determines the position of the plane relative to the origin,
- $x$, $y$, and $z$ are the coordinates of any point on the plane.

#### 2. **Point-Normal Form**:
If a plane passes through a point $P_0(x_0, y_0, z_0)$ and has a normal vector $\mathbf{n} = (A, B, C)$, the equation of the plane can be written as:

$A(x - x_0) + B(y - y_0) + C(z - z_0) = 0$

Where:
- $(x_0, y_0, z_0)$ is the known point on the plane,
- $(x, y, z)$ are the coordinates of any point on the plane.

#### 3. **Intercept Form**:
For a plane that intersects the coordinate axes at points $A(a, 0, 0)$, $B(0, b, 0)$, and $C(0, 0, c)$, the equation of the plane can be expressed as:

$\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1$

Where:
- $a$, $b$, and $c$ are the intercepts of the plane with the x-, y-, and z-axes, respectively.

#### 4. **Parametric Form**:
The parametric equation of a plane can be given using two vectors $\mathbf{v}_1$ and $\mathbf{v}_2$ that lie on the plane, and a point $P_0(x_0, y_0, z_0)$ on the plane:

$\mathbf{r}(u, v) = \mathbf{P_0} + u\mathbf{v_1} + v\mathbf{v_2}$

Where:
- $\mathbf{r}(u, v)$ represents any point on the plane,
- $u$ and $v$ are parameters that vary over the plane,
- $\mathbf{P_0}$ is a point on the plane,
- $\mathbf{v_1}$ and $\mathbf{v_2}$ are direction vectors that lie on the plane.

### Conclusion:

The equation of a plane in space can be expressed in several forms, depending on the available information. The most common forms are the general form, the point-normal form, the intercept form, and the parametric form.


### Problem:
Find the equation of the plane passing through the points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$.

### Solution:

The equation of a plane can be written in the point-normal form:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

Where:
- $(x_0, y_0, z_0)$ is a point on the plane (we can use any of the given points),
- $A$, $B$, and $C$ are the components of the normal vector to the plane.

#### Step 1: Find the normal vector

To find the normal vector $\mathbf{n} = (A, B, C)$, we first find two vectors that lie on the plane by subtracting the coordinates of the given points:

$$
\mathbf{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2)
$$

$$
\mathbf{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1)
$$

Now, we find the cross product of $\mathbf{AB}$ and $\mathbf{AC}$ to get the normal vector $\mathbf{n}$:

$$
\mathbf{n} = \mathbf{AB} \times \mathbf{AC} = \begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
2 & 2 & 2 \\
1 & -1 & 1
\end{vmatrix}
$$

Expanding the determinant:

$$
\mathbf{n} = \hat{i} \begin{vmatrix} 2 & 2 \\ -1 & 1 \end{vmatrix} - \hat{j} \begin{vmatrix} 2 & 2 \\ 1 & 1 \end{vmatrix} + \hat{k} \begin{vmatrix} 2 & 2 \\ 1 & -1 \end{vmatrix}
$$

$$
\mathbf{n} = \hat{i} \cdot (2 \cdot 1 - 2 \cdot -1) - \hat{j} \cdot (2 \cdot 1 - 2 \cdot 1) + \hat{k} \cdot (2 \cdot -1 - 2 \cdot 1)
$$

$$
\mathbf{n} = \hat{i} \cdot (2 + 2) - \hat{j} \cdot (2 - 2) + \hat{k} \cdot (-2 - 2)
$$

$$
\mathbf{n} = \hat{i} \cdot 4 - \hat{j} \cdot 0 + \hat{k} \cdot (-4)
$$

$$
\mathbf{n} = (4, 0, -4)
$$

Thus, the normal vector to the plane is $\mathbf{n} = (4, 0, -4)$.

#### Step 2: Equation of the plane

Now that we have the normal vector $\mathbf{n} = (4, 0, -4)$ and the point $A(1, 2, 3)$, we can substitute them into the point-normal form of the plane equation:

$$
4(x - 1) + 0(y - 2) - 4(z - 3) = 0
$$

Simplifying:

$$
4(x - 1) - 4(z - 3) = 0
$$

$$
4x - 4 - 4z + 12 = 0
$$

$$
4x - 4z + 8 = 0
$$

Dividing through by 4:

$$
x - z + 2 = 0
$$

Thus, the equation of the plane is:

$$
x - z + 2 = 0
$$

### Conclusion:
The equation of the plane passing through the points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$ is:

$$
x - z + 2 = 0
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Points A, B, and C
A = np.array([1, 2, 3])
B = np.array([3, 4, 5])
C = np.array([2, 1, 4])

# Vectors AB and AC
AB = B - A
AC = C - A

# Normal vector to the plane (cross product of AB and AC)
normal_vector = np.cross(AB, AC)

# Define the equation of the plane: Ax + By + Cz + D = 0
A_val, B_val, C_val = normal_vector
D_val = -np.dot(normal_vector, A)  # Calculate D using point A

# Create a grid for plotting the plane
x = np.linspace(-1, 5, 10)
y = np.linspace(-1, 5, 10)
X, Y = np.meshgrid(x, y)

# Solve for Z in terms of X and Y from the plane equation
Z = (-A_val * X - B_val * Y - D_val) / C_val

# Plot the points A, B, and C
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='blue')

# Plot the points A, B, C
ax.scatter(A[0], A[1], A[2], color='r', s=100, label='A(1,2,3)')
ax.scatter(B[0], B[1], B[2], color='g', s=100, label='B(3,4,5)')
ax.scatter(C[0], C[1], C[2], color='b', s=100, label='C(2,1,4)')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane Passing Through Points A, B, and C')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

### Problem:
Find the equation of the plane passing through the point \( A(1, 2, 3) \) and parallel to the plane \( 2x + 3y + 4z = 5 \).

### Solution:

When two planes are parallel, they have the same normal vector.

#### Step 1: Identify the normal vector of the given plane

The equation of the given plane is:

$$
2x + 3y + 4z = 5
$$

The normal vector to the plane is the vector of coefficients of \( x \), \( y \), and \( z \). Hence, the normal vector for this plane is:

$$
\mathbf{n} = (2, 3, 4)
$$

#### Step 2: Use the normal vector for the new plane

Since the new plane is parallel to the given plane, it will have the same normal vector \( \mathbf{n} = (2, 3, 4) \).

#### Step 3: Equation of the new plane

The equation of a plane can be written in point-normal form as:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

Where:
- \( (x_0, y_0, z_0) \) is the point on the plane (in this case, \( A(1, 2, 3) \)),
- \( A, B, C \) are the components of the normal vector to the plane (in this case, \( (2, 3, 4) \)).

Substitute the values into the equation:

$$
2(x - 1) + 3(y - 2) + 4(z - 3) = 0
$$

Simplifying:

$$
2(x - 1) + 3(y - 2) + 4(z - 3) = 0
$$

$$
2x - 2 + 3y - 6 + 4z - 12 = 0
$$

$$
2x + 3y + 4z - 20 = 0
$$

Thus, the equation of the plane passing through \( A(1, 2, 3) \) and parallel to the plane \( 2x + 3y + 4z = 5 \) is:

$$
2x + 3y + 4z = 20
$$

### Conclusion:
The equation of the plane passing through the point \( A(1, 2, 3) \) and parallel to the plane \( 2x + 3y + 4z = 5 \) is:

$$
2x + 3y + 4z = 20
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the normal vector and point A
normal_vector = np.array([2, 3, 4])
point_A = np.array([1, 2, 3])

# Create a grid for plotting the planes
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Equation of the planes: 2x + 3y + 4z = 20 (new plane), 2x + 3y + 4z = 5 (given plane)
# Solve for Z in terms of X and Y from the plane equation: Z = (20 - 2x - 3y) / 4
Z_new = (20 - 2*X - 3*Y) / 4
Z_given = (5 - 2*X - 3*Y) / 4

# Plot the planes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the new plane (parallel plane)
ax.plot_surface(X, Y, Z_new, alpha=0.5, rstride=100, cstride=100, color='blue', label='Plane: 2x + 3y + 4z = 20')

# Plot the given plane
ax.plot_surface(X, Y, Z_given, alpha=0.5, rstride=100, cstride=100, color='red', label='Plane: 2x + 3y + 4z = 5')

# Plot the point A(1, 2, 3)
ax.scatter(point_A[0], point_A[1], point_A[2], color='green', s=100, label='Point A(1,2,3)')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Planes Parallel to Each Other and Point A')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

### Problem:
Find the line of intersection of the planes \( 2x + 3y + 4z = 5 \) and \( 3x + 4y + 2z = 6 \).

### Solution:

To find the line of intersection of two planes, we need to solve the system of equations that represent the planes. The two equations are:

1. \( 2x + 3y + 4z = 5 \)
2. \( 3x + 4y + 2z = 6 \)

The line of intersection is the set of points that satisfy both equations simultaneously.

#### Step 1: Solve the system of equations

We will solve these two equations by eliminating one of the variables.

- Multiply the first equation by 3:
  $$
  6x + 9y + 12z = 15
  $$

- Multiply the second equation by 2:
  $$
  6x + 8y + 4z = 12
  $$

Now subtract the second equation from the first equation:

$$
(6x + 9y + 12z) - (6x + 8y + 4z) = 15 - 12
$$

Simplifying:

$$
y + 8z = 3
$$

This gives us the relationship between \( y \) and \( z \):
$$
y = 3 - 8z
$$

#### Step 2: Substitute into one of the original equations

Substitute \( y = 3 - 8z \) into the first equation \( 2x + 3y + 4z = 5 \):

$$
2x + 3(3 - 8z) + 4z = 5
$$

Simplifying:

$$
2x + 9 - 24z + 4z = 5
$$

$$
2x - 20z = -4
$$

Solving for \( x \):

$$
x = 10z - 2
$$

#### Step 3: Parametric form of the line

Now we have the parametric equations for the line of intersection:

- \( x = 10z - 2 \)
- \( y = 3 - 8z \)
- \( z = z \)

Thus, the parametric form of the line is:

$$
x = 10z - 2
$$
$$
y = 3 - 8z
$$
$$
z = z
$$

### Conclusion:
The line of intersection of the planes \( 2x + 3y + 4z = 5 \) and \( 3x + 4y + 2z = 6 \) is given by the parametric equations:

$$
x = 10z - 2, \quad y = 3 - 8z, \quad z = z
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid for plotting
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Equation of the first plane: 2x + 3y + 4z = 5
# Solve for Z: Z = (5 - 2x - 3y) / 4
Z1 = (5 - 2*X - 3*Y) / 4

# Equation of the second plane: 3x + 4y + 2z = 6
# Solve for Z: Z = (6 - 3x - 4y) / 2
Z2 = (6 - 3*X - 4*Y) / 2

# Create the figure for plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the first plane
ax.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100, color='blue', label='Plane 1: 2x + 3y + 4z = 5')

# Plot the second plane
ax.plot_surface(X, Y, Z2, alpha=0.5, rstride=100, cstride=100, color='red', label='Plane 2: 3x + 4y + 2z = 6')

# Plot the line of intersection of the planes
z_values = np.linspace(-5, 5, 100)
x_values = 10 * z_values - 2
y_values = 3 - 8 * z_values

# Plot the line in 3D
ax.plot(x_values, y_values, z_values, color='green', label='Line of Intersection')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Planes and Line of Intersection')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

### Problem:
Find the equation of the plane passing through point \( A(1, 2, 3) \) and parallel to the vectors \( \mathbf{v}_1 = [1, 0, 1] \) and \( \mathbf{v}_2 = [0, 1, -1] \).

### Solution:

The equation of a plane in 3D space can be written in the point-normal form:

$$
A(x - x_0) + B(y - y_0) + C(z - z_0) = 0
$$

Where:
- \( (x_0, y_0, z_0) \) is a point on the plane (in this case, \( A(1, 2, 3) \)),
- \( (A, B, C) \) is the normal vector to the plane.

#### Step 1: Find the normal vector to the plane

The normal vector \( \mathbf{n} \) of the plane can be found by taking the cross product of the two given vectors \( \mathbf{v}_1 = [1, 0, 1] \) and \( \mathbf{v}_2 = [0, 1, -1] \):

$$
\mathbf{n} = \mathbf{v}_1 \times \mathbf{v}_2
$$

We calculate the cross product:

$$
\mathbf{n} = \begin{vmatrix} \hat{i} & \hat{j} & \hat{k} \\ 1 & 0 & 1 \\ 0 & 1 & -1 \end{vmatrix}
$$

Expanding the determinant:

$$
\mathbf{n} = \hat{i}(0 \times -1 - 1 \times 1) - \hat{j}(1 \times -1 - 1 \times 0) + \hat{k}(1 \times 1 - 0 \times 0)
$$

Simplifying:

$$
\mathbf{n} = \hat{i}(-1) - \hat{j}(-1) + \hat{k}(1)
$$

Thus, the normal vector is:

$$
\mathbf{n} = [-1, 1, 1]
$$

#### Step 2: Write the equation of the plane

Now that we have the normal vector \( \mathbf{n} = [-1, 1, 1] \) and the point \( A(1, 2, 3) \), we can substitute these values into the point-normal form of the plane equation:

$$
-1(x - 1) + 1(y - 2) + 1(z - 3) = 0
$$

Expanding:

$$
-x + 1 + y - 2 + z - 3 = 0
$$

Simplifying:

$$
-x + y + z - 4 = 0
$$

Or equivalently:

$$
x - y - z = -4
$$

### Conclusion:
The equation of the plane passing through \( A(1, 2, 3) \) and parallel to the vectors \( \mathbf{v}_1 = [1, 0, 1] \) and \( \mathbf{v}_2 = [0, 1, -1] \) is:

$$
x - y - z = -4
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the point A(1, 2, 3)
point = np.array([1, 2, 3])

# Define the vectors v1 = [1, 0, 1] and v2 = [0, 1, -1]
v1 = np.array([1, 0, 1])
v2 = np.array([0, 1, -1])

# Compute the normal vector by taking the cross product of v1 and v2
normal_vector = np.cross(v1, v2)

# Define the grid for plotting
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Equation of the plane: x - y - z = -4
# Solve for Z: z = x - y + 4
Z = X - Y + 4

# Create the figure for plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='blue', label='Plane: x - y - z = -4')

# Plot the point A(1, 2, 3)
ax.scatter(point[0], point[1], point[2], color='red', s=100, label='Point A(1, 2, 3)')

# Plot the vectors v1 and v2 from point A
ax.quiver(point[0], point[1], point[2], v1[0], v1[1], v1[2], color='green', length=2, normalize=True, label='Vector v1')
ax.quiver(point[0], point[1], point[2], v2[0], v2[1], v2[2], color='orange', length=2, normalize=True, label='Vector v2')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane Passing Through Point A and Parallel to Vectors v1 and v2')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

### Problem:
We have the plane \( 2x + 3y + 4z = 5 \). Find an example of a plane parallel and perpendicular to it.

### Solution:

#### 1. Plane Parallel to the Given Plane:

Two planes are parallel if they have the same normal vector. The normal vector of the plane \( 2x + 3y + 4z = 5 \) is given by \( \mathbf{n} = [2, 3, 4] \).

So, any plane parallel to this one will have the same normal vector, but the constant term will be different.

For example, the plane:

$$
2x + 3y + 4z = 10
$$

is parallel to the given plane because it has the same normal vector \( \mathbf{n} = [2, 3, 4] \).

#### 2. Plane Perpendicular to the Given Plane:

Two planes are perpendicular if the dot product of their normal vectors is zero. To find a perpendicular plane, we need a normal vector that is orthogonal to the normal vector of the given plane \( \mathbf{n} = [2, 3, 4] \).

Lets pick a vector \( \mathbf{m} = [a, b, c] \) such that the dot product between \( \mathbf{n} \) and \( \mathbf{m} \) is zero:

$$
2a + 3b + 4c = 0
$$

One possible solution is to take \( a = 1 \), \( b = -1 \), and \( c = 0 \). This gives the normal vector \( \mathbf{m} = [1, -1, 0] \).

So, a plane perpendicular to the given plane can have the equation:

$$
x - y = 1
$$

This plane has the normal vector \( \mathbf{m} = [1, -1, 0] \), which is orthogonal to the normal vector \( \mathbf{n} = [2, 3, 4] \) of the given plane.

### Conclusion:

- A plane parallel to \( 2x + 3y + 4z = 5 \) is \( 2x + 3y + 4z = 10 \).
- A plane perpendicular to \( 2x + 3y + 4z = 5 \) is \( x - y = 1 \).


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create grid
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Given planes equations
Z1 = (5 - 2*X - 3*Y) / 4  # Plane 2x + 3y + 4z = 5
Z2 = (10 - 2*X - 3*Y) / 4  # Parallel plane 2x + 3y + 4z = 10
Z3 = X - Y - 1  # Perpendicular plane x - y = 1

# Create the figure for plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the planes
ax.plot_surface(X, Y, Z1, alpha=0.5, rstride=100, cstride=100, color='blue', label='Plane: 2x + 3y + 4z = 5')
ax.plot_surface(X, Y, Z2, alpha=0.5, rstride=100, cstride=100, color='green', label='Plane: 2x + 3y + 4z = 10')
ax.plot_surface(X, Y, Z3, alpha=0.5, rstride=100, cstride=100, color='red', label='Plane: x - y = 1')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Planes: Given, Parallel, and Perpendicular')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

### Problem:
We have the plane \( 2x + 3y + 4z = 5 \) and the point \( A(1, 2, 3) \). Find the distance from point \( A \) to this plane.

### Solution:

#### Formula for Distance from a Point to a Plane:

The formula to calculate the distance \( d \) from a point \( (x_1, y_1, z_1) \) to a plane \( Ax + By + Cz + D = 0 \) is:

$$
d = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}}
$$

#### Given:

- The equation of the plane is \( 2x + 3y + 4z = 5 \), so we can rewrite it as:

  $$ 2x + 3y + 4z - 5 = 0 $$

  Therefore, \( A = 2 \), \( B = 3 \), \( C = 4 \), and \( D = -5 \).

- The coordinates of point \( A \) are \( (1, 2, 3) \).

#### Applying the Formula:

Substitute \( A = 2 \), \( B = 3 \), \( C = 4 \), \( D = -5 \), and the coordinates of point \( A(1, 2, 3) \) into the formula:

$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}
$$

Simplifying the numerator:

$$
d = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}}
$$

$$
d = \frac{|15|}{\sqrt{29}}
$$

Thus, the distance from point \( A(1, 2, 3) \) to the plane \( 2x + 3y + 4z = 5 \) is:

$$
d = \frac{15}{\sqrt{29}}
$$

This is the distance from point \( A \) to the given plane.


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the plane equation
def plane_eq(x, y):
    return (5 - 2*x - 3*y) / 4

# Create grid for x and y values
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)
Z = plane_eq(X, Y)

# Coordinates of point A
A = np.array([1, 2, 3])

# Calculate the normal vector to the plane
normal_vector = np.array([2, 3, 4])

# Calculate the distance from point A to the plane
numerator = abs(np.dot(normal_vector, A) - 5)
denominator = np.linalg.norm(normal_vector)
distance = numerator / denominator

# Plot the plane and point A
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface (plane)
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='blue', label='Plane: 2x + 3y + 4z = 5')

# Plot point A
ax.scatter(A[0], A[1], A[2], color='red', s=100, label=f'Point A(1, 2, 3)')

# Plot the perpendicular from point A to the plane
point_on_plane = np.array([A[0], A[1], plane_eq(A[0], A[1])])
ax.plot([A[0], point_on_plane[0]], [A[1], point_on_plane[1]], [A[2], point_on_plane[2]], color='green', label=f'Distance = {distance:.2f}')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Distance from Point A to Plane')

# Show legend
ax.legend()

# Show the plot
plt.show()

# Output the distance
print(f"The distance from point A(1, 2, 3) to the plane is: {distance:.2f}")
```

### Problem:
We have a plane that intersects the coordinate axes at points \( A(2, 0, 0) \), \( B(0, 3, 0) \), and \( C(0, 0, 4) \). Find the equation of the plane.

### Solution:

#### General Equation of the Plane:
The general equation of a plane in 3D space is given by:

$$
\frac{x}{a} + \frac{y}{b} + \frac{z}{c} = 1
$$

where \( a \), \( b \), and \( c \) are the intercepts of the plane with the coordinate axes.

#### Intercepts of the Plane:
- The point \( A(2, 0, 0) \) gives the intercept along the \( x \)-axis, so \( a = 2 \).
- The point \( B(0, 3, 0) \) gives the intercept along the \( y \)-axis, so \( b = 3 \).
- The point \( C(0, 0, 4) \) gives the intercept along the \( z \)-axis, so \( c = 4 \).

Thus, the equation of the plane is:

$$
\frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1
$$

#### Final Equation:
The equation of the plane is:

$$
\frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid for x, y
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Define the equation of the plane: (x/2) + (y/3) + (z/4) = 1
# Solving for Z in terms of X and Y
Z = 4 * (1 - X/2 - Y/3)

# Coordinates of points A, B, C
A = np.array([2, 0, 0])
B = np.array([0, 3, 0])
C = np.array([0, 0, 4])

# Plotting the plane and points
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface (plane)
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='blue', label=r'$\frac{x}{2} + \frac{y}{3} + \frac{z}{4} = 1$')

# Plot the points A, B, C
ax.scatter(A[0], A[1], A[2], color='red', s=100, label='Point A (2, 0, 0)')
ax.scatter(B[0], B[1], B[2], color='green', s=100, label='Point B (0, 3, 0)')
ax.scatter(C[0], C[1], C[2], color='yellow', s=100, label='Point C (0, 0, 4)')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane Intersection with Coordinate Axes')

# Show legend
ax.legend()

# Show the plot
plt.show()
```

### Problem:
We are asked to calculate the angle between the plane \( x + y + z = 1 \) and the plane \( x = 0 \) (the \( yz \)-plane).

### Solution:

#### Step 1: The Normal Vectors of the Planes
- The equation of a plane is generally given by \( Ax + By + Cz = D \), where the vector \( \vec{n} = [A, B, C] \) is the normal vector of the plane.
  
- For the plane \( x + y + z = 1 \), the normal vector is:
  $$
  \vec{n_1} = [1, 1, 1]
  $$

- For the plane \( x = 0 \) (the \( yz \)-plane), the normal vector is:
  $$
  \vec{n_2} = [1, 0, 0]
  $$

#### Step 2: Calculate the Angle Between the Normal Vectors
The angle \( \theta \) between two vectors \( \vec{n_1} \) and \( \vec{n_2} \) can be calculated using the following formula:
$$
\cos \theta = \frac{\vec{n_1} \cdot \vec{n_2}}{\|\vec{n_1}\| \|\vec{n_2}\|}
$$

Where:
- \( \vec{n_1} \cdot \vec{n_2} \) is the dot product of the vectors.
- \( \|\vec{n_1}\| \) and \( \|\vec{n_2}\| \) are the magnitudes (norms) of the vectors.

The dot product is calculated as:
$$
\vec{n_1} \cdot \vec{n_2} = 1 \cdot 1 + 1 \cdot 0 + 1 \cdot 0 = 1
$$

The magnitudes of the normal vectors are:
$$
\|\vec{n_1}\| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}
$$
$$
\|\vec{n_2}\| = \sqrt{1^2 + 0^2 + 0^2} = 1
$$

Thus:
$$
\cos \theta = \frac{1}{\sqrt{3} \times 1} = \frac{1}{\sqrt{3}}
$$

#### Step 3: Find the Angle \( \theta \)
We can find the angle \( \theta \) by taking the inverse cosine:
$$
\theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right)
$$

#### Final Answer:
The angle \( \theta \) between the plane \( x + y + z = 1 \) and the \( yz \)-plane \( x = 0 \) is:
$$
\theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right) \approx 54.74^\circ
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the plane x + y + z = 1
def plane1(x, y):
    return 1 - x - y

# Create grid for x and y
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Z values for the plane x + y + z = 1
Z1 = plane1(X, Y)

# Plotting the planes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot plane x + y + z = 1
ax.plot_surface(X, Y, Z1, alpha=0.5, color='blue', label=r'$x + y + z = 1$')

# Plane x = 0 (yz-plane)
ax.plot_surface(np.zeros_like(X), Y, Z1, alpha=0.5, color='green', label=r'$x = 0$ (yz-plane)')

# Axes Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Planes: $x + y + z = 1$ and $x = 0$ (yz-plane)')

# Display legend
ax.legend()

# Show the plot
plt.show()
```

### Problem:
We are asked to find a vector perpendicular to the plane \( x + y + z = 1 \).

### Solution:

#### Step 1: The Equation of a Plane
The general equation of a plane is given by:
$$
Ax + By + Cz = D
$$
Where the vector \( \vec{n} = [A, B, C] \) is the normal vector to the plane.

#### Step 2: Identify the Normal Vector
For the plane \( x + y + z = 1 \), we can compare this equation with the general form \( Ax + By + Cz = D \). Here, we have:
- \( A = 1 \)
- \( B = 1 \)
- \( C = 1 \)

Thus, the normal vector \( \vec{n} \) to the plane is:
$$
\vec{n} = [1, 1, 1]
$$

#### Final Answer:
The vector perpendicular to the plane \( x + y + z = 1 \) is:
$$
\vec{n} = [1, 1, 1]
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the plane equation x + y + z = 1
def plane(x, y):
    return 1 - x - y

# Create grid for x and y
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)

# Z values for the plane x + y + z = 1
Z = plane(X, Y)

# Normal vector to the plane
normal_vector = np.array([1, 1, 1])

# Create a point on the plane (e.g., x = 0, y = 0, z = 1)
point_on_plane = np.array([0, 0, 1])

# Plotting the plane and normal vector
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the plane
ax.plot_surface(X, Y, Z, alpha=0.5, color='blue', label=r'$x + y + z = 1$')

# Plot the normal vector
ax.quiver(point_on_plane[0], point_on_plane[1], point_on_plane[2],
          normal_vector[0], normal_vector[1], normal_vector[2],
          length=2, color='red', label='Normal vector [1, 1, 1]')

# Set labels for axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plane: $x + y + z = 1$ and Normal Vector [1, 1, 1]')

# Show the legend
ax.legend()

# Show the plot
plt.show()
```
