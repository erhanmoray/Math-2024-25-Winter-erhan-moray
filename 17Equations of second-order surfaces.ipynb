### Equations of Second-Order Surfaces

Second-order surfaces, also known as **quadratic surfaces**, are represented by equations that are quadratic in the variables \(x\), \(y\), and \(z\). These surfaces can be classified into various types based on the form of their equation.

#### 1. **Ellipsoid**
The equation of an ellipsoid is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1
$$
Where:
- \(a\), \(b\), and \(c\) are the semi-axes along the \(x\), \(y\), and \(z\)-axes, respectively.

#### 2. **Hyperboloid of One Sheet**
The equation of a hyperboloid of one sheet is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1
$$
Where:
- \(a\), \(b\), and \(c\) are constants that determine the shape of the hyperboloid.

#### 3. **Hyperboloid of Two Sheets**
The equation of a hyperboloid of two sheets is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1
$$
Where:
- \(a\), \(b\), and \(c\) are constants.

#### 4. **Elliptic Paraboloid**
The equation of an elliptic paraboloid is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} = z
$$
Where:
- \(a\) and \(b\) control the curvature along the \(x\)- and \(y\)-axes.

#### 5. **Hyperbolic Paraboloid**
The equation of a hyperbolic paraboloid is:
$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = z
$$
Where:
- \(a\) and \(b\) control the curvature along the \(x\)- and \(y\)-axes.

#### 6. **Cone**
The equation of a cone is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} = \frac{z^2}{c^2}
$$
Where:
- \(a\), \(b\), and \(c\) determine the cone's dimensions and orientation.

#### 7. **Elliptic Cylinder**
The equation of an elliptic cylinder is:
$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1
$$
Where:
- The cylinder extends infinitely along the \(z\)-axis.

#### 8. **Hyperbolic Cylinder**
The equation of a hyperbolic cylinder is:
$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$$
Where:
- The cylinder extends infinitely along the \(z\)-axis.

#### 9. **Paraboloid of Revolution**
The equation of a paraboloid of revolution is:
$$
z = \frac{x^2}{a^2} + \frac{y^2}{b^2}
$$
Where:
- \(a\) and \(b\) determine the curvature of the paraboloid.

#### 10. **Cylinder**
The equation of a cylinder is:
$$
x^2 + y^2 = r^2
$$
Where:
- \(r\) is the radius of the cylinder, and the cylinder extends infinitely along the \(z\)-axis.

### Summary
Second-order surfaces are widely used in geometry and physics to describe various natural and man-made objects. Their classification into ellipsoids, hyperboloids, paraboloids, and cylinders allows for detailed analysis and modeling of these shapes.


### Equation of a Sphere

The **equation of a sphere** in 3D space is based on the **distance formula**, which calculates the distance between any point on the sphere and its center. A sphere is the set of all points that are equidistant from a fixed point, called the center. The general equation of a sphere with center \( (x_0, y_0, z_0) \) and radius \( r \) is:

$$
(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2
$$

Where:
- \( (x, y, z) \) represents any point on the surface of the sphere,
- \( (x_0, y_0, z_0) \) is the center of the sphere,
- \( r \) is the radius of the sphere, which is the constant distance from any point on the sphere to the center.

### Step-by-step process to find the equation of the sphere

#### Given:
- **Center of the sphere**: \( P = (1, 2, 3) \)
- **Radius**: \( r = 3 \)

#### Step 1: General Equation of a Sphere
The general form of the equation of a sphere is:

$$
(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2
$$

This equation is derived from the **distance formula** in 3D, which states that the distance between a point \( (x, y, z) \) and the center \( (x_0, y_0, z_0) \) is:

$$
d = \sqrt{(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2}
$$

Since the distance from the center to any point on the sphere is equal to the radius \( r \), we set \( d = r \) and square both sides of the equation to get the general form.

#### Step 2: Substitute the Given Values
Substitute the values of the center \( P = (1, 2, 3) \) and radius \( r = 3 \) into the general equation:

- \( x_0 = 1 \)
- \( y_0 = 2 \)
- \( z_0 = 3 \)
- \( r = 3 \)

This results in:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 3^2
$$

#### Step 3: Simplify the Equation
Simplify the right-hand side:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

Thus, the equation of the sphere with center \( (1, 2, 3) \) and radius 3 is:

$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

### Explanation:
- The terms \( (x - 1)^2 \), \( (y - 2)^2 \), and \( (z - 3)^2 \) represent the squared differences between the coordinates of any point \( (x, y, z) \) on the sphere and the center of the sphere \( (1, 2, 3) \).
- The right side of the equation is \( 9 \), which is the square of the radius \( r = 3 \).
- The equation represents all the points \( (x, y, z) \) that are at a distance of 3 units from the center \( (1, 2, 3) \).

This is the **standard form** of the equation of a sphere in 3D space.


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given center and radius of the sphere
center = np.array([1, 2, 3])  # Center of the sphere at (1, 2, 3)
radius = 3  # Radius of the sphere

# Create a grid of values for the angles theta (polar) and phi (azimuthal)
theta = np.linspace(0, np.pi, 100)  # Polar angle from 0 to pi
phi = np.linspace(0, 2 * np.pi, 100)  # Azimuthal angle from 0 to 2pi

# Generate meshgrid for theta and phi
theta, phi = np.meshgrid(theta, phi)

# Parametric equations for the sphere using spherical coordinates
x = center[0] + radius * np.sin(theta) * np.cos(phi)
y = center[1] + radius * np.sin(theta) * np.sin(phi)
z = center[2] + radius * np.cos(theta)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the sphere
ax.plot_surface(x, y, z, color='b', alpha=0.6, rstride=5, cstride=5)

# Label the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Title of the plot
ax.set_title('3D Visualization of Sphere: Center (1, 2, 3), Radius = 3')

# Show the plot
plt.show()
```

### Problem:
Do the spheres with equations \(x^2 + y^2 + z^2 = 1\) and \(x^2 + y^2 + z^2 = 2\) have any common points?

### Geometric Analysis:
- **Equation of Sphere 1**: \(x^2 + y^2 + z^2 = 1\)
  - This is the equation of a sphere with center at the origin \( (0, 0, 0) \) and radius \(1\).

- **Equation of Sphere 2**: \(x^2 + y^2 + z^2 = 2\)
  - This is the equation of a sphere with center at the origin \( (0, 0, 0) \) and radius \( \sqrt{2} \).

### Intersection of Spheres:
- Since both spheres have the same center but different radii (\(1\) and \( \sqrt{2} \)), they are concentric spheres.
- **Concentric spheres** do not intersect, as one sphere is entirely inside the other and there is no common point.

### Conclusion:
- The spheres with the equations \(x^2 + y^2 + z^2 = 1\) and \(x^2 + y^2 + z^2 = 2\) do **not** intersect because they are concentric with different radii.


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters for the spheres
center = np.array([0, 0, 0])  # Both spheres have the same center at the origin

# Radii of the spheres
radius1 = 1  # Radius of the first sphere
radius2 = np.sqrt(2)  # Radius of the second sphere

# Create a grid of points (theta and phi) for parametric equation of spheres
theta = np.linspace(0, np.pi, 100)  # Polar angle from 0 to pi
phi = np.linspace(0, 2 * np.pi, 100)  # Azimuthal angle from 0 to 2pi
theta, phi = np.meshgrid(theta, phi)

# Parametric equations for the two spheres
# Sphere 1: x^2 + y^2 + z^2 = 1
x1 = center[0] + radius1 * np.sin(theta) * np.cos(phi)
y1 = center[1] + radius1 * np.sin(theta) * np.sin(phi)
z1 = center[2] + radius1 * np.cos(theta)

# Sphere 2: x^2 + y^2 + z^2 = 2
x2 = center[0] + radius2 * np.sin(theta) * np.cos(phi)
y2 = center[1] + radius2 * np.sin(theta) * np.sin(phi)
z2 = center[2] + radius2 * np.cos(theta)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the two spheres
ax.plot_surface(x1, y1, z1, color='blue', alpha=0.6, rstride=5, cstride=5, label='Sphere 1: x + y + z = 1')
ax.plot_surface(x2, y2, z2, color='red', alpha=0.6, rstride=5, cstride=5, label='Sphere 2: x + y + z = 2')

# Label the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Title of the plot
ax.set_title('Visualization of Two Spheres')

# Show the plot
plt.show()
```

### Problem:
We are asked to find the curve formed by the intersection of two spheres:
1. The first sphere with the equation:
   $$
   x^2 + y^2 + z^2 = 1
   $$
2. The second sphere with the equation:
   $$
   (x - 1)^2 + y^2 + z^2 = 1
   $$

### Geometric Analysis:
- The first sphere has its center at $(0, 0, 0)$ and radius $1$.
- The second sphere has its center at $(1, 0, 0)$ and radius $1$.
- These two spheres are intersecting, and the curve of intersection will be a circle because the intersection of two spheres is always a circle when the spheres are not identical and do not contain one another.

### Equation of the Curve:
To find the equation of the curve formed by the intersection, we will subtract the second equation from the first.

1. Start with the two equations:

   $$
   x^2 + y^2 + z^2 = 1 \quad \text{(Equation of Sphere 1)}
   $$
   and
   $$
   (x - 1)^2 + y^2 + z^2 = 1 \quad \text{(Equation of Sphere 2)}.
   $$

2. Expanding the second equation:

   $$
   (x - 1)^2 + y^2 + z^2 = 1
   $$
   Expanding $(x - 1)^2$:
   $$
   x^2 - 2x + 1 + y^2 + z^2 = 1.
   $$

3. Now subtract the first equation from the second:
   
   $$
   (x^2 - 2x + 1 + y^2 + z^2) - (x^2 + y^2 + z^2) = 1 - 1.
   $$

   Simplifying:
   $$
   -2x + 1 = 0.
   $$

4. Solving for $x$:
   $$
   x = \frac{1}{2}.
   $$

### Equation of the Curve:
Since $x = \frac{1}{2}$, this equation describes a vertical plane at $x = \frac{1}{2}$ where the two spheres intersect. The equation of the intersection curve can be derived by substituting $x = \frac{1}{2}$ into the first sphere's equation.

Substitute $x = \frac{1}{2}$ into $x^2 + y^2 + z^2 = 1$:
$$
\left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1.
$$
Simplifying:
$$
\frac{1}{4} + y^2 + z^2 = 1.
$$
$$
y^2 + z^2 = 1 - \frac{1}{4}.
$$
$$
y^2 + z^2 = \frac{3}{4}.
$$

Thus, the equation of the curve of intersection is:
$$
x = \frac{1}{2}, \quad y^2 + z^2 = \frac{3}{4}.
$$

### Conclusion:
- The intersection of the two spheres forms a circle.
- The circle lies in the plane $x = \frac{1}{2}$, and its radius is $\frac{\sqrt{3}}{2}$.
- The equation of the curve of intersection is:
  $$
  x = \frac{1}{2}, \quad y^2 + z^2 = \frac{3}{4}.
  $$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters
u = np.linspace(0, 2 * np.pi, 100)  # Angles for u
v = np.linspace(0, np.pi, 100)      # Angles for v

# First sphere: Center (0, 0, 0), Radius 1
x1 = np.outer(np.cos(u), np.sin(v))
y1 = np.outer(np.sin(u), np.sin(v))
z1 = np.outer(np.ones(np.size(u)), np.cos(v))

# Second sphere: Center (1, 0, 0), Radius 1
x2 = np.outer(np.cos(u), np.sin(v)) - 1  # Shifted by x = 1
y2 = np.outer(np.sin(u), np.sin(v))
z2 = np.outer(np.ones(np.size(u)), np.cos(v))

# Create the plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the first sphere
ax.plot_surface(x1, y1, z1, color='b', alpha=0.3, rstride=5, cstride=5)

# Plot the second sphere
ax.plot_surface(x2, y2, z2, color='g', alpha=0.3, rstride=5, cstride=5)

# Calculate the intersection curve
# The intersection curve's x value is known to be 0.5
x_curve = 0.5 * np.ones_like(u)  # x coordinates
y_curve = np.sqrt(3 / 4) * np.sin(u)  # y coordinates
z_curve = np.sqrt(3 / 4) * np.cos(u)  # z coordinates

# Plot the intersection curve
ax.plot(y_curve, z_curve, x_curve, color='r', linewidth=3)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Title
plt.title("Intersection of Two Spheres and the Intersection Curve")
plt.show()
```

### Equation of the Tangent Plane to the Paraboloid at Point P(1, 0, 1)

We are given the paraboloid equation:

$$
z = (x - 1)^2 + y^2 + 1
$$

We need to find the equation of the tangent plane at point \( P(1, 0, 1) \).

1. **Compute the Gradient:** The gradient of the surface gives the normal vector to the surface. The gradient is:

$$
\nabla z = \left( \frac{\partial z}{\partial x}, \frac{\partial z}{\partial y} \right)
$$

2. **Compute Partial Derivatives:**

$$
\frac{\partial z}{\partial x} = 2(x - 1)
$$

$$
\frac{\partial z}{\partial y} = 2y
$$

3. **Evaluate the Partial Derivatives at \( P(1, 0, 1) \):**

$$
\frac{\partial z}{\partial x}(1, 0) = 2(1 - 1) = 0
$$

$$
\frac{\partial z}{\partial y}(1, 0) = 2(0) = 0
$$

4. **Write the Equation of the Tangent Plane:** The equation of the tangent plane is:

$$
z - 1 = 0(x - 1) + 0(y - 0)
$$

Simplifying, we get:

$$
z = 1
$$

Thus, the equation of the tangent plane to the paraboloid \( z = (x - 1)^2 + y^2 + 1 \) at the point \( P(1, 0, 1) \) is:

$$
z = 1
$$


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-2, 4, 100)  # x ranges from -2 to 4
y = np.linspace(-2, 4, 100)  # y ranges from -2 to 4
x, y = np.meshgrid(x, y)  # Create a grid for plotting

# Define the equation of the paraboloid: z = (x - 1)^2 + y^2 + 1
z_paraboloid = (x - 1)**2 + y**2 + 1

# Define the equation of the tangent plane at P(1, 0, 1): z = 2(x - 1) + 2y + 1
z_tangent_plane = 2*(x - 1) + 2*y + 1

# Create a figure for 3D plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the paraboloid surface with a color map
ax.plot_surface(x, y, z_paraboloid, cmap='viridis', alpha=0.7, label="Paraboloid")

# Plot the tangent plane with a semi-transparent red color
ax.plot_surface(x, y, z_tangent_plane, color='r', alpha=0.5, label="Tangent Plane")

# Mark the point P(1, 0, 1) on the plot
ax.scatter(1, 0, 1, color='r', s=100, label="Point P(1, 0, 1)", edgecolors='k')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Tangent Plane to Paraboloid at P(1, 0, 1)')

# Add a legend to describe the plot elements
ax.legend()

# Show the plot
plt.show()
```
