# Limits of Sequences

A **sequence** is an ordered list of numbers, typically represented as $a_1, a_2, a_3, \dots$, where each number in the list is called a **term**. The **limit** of a sequence refers to the value that the terms of the sequence approach as the index of the terms increases indefinitely.

## Definition of the Limit of a Sequence

Let $ \{a_n\} $ be a sequence of real numbers. We say that the sequence $ \{a_n\} $ has a limit $ L $ if, for every $ \epsilon > 0 $, there exists a positive integer $ N $ such that for all $ n \geq N $, the inequality

$$
|a_n - L| < \epsilon
$$

holds true. In other words, the terms of the sequence become arbitrarily close to $ L $ as $ n $ gets larger.

We denote the limit of the sequence as:

$$
\lim_{n \to \infty} a_n = L
$$

### Intuition behind the Definition

The definition of the limit of a sequence means that as we move further along the sequence (i.e., as $ n \to \infty $), the terms $ a_n $ get closer and closer to the value $ L $. The term $ L $ is called the **limit** of the sequence.

### Example: Limit of the Sequence $ \frac{1}{n} $

Consider the sequence $ a_n = \frac{1}{n} $, where $ n = 1, 2, 3, \dots $.

Let's calculate its limit as $ n \to \infty $:

$$
\lim_{n \to \infty} \frac{1}{n} = 0
$$

As $ n $ increases, $ \frac{1}{n} $ gets smaller and smaller, and it approaches 0. Therefore, the limit of this sequence is $ 0 $.

### Example: Limit of the Sequence $ (-1)^n $

Consider the sequence $ a_n = (-1)^n $, where $ n = 1, 2, 3, \dots $.

Let's calculate its limit as $ n \to \infty $:

$$
\lim_{n \to \infty} (-1)^n
$$

This sequence does not have a limit because the terms alternate between -1 and 1. The terms do not approach any specific value as $ n \to \infty $. Therefore, we say that the sequence **does not converge**.

### Convergence and Divergence

- A sequence is said to **converge** if it has a limit. That is, if $ \lim_{n \to \infty} a_n = L $, the sequence converges to $ L $.
  
- A sequence is said to **diverge** if it does not have a finite limit. That is, if $ \lim_{n \to \infty} a_n $ does not exist or is infinite, the sequence diverges.

### Common Examples of Sequences with Limits

1. **Constant Sequence**: If $ a_n = c $, where $ c $ is a constant, then:

   $$
   \lim_{n \to \infty} a_n = c
   $$

2. **Arithmetic Sequence**: Consider the sequence $ a_n = 2n + 3 $, where $ n = 1, 2, 3, \dots $. As $ n \to \infty $, the terms increase without bound, and therefore the sequence diverges to infinity:

   $$
   \lim_{n \to \infty} a_n = \infty
   $$

3. **Geometric Sequence**: Consider the sequence $ a_n = \left(\frac{1}{2}\right)^n $, where $ n = 1, 2, 3, \dots $. As $ n \to \infty $, the terms get closer and closer to 0, so:

   $$
   \lim_{n \to \infty} a_n = 0
   $$

### The Squeeze Theorem

If $ a_n \leq b_n \leq c_n $ for all $ n $ sufficiently large, and if:

$$
\lim_{n \to \infty} a_n = \lim_{n \to \infty} c_n = L
$$

then:

$$
\lim_{n \to \infty} b_n = L
$$

This is known as the **Squeeze Theorem** and can be useful in proving the limit of certain sequences.

---

## Summary

- The limit of a sequence is the value that the terms of the sequence approach as the index increases.
- A sequence converges if it has a limit and diverges if it does not.
- Key examples of sequences include constant sequences, arithmetic sequences, and geometric sequences.
- The Squeeze Theorem is a useful tool for finding the limit of certain sequences.

## Calculations of Limits

### First Limit:
We want to calculate the following limit:

$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}
$$

#### Step-by-Step Calculation:
1. **Identify the highest power of \( n \)** in both the numerator and denominator. Here, the highest power of \( n \) is \( n^2 \) in both parts.
2. **Divide both the numerator and denominator by \( n^2 \)**:
   
   $$
   \lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \lim_{n \to \infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}
   $$

3. **Evaluate the limit as \( n \to \infty \)**. As \( n \) approaches infinity, \( \frac{3}{n} \) and \( \frac{2}{n} \) both tend to 0:

   $$
   \lim_{n \to \infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}} = \frac{1 + 0}{2 - 0} = \frac{1}{2}
   $$

Thus, the value of the limit is:

$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \frac{1}{2}
$$

### Second Limit:
We now calculate the following limit:

$$
\lim_{n \to \infty} \frac{(2n + 3)}{3n^3 - 1}
$$

#### Step-by-Step Calculation:
1. **Identify the highest power of \( n \)** in both the numerator and denominator. The highest power in the numerator is \( n \), and in the denominator, it is \( n^3 \).
2. **Divide both the numerator and denominator by \( n^3 \)**:

   $$
   \lim_{n \to \infty} \frac{(2n + 3)}{3n^3 - 1} = \lim_{n \to \infty} \frac{\frac{2}{n^2} + \frac{3}{n^3}}{3 - \frac{1}{n^3}}
   $$

3. **Evaluate the limit as \( n \to \infty \)**. As \( n \to \infty \), \( \frac{2}{n^2} \), \( \frac{3}{n^3} \), and \( \frac{1}{n^3} \) all tend to 0:

   $$
   \lim_{n \to \infty} \frac{\frac{2}{n^2} + \frac{3}{n^3}}{3 - \frac{1}{n^3}} = \frac{0 + 0}{3 - 0} = 0
   $$

Thus, the value of the limit is:

$$
\lim_{n \to \infty} \frac{(2n + 3)}{3n^3 - 1} = 0
$$

## Final Answer:
- The first limit is \( \frac{1}{2} \).
- The second limit is \( 0 \).

## Proof using the Squeeze Theorem

We are tasked with proving the following limit using the squeeze theorem:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$$

### Step-by-Step Proof:

1. **Recall the Squeeze Theorem**:  
   The Squeeze Theorem states that if we have three functions \( f(n) \), \( g(n) \), and \( h(n) \) such that:

   $$
   f(n) \leq g(n) \leq h(n) \quad \text{for all} \quad n
   $$

   and if:

   $$
   \lim_{n \to \infty} f(n) = \lim_{n \to \infty} h(n) = L,
   $$

   then:

   $$
   \lim_{n \to \infty} g(n) = L.
   $$

2. **Apply the Squeeze Theorem**:  
   We know that for all \( n \):

   $$
   -1 \leq \sin(n) \leq 1
   $$

   Dividing through by \( n \) (which is positive as \( n \to \infty \)):

   $$
   -\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
   $$

3. **Take the limit of the bounds**:  
   Now, take the limit as \( n \to \infty \) of both sides:

   $$
   \lim_{n \to \infty} -\frac{1}{n} = 0
   $$

   and

   $$
   \lim_{n \to \infty} \frac{1}{n} = 0
   $$

4. **Conclusion**:  
   Since:

   $$
   -\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
   $$

   and both the upper and lower bounds approach 0 as \( n \to \infty \), by the Squeeze Theorem:

   $$
   \lim_{n \to \infty} \frac{\sin(n)}{n} = 0
   $$

Thus, we have proven that:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0.
$$

## Finding the Limit of the Sequence

We are tasked with finding the limit of the following sequence:

$$
a_n = \left( 1 + \frac{1}{n} \right)^n
$$

### Step-by-Step Solution:

1. **Recognize the form of the sequence**:  
   The sequence has the form of a well-known limit expression. We recognize that:

   $$
   \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n
   $$

   is a standard limit that converges to the number \( e \), where \( e \) is the base of the natural logarithm.

2. **Intuition**:  
   As \( n \) increases, the expression \( \left( 1 + \frac{1}{n} \right) \) gets closer to 1, but it is raised to the power \( n \), so it mimics the definition of the exponential function.

3. **Conclusion**:  
   Therefore, the limit of the sequence is:

   $$
   \lim_{n \to \infty} a_n = e
   $$

Thus, we have:

$$
\lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e.
$$
