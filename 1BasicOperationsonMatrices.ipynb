# Basic Operations on Matrices

Matrices are mathematical objects used for representing and solving systems of linear equations, transformations, and many other applications. Below are the basic matrix operations.

### 1. Matrix Addition

Matrix addition is defined when two matrices have the same dimensions (i.e., the same number of rows and columns). To add two matrices, simply add their corresponding elements.

#### Example:

Let $A$ and $B$ be two $2 \times 2$ matrices:

$$
A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}, \quad B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
$$

The sum of $A$ and $B$ is:

$$
A + B = \begin{pmatrix} 1+5 & 2+6 \\ 3+7 & 4+8 \end{pmatrix} = \begin{pmatrix} 6 & 8 \\ 10 & 12 \end{pmatrix}
$$

### 2. Matrix Subtraction

Matrix subtraction is similar to matrix addition. It is defined when two matrices have the same dimensions, and you subtract their corresponding elements.

#### Example:

Let $A$ and $B$ be as defined above:

$$
A - B = \begin{pmatrix} 1-5 & 2-6 \\ 3-7 & 4-8 \end{pmatrix} = \begin{pmatrix} -4 & -4 \\ -4 & -4 \end{pmatrix}
$$

### 3. Scalar Multiplication

Scalar multiplication is the multiplication of a matrix by a scalar (a constant number). To multiply a matrix by a scalar, multiply each element of the matrix by that scalar.

#### Example:

Let $A$ be the matrix:

$$
A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
$$

If we multiply matrix $A$ by a scalar $k = 2$:

$$
kA = 2 \times \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} = \begin{pmatrix} 2 & 4 \\ 6 & 8 \end{pmatrix}
$$

### 4. Matrix Multiplication

Matrix multiplication is defined when the number of columns in the first matrix is equal to the number of rows in the second matrix. The element in the $i$-th row and $j$-th column of the product is obtained by taking the dot product of the $i$-th row of the first matrix and the $j$-th column of the second matrix.

#### Example:

Let $A$ be a $2 \times 3$ matrix and $B$ be a $3 \times 2$ matrix:

$$
A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{pmatrix}, \quad B = \begin{pmatrix} 7 & 8 \\ 9 & 10 \\ 11 & 12 \end{pmatrix}
$$

The product of $A$ and $B$ is:

$$
AB = \begin{pmatrix} (1 \times 7 + 2 \times 9 + 3 \times 11) & (1 \times 8 + 2 \times 10 + 3 \times 12) \\ (4 \times 7 + 5 \times 9 + 6 \times 11) & (4 \times 8 + 5 \times 10 + 6 \times 12) \end{pmatrix}
$$

Simplifying:

$$
AB = \begin{pmatrix} 58 & 64 \\ 139 & 154 \end{pmatrix}
$$

### 5. Matrix Transpose

The transpose of a matrix $A$ is obtained by swapping its rows and columns.

#### Example:

Let $A$ be a $2 \times 3$ matrix:

$$
A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{pmatrix}
$$

The transpose of $A$ is:

$$
A^T = \begin{pmatrix} 1 & 4 \\ 2 & 5 \\ 3 & 6 \end{pmatrix}
$$

### 6. Matrix Determinant (for Square Matrices)

The determinant is a scalar value that can be computed from a square matrix. For a $2 \times 2$ matrix:

$$
A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}
$$

The determinant of $A$ is:

$$
\text{det}(A) = ad - bc
$$

#### Example:

Let:

$$
A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
$$

The determinant of $A$ is:

$$
\text{det}(A) = (1 \times 4) - (2 \times 3) = 4 - 6 = -2
$$

### 7. Inverse of a Matrix (for Square Matrices)

The inverse of a matrix $A$ is denoted by $A^{-1}$, and it exists only if the determinant of $A$ is non-zero. The inverse of a $2 \times 2$ matrix is given by:

$$
A^{-1} = \frac{1}{\text{det}(A)} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix}
$$

#### Example:

Let:

$$
A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
$$

The determinant of $A$ is:

$$
\text{det}(A) = (1 \times 4) - (2 \times 3) = -2
$$

The inverse of $A$ is:

$$
A^{-1} = \frac{1}{-2} \begin{pmatrix} 4 & -2 \\ -3 & 1 \end{pmatrix} = \begin{pmatrix} -2 & 1 \\ 1.5 & -0.5 \end{pmatrix}
$$

---

### Conclusion

These are the basic operations that can be performed on matrices. These operations are fundamental to matrix algebra and are widely used in solving systems of equations, transformations, and many areas of mathematics and applied sciences.


# Matrices and Operations

We are given the following matrices:

### Matrix A:
$$
A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
$$

### Matrix B:
$$
B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
$$

### Matrix C:
$$
C = \begin{pmatrix} -1 & 2 \\ 3 & 0 \end{pmatrix}
$$

### Matrix D:
$$
D = \begin{pmatrix} -1 & 2 & 3 \\ 4 & 0 & 6 \end{pmatrix}
$$

### Matrix E:
$$
E = \begin{pmatrix} 1 & 2 \\ 4 & 5 \\ 7 & 8 \end{pmatrix}
$$

---

### Matrix Operations

#### 1. Matrix Addition (Example: \( A + B \)):

Matrix addition is possible only if matrices have the same dimensions.

**For \( A + B \):**
Since both \( A \) and \( B \) are 2x2 matrices, we can add them:

$$
A + B = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} + \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} = \begin{pmatrix} 1+5 & 2+6 \\ 3+7 & 4+8 \end{pmatrix} = \begin{pmatrix} 6 & 8 \\ 10 & 12 \end{pmatrix}
$$

#### 2. Matrix Subtraction (Example: \( A - C \)):

Matrix subtraction is also defined for matrices with the same dimensions.

**For \( A - C \):**
Since both \( A \) and \( C \) are 2x2 matrices:

$$
A - C = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} - \begin{pmatrix} -1 & 2 \\ 3 & 0 \end{pmatrix} = \begin{pmatrix} 1 - (-1) & 2 - 2 \\ 3 - 3 & 4 - 0 \end{pmatrix} = \begin{pmatrix} 2 & 0 \\ 0 & 4 \end{pmatrix}
$$

#### 3. Matrix Multiplication (Example: \( A \times B \)):

Matrix multiplication is defined when the number of columns in the first matrix is equal to the number of rows in the second matrix.

**For \( A \times B \):**

Both \( A \) and \( B \) are 2x2 matrices, so the multiplication is valid.

$$
A \times B = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \times \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} = \begin{pmatrix} (1 \times 5 + 2 \times 7) & (1 \times 6 + 2 \times 8) \\ (3 \times 5 + 4 \times 7) & (3 \times 6 + 4 \times 8) \end{pmatrix}
$$

Simplifying:

$$
A \times B = \begin{pmatrix} 5 + 14 & 6 + 16 \\ 15 + 28 & 18 + 32 \end{pmatrix} = \begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}
$$

#### 4. Matrix Transpose (Example: \( A^T \)):

The transpose of a matrix is obtained by swapping its rows and columns.

**For \( A^T \):**
Since \( A \) is a 2x2 matrix:

$$
A^T = \begin{pmatrix} 1 & 3 \\ 2 & 4 \end{pmatrix}
$$

#### 5. Matrix Determinant (Example: \( \text{det}(A) \)):

The determinant of a 2x2 matrix is calculated as:

$$
\text{det}(A) = (1 \times 4) - (2 \times 3) = 4 - 6 = -2
$$

---

### Conclusion

These are some basic matrix operations you can perform on the given matrices. The operations demonstrated include matrix addition, subtraction, multiplication, transpose, and determinant calculation.

# Scalar Multiplication of Matrices

Given matrices:

### Matrix A:
$$
A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
$$

### Matrix B:
$$
B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
$$

### Matrix C:
$$
C = \begin{pmatrix} -1 & 2 \\ 3 & 0 \end{pmatrix}
$$

### Matrix D:
$$
D = \begin{pmatrix} -1 & 2 & 3 \\ 4 & 0 & 6 \end{pmatrix}
$$

---

### 1. **Scalar Multiplication: $\frac{1}{2} A$**

We multiply each element of matrix $A$ by $\frac{1}{2}$:

$$
\frac{1}{2} A = \frac{1}{2} \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} = \begin{pmatrix} \frac{1}{2} & 1 \\ \frac{3}{2} & 2 \end{pmatrix}
$$

---

### 2. **Scalar Multiplication: $2B$**

We multiply each element of matrix $B$ by 2:

$$
2B = 2 \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} = \begin{pmatrix} 10 & 12 \\ 14 & 16 \end{pmatrix}
$$

---

### 3. **Scalar Multiplication: $-3C$**

We multiply each element of matrix $C$ by $-3$:

$$
-3C = -3 \begin{pmatrix} -1 & 2 \\ 3 & 0 \end{pmatrix} = \begin{pmatrix} 3 & -6 \\ -9 & 0 \end{pmatrix}
$$

---

### 4. **Scalar Multiplication: $4D$**

We multiply each element of matrix $D$ by 4:

$$
4D = 4 \begin{pmatrix} -1 & 2 & 3 \\ 4 & 0 & 6 \end{pmatrix} = \begin{pmatrix} -4 & 8 & 12 \\ 16 & 0 & 24 \end{pmatrix}
$$

---

### Conclusion

- $\frac{1}{2} A = \begin{pmatrix} \frac{1}{2} & 1 \\ \frac{3}{2} & 2 \end{pmatrix}$
- $2B = \begin{pmatrix} 10 & 12 \\ 14 & 16 \end{pmatrix}$
- $-3C = \begin{pmatrix} 3 & -6 \\ -9 & 0 \end{pmatrix}$
- $4D = \begin{pmatrix} -4 & 8 & 12 \\ 16 & 0 & 24 \end{pmatrix}$


# Matrix Products

Given matrices:

### Matrix A:
$$
A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}
$$

### Matrix B:
$$
B = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
$$

### Matrix D:
$$
D = \begin{pmatrix} -1 & 2 & 3 \\ 4 & 0 & 6 \end{pmatrix}
$$

### Matrix E:
$$
E = \begin{pmatrix} 1 & 2 \\ 4 & 5 \\ 7 & 8 \end{pmatrix}
$$

---

### 1. **Matrix Product: $A \cdot B$**

To multiply matrices, we perform dot products of rows and columns:

$$
A \cdot B = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \cdot \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} = \begin{pmatrix} 1(5) + 2(7) & 1(6) + 2(8) \\ 3(5) + 4(7) & 3(6) + 4(8) \end{pmatrix}
$$

$$
A \cdot B = \begin{pmatrix} 5 + 14 & 6 + 16 \\ 15 + 28 & 18 + 32 \end{pmatrix} = \begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}
$$

---

### 2. **Matrix Product: $B \cdot A$**

Similarly, multiply matrices $B$ and $A$:

$$
B \cdot A = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix} \cdot \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} = \begin{pmatrix} 5(1) + 6(3) & 5(2) + 6(4) \\ 7(1) + 8(3) & 7(2) + 8(4) \end{pmatrix}
$$

$$
B \cdot A = \begin{pmatrix} 5 + 18 & 10 + 24 \\ 7 + 24 & 14 + 32 \end{pmatrix} = \begin{pmatrix} 23 & 34 \\ 31 & 46 \end{pmatrix}
$$

---

### 3. **Matrix Product: $A \cdot D$**

Now, multiply matrices $A$ and $D$:

$$
A \cdot D = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \cdot \begin{pmatrix} -1 & 2 & 3 \\ 4 & 0 & 6 \end{pmatrix} = \begin{pmatrix} 1(-1) + 2(4) & 1(2) + 2(0) & 1(3) + 2(6) \\ 3(-1) + 4(4) & 3(2) + 4(0) & 3(3) + 4(6) \end{pmatrix}
$$

$$
A \cdot D = \begin{pmatrix} -1 + 8 & 2 + 0 & 3 + 12 \\ -3 + 16 & 6 + 0 & 9 + 24 \end{pmatrix} = \begin{pmatrix} 7 & 2 & 15 \\ 13 & 6 & 33 \end{pmatrix}
$$

---

### 4. **Matrix Product: $D \cdot E$**

Finally, multiply matrices $D$ and $E$:

$$
D \cdot E = \begin{pmatrix} -1 & 2 & 3 \\ 4 & 0 & 6 \end{pmatrix} \cdot \begin{pmatrix} 1 & 2 \\ 4 & 5 \\ 7 & 8 \end{pmatrix}
$$

Since the number of columns in $D$ (3 columns) matches the number of rows in $E$ (3 rows), this matrix multiplication is valid. Perform the multiplication:

$$
D \cdot E = \begin{pmatrix} (-1)(1) + 2(4) + 3(7) & (-1)(2) + 2(5) + 3(8) \\ 4(1) + 0(4) + 6(7) & 4(2) + 0(5) + 6(8) \end{pmatrix}
$$

$$
D \cdot E = \begin{pmatrix} -1 + 8 + 21 & -2 + 10 + 24 \\ 4 + 0 + 42 & 8 + 0 + 48 \end{pmatrix} = \begin{pmatrix} 28 & 32 \\ 46 & 56 \end{pmatrix}
$$

---

### Conclusion

- $A \cdot B = \begin{pmatrix} 19 & 22 \\ 43 & 50 \end{pmatrix}$
- $B \cdot A = \begin{pmatrix} 23 & 34 \\ 31 & 46 \end{pmatrix}$
- $A \cdot D = \begin{pmatrix} 7 & 2 & 15 \\ 13 & 6 & 33 \end{pmatrix}$
- $D \cdot E = \begin{pmatrix} 28 & 32 \\ 46 & 56 \end{pmatrix}$
