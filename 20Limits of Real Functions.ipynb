# Limits of Real Functions

In calculus, the concept of a limit describes the behavior of a function as the input approaches a particular value. Limits are fundamental in defining derivatives, integrals, and continuity. This section will cover the concept of limits of real functions and how they are used in calculus.

## Definition of a Limit

Given a function \( f(x) \), the limit of \( f(x) \) as \( x \) approaches a number \( a \) is the value that \( f(x) \) gets closer to as \( x \) gets closer to \( a \).

The limit is written as:

$$
\lim_{x \to a} f(x) = L
$$

where \( L \) is the value that \( f(x) \) approaches as \( x \) gets arbitrarily close to \( a \).

### Left-Hand and Right-Hand Limits

When evaluating limits, we may consider one-sided limits:

- **Left-Hand Limit**: The limit of \( f(x) \) as \( x \) approaches \( a \) from the left (\( x \to a^- \)):
  $$
  \lim_{x \to a^-} f(x)
  $$

- **Right-Hand Limit**: The limit of \( f(x) \) as \( x \) approaches \( a \) from the right (\( x \to a^+ \)):
  $$
  \lim_{x \to a^+} f(x)
  $$

For the limit to exist at \( a \), both the left-hand limit and right-hand limit must be equal.

## Basic Limit Rules

The following basic properties and rules are used to evaluate limits of real functions:

1. **Sum Rule**:
   $$
   \lim_{x \to a} \left( f(x) + g(x) \right) = \lim_{x \to a} f(x) + \lim_{x \to a} g(x)
   $$

2. **Product Rule**:
   $$
   \lim_{x \to a} \left( f(x) \cdot g(x) \right) = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x)
   $$

3. **Quotient Rule**:
   $$
   \lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)} \quad \text{if} \quad \lim_{x \to a} g(x) \neq 0
   $$

4. **Constant Multiple Rule**:
   $$
   \lim_{x \to a} \left( c \cdot f(x) \right) = c \cdot \lim_{x \to a} f(x)
   $$

5. **Power Rule**:
   $$
   \lim_{x \to a} \left( f(x) \right)^n = \left( \lim_{x \to a} f(x) \right)^n \quad \text{for any integer } n
   $$

## Examples

### Example 1: Limit of a Polynomial Function

Consider the function:
$$
f(x) = 3x^2 + 2x - 1
$$

To find the limit as \( x \) approaches 2:
$$
\lim_{x \to 2} \left( 3x^2 + 2x - 1 \right)
$$

We simply substitute \( x = 2 \) into the function:

$$
f(2) = 3(2)^2 + 2(2) - 1 = 12 + 4 - 1 = 15
$$

So:
$$
\lim_{x \to 2} \left( 3x^2 + 2x - 1 \right) = 15
$$

### Example 2: Limit of a Rational Function

Consider the function:
$$
f(x) = \frac{2x^2 + 3x - 5}{x - 1}
$$

To find the limit as \( x \) approaches 1:
$$
\lim_{x \to 1} \frac{2x^2 + 3x - 5}{x - 1}
$$

Substitute \( x = 1 \) directly:

$$
f(1) = \frac{2(1)^2 + 3(1) - 5}{1 - 1} = \frac{0}{0}
$$

This is an indeterminate form. To evaluate the limit, we can try simplifying the expression using factorization or L'Hopital's Rule. For this case, the limit is:

$$
\lim_{x \to 1} \frac{2x^2 + 3x - 5}{x - 1} = 5
$$

## Conclusion

Understanding the limits of real functions is essential in calculus, as they form the foundation for defining continuity, derivatives, and integrals. By applying the basic limit rules and techniques, we can evaluate limits for various functions effectively.

# Limit Computation

We are asked to compute the following limit:

$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}
$$

### Step-by-Step Solution

To compute this limit, we can first simplify the expression by factoring out the highest power of \( x \) from both the numerator and the denominator.

#### Step 1: Factor out the highest power of \( x \)

The highest power of \( x \) in the numerator is \( x^3 \), and the highest power of \( x \) in the denominator is \( x^4 \).

So, we can factor out \( x^3 \) in the numerator and \( x^4 \) in the denominator:

$$
\frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{x^3(1 + \frac{2}{x})}{x^4(1 - \frac{3}{x})}
$$

#### Step 2: Simplify the expression

We can now cancel \( x^3 \) from the numerator and denominator:

$$
= \frac{1 + \frac{2}{x}}{x(1 - \frac{3}{x})}
$$

#### Step 3: Take the limit as \( x \to \infty \)

As \( x \to \infty \), the terms \( \frac{2}{x} \) and \( \frac{3}{x} \) approach 0. Therefore, we can evaluate the limit:

$$
\lim_{x \to \infty} \frac{1 + \frac{2}{x}}{x(1 - \frac{3}{x})} = \frac{1 + 0}{\infty(1 - 0)} = \frac{1}{\infty} = 0
$$

### Final Answer:

$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = 0
$$

# Limit Computation

We are asked to compute the following limit:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}
$$

### Step-by-Step Solution

#### Step 1: Apply the standard limit formula

We know that:

$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1
$$

We can rewrite the given expression in terms of this standard limit. Notice that \( \sin(3x) \) has a factor of \( 3x \). So, we factor out the 3 from \( \sin(3x) \) to apply the standard limit:

$$
\frac{\sin(3x)}{2x + 1} = \frac{\sin(3x)}{3x} \cdot \frac{3x}{2x + 1}
$$

#### Step 2: Take the limit of each part

Now, we can take the limit of each part separately:

1. The first part:

$$
\lim_{x \to 0} \frac{\sin(3x)}{3x} = 1
$$

2. The second part:

$$
\lim_{x \to 0} \frac{3x}{2x + 1} = \frac{0}{1} = 0
$$

#### Step 3: Multiply the limits

Now, we multiply the two results:

$$
1 \cdot 0 = 0
$$

### Final Answer:

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} = 0
$$

# Asymptotes of the Given Functions

We are asked to find the asymptotes of the following functions:

1. **Function 1:**
   $$
   f(x) = \frac{x^2 - 1}{x^2 + 1}
   $$

2. **Function 2:**
   $$
   g(x) = \frac{\sin(x)}{x^2 + 1}
   $$

### Step 1: Asymptotes of Function 1

To find the asymptotes of $f(x) = \frac{x^2 - 1}{x^2 + 1}$, we consider two types of asymptotes: vertical and horizontal.

#### Vertical Asymptotes

A vertical asymptote occurs when the denominator equals zero, and the numerator is non-zero. For $f(x)$:

- The denominator is $x^2 + 1$. Set it equal to zero:

$$
x^2 + 1 = 0
$$

- This equation has no real solutions because $x^2 + 1 \geq 1$ for all real values of $x$.
- Therefore, **there are no vertical asymptotes**.

#### Horizontal Asymptotes

To find the horizontal asymptotes, we examine the behavior of $f(x)$ as $x \to \infty$ or $x \to -\infty$. For large values of $x$, we divide the numerator and denominator by $x^2$:

$$
f(x) = \frac{x^2 - 1}{x^2 + 1} = \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}}
$$

As $x \to \infty$, both $\frac{1}{x^2} \to 0$. Therefore, the function approaches:

$$
f(x) \to \frac{1 - 0}{1 + 0} = 1
$$

Thus, the **horizontal asymptote** is:

$$
y = 1
$$

### Step 2: Asymptotes of Function 2

Now, for the second function $g(x) = \frac{\sin(x)}{x^2 + 1}$, let's analyze the asymptotes.

#### Vertical Asymptotes

We check if the denominator can be zero:

$$
x^2 + 1 = 0
$$

This equation also has no real solutions. Therefore, **there are no vertical asymptotes**.

#### Horizontal Asymptotes

For large $|x|$, the denominator $x^2 + 1$ grows faster than the numerator $\sin(x)$, which is bounded between $-1$ and $1$. So, as $x \to \infty$ or $x \to -\infty$, $g(x)$ approaches $0$:

$$
g(x) \to \frac{\sin(x)}{x^2 + 1} \to 0
$$

Thus, the **horizontal asymptote** is:

$$
y = 0
$$

### Final Results

- The asymptote for $f(x) = \frac{x^2 - 1}{x^2 + 1}$ is $y = 1$ (horizontal asymptote).
- The asymptote for $g(x) = \frac{\sin(x)}{x^2 + 1}$ is $y = 0$ (horizontal asymptote).
