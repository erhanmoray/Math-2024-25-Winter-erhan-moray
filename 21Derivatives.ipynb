# Derivatives

A **derivative** represents the rate of change of a function with respect to a variable. It can be thought of as the slope of the tangent line to the graph of the function at any given point.

### Definition of Derivative

The derivative of a function $f(x)$ at a point $x = a$ is defined as:

$$
f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}
$$

### Common Rules for Differentiation

1. **Power Rule:**

   For any real number $n$, the derivative of $f(x) = x^n$ is:

   $$
   f'(x) = nx^{n-1}
   $$

2. **Sum Rule:**

   If $f(x) = g(x) + h(x)$, then:

   $$
   f'(x) = g'(x) + h'(x)
   $$

3. **Product Rule:**

   If $f(x) = g(x) \cdot h(x)$, then:

   $$
   f'(x) = g'(x)h(x) + g(x)h'(x)
   $$

4. **Quotient Rule:**

   If $f(x) = \frac{g(x)}{h(x)}$, then:

   $$
   f'(x) = \frac{g'(x)h(x) - g(x)h'(x)}{[h(x)]^2}
   $$

5. **Chain Rule:**

   If $f(x) = g(h(x))$, then:

   $$
   f'(x) = g'(h(x)) \cdot h'(x)
   $$

### Examples of Derivatives

#### Example 1: Power Rule

Find the derivative of the function $f(x) = x^3$.

Using the power rule:

$$
f'(x) = 3x^{3-1} = 3x^2
$$

#### Example 2: Product Rule

Find the derivative of $f(x) = x^2 \cdot \sin(x)$.

Using the product rule:

$$
f'(x) = 2x \cdot \sin(x) + x^2 \cdot \cos(x)
$$

#### Example 3: Quotient Rule

Find the derivative of $f(x) = \frac{x^2}{x + 1}$.

Using the quotient rule:

$$
f'(x) = \frac{(2x)(x+1) - (x^2)(1)}{(x+1)^2}
$$

Simplifying:

$$
f'(x) = \frac{x^2 + 2x - x^2}{(x+1)^2} = \frac{2x}{(x+1)^2}
$$

#### Example 4: Chain Rule

Find the derivative of $f(x) = \sin(x^2)$.

Using the chain rule:

$$
f'(x) = \cos(x^2) \cdot 2x
$$

Thus:

$$
f'(x) = 2x \cos(x^2)
$$

### Higher Order Derivatives

- The second derivative, denoted $f''(x)$, is the derivative of the first derivative. It represents the rate of change of the rate of change of the function.

$$
f''(x) = \frac{d}{dx} f'(x)
$$

- Similarly, the third derivative, denoted $f^{(3)}(x)$, is the derivative of the second derivative, and so on.

### Notation for Derivatives

- $f'(x)$: First derivative of $f(x)$.
- $f''(x)$: Second derivative of $f(x)$.
- $f^{(n)}(x)$: $n$-th derivative of $f(x)$.

### Implicit Differentiation

Sometimes, we have equations involving both $x$ and $y$, and we want to find $\frac{dy}{dx}$. In such cases, we use **implicit differentiation**. For example:

If $x^2 + y^2 = 1$, differentiate both sides with respect to $x$:

$$
2x + 2y \frac{dy}{dx} = 0
$$

Solve for $\frac{dy}{dx}$:

$$
\frac{dy}{dx} = -\frac{x}{y}
$$

### Derivatives of Common Functions

- The derivative of $e^x$ is:

  $$
  \frac{d}{dx} e^x = e^x
  $$

- The derivative of $\ln(x)$ is:

  $$
  \frac{d}{dx} \ln(x) = \frac{1}{x}
  $$

- The derivative of $\sin(x)$ is:

  $$
  \frac{d}{dx} \sin(x) = \cos(x)
  $$

- The derivative of $\cos(x)$ is:

  $$
  \frac{d}{dx} \cos(x) = -\sin(x)
  $$

### Conclusion

The derivative is a powerful tool for understanding how a function changes. By applying the rules of differentiation, we can compute the rate of change of functions, find tangent lines, and understand the behavior of functions.

# Derivative Proof

We are asked to prove that:

$$
\frac{d}{dx}\left(\ln(\sin(x))\right) = \cot(x)
$$

### Proof:

We will apply the chain rule for derivatives.

First, let's rewrite the expression:

$$
y = \ln(\sin(x))
$$

Now, differentiate both sides with respect to \(x\):

$$
\frac{dy}{dx} = \frac{d}{dx} \left( \ln(\sin(x)) \right)
$$

The chain rule tells us that the derivative of \( \ln(u) \) with respect to \(x\) is:

$$
\frac{d}{dx} \ln(u) = \frac{1}{u} \cdot \frac{du}{dx}
$$

In our case, \( u = \sin(x) \), so we get:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \frac{d}{dx} (\sin(x))
$$

The derivative of \( \sin(x) \) with respect to \(x\) is \( \cos(x) \):

$$
\frac{d}{dx} \sin(x) = \cos(x)
$$

Thus, we have:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \cos(x)
$$

Simplifying the right-hand side:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{\cos(x)}{\sin(x)}
$$

Finally, recall that \( \cot(x) = \frac{\cos(x)}{\sin(x)} \), so we have:

$$
\frac{d}{dx} \ln(\sin(x)) = \cot(x)
$$

Thus, we have proven the identity:

$$
\frac{d}{dx} \left( \ln(\sin(x)) \right) = \cot(x)
$$

# Verification of the Second Derivative

We are given the function:

$$
f(x) = \cos(x)
$$

We need to verify that:

$$
f''(x) = -f(x)
$$

### Step 1: Compute the first derivative \( f'(x) \)

The first derivative of \( f(x) = \cos(x) \) is:

$$
f'(x) = -\sin(x)
$$

### Step 2: Compute the second derivative \( f''(x) \)

Now, differentiate \( f'(x) = -\sin(x) \) with respect to \( x \):

$$
f''(x) = -\cos(x)
$$

### Step 3: Compare \( f''(x) \) with \( -f(x) \)

Recall that \( f(x) = \cos(x) \). Therefore:

$$
-f(x) = -\cos(x)
$$

We can see that:

$$
f''(x) = -\cos(x) = -f(x)
$$

Thus, we have verified that:

$$
f''(x) = -f(x)
$$

# Using de l'Hôpital's Rule to Find Improper Limits

We are tasked with finding the following limits using de l'Hôpital's Rule:

1. $$ \lim_{x \to 0} \frac{\sin(x)}{x} $$

2. $$ \lim_{x \to \infty} \frac{\ln(x)}{x} $$

3. $$ \lim_{x \to \infty} \frac{e^x}{x} $$

### 1. Find $$ \lim_{x \to 0} \frac{\sin(x)}{x} $$

At \( x = 0 \), both the numerator and the denominator approach 0, so we have a \( \frac{0}{0} \) indeterminate form. We can apply de l'Hôpital's Rule, which states that:

$$
\lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f'(x)}{g'(x)} \quad \text{if the limit exists.}
$$

For our limit:

$$
f(x) = \sin(x), \quad g(x) = x
$$

Now, compute the derivatives:

$$
f'(x) = \cos(x), \quad g'(x) = 1
$$

Applying de l'Hôpital's Rule:

$$
\lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1} = 1
$$

Thus, the limit is:

$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1
$$

### 2. Find $$ \lim_{x \to \infty} \frac{\ln(x)}{x} $$

At \( x \to \infty \), both the numerator and denominator approach infinity, so we have an indeterminate form \( \frac{\infty}{\infty} \). We can apply de l'Hôpital's Rule again.

For our limit:

$$
f(x) = \ln(x), \quad g(x) = x
$$

Now, compute the derivatives:

$$
f'(x) = \frac{1}{x}, \quad g'(x) = 1
$$

Applying de l'Hôpital's Rule:

$$
\lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{\frac{1}{x}}{1} = \lim_{x \to \infty} \frac{1}{x} = 0
$$

Thus, the limit is:

$$
\lim_{x \to \infty} \frac{\ln(x)}{x} = 0
$$

### 3. Find $$ \lim_{x \to \infty} \frac{e^x}{x} $$

At \( x \to \infty \), both the numerator and denominator approach infinity, so we have an indeterminate form \( \frac{\infty}{\infty} \). We can apply de l'Hôpital's Rule again.

For our limit:

$$
f(x) = e^x, \quad g(x) = x
$$

Now, compute the derivatives:

$$
f'(x) = e^x, \quad g'(x) = 1
$$

Applying de l'Hôpital's Rule:

$$
\lim_{x \to \infty} \frac{e^x}{x} = \lim_{x \to \infty} \frac{e^x}{1} = \infty
$$

Thus, the limit is:

$$
\lim_{x \to \infty} \frac{e^x}{x} = \infty
$$

### Final Results:

1. $$ \lim_{x \to 0} \frac{\sin(x)}{x} = 1 $$

2. $$ \lim_{x \to \infty} \frac{\ln(x)}{x} = 0 $$

3. $$ \lim_{x \to \infty} \frac{e^x}{x} = \infty $$

# Finding Velocity and Acceleration

We are given the position of a particle as a function of time:

$$
x(t) = 3t^2 - 6t + 1
$$

### 1. Velocity Function

The velocity is the first derivative of the position function with respect to time. Therefore, we need to compute the derivative of \(x(t)\) to get the velocity function \(V(t)\).

$$
V(t) = \frac{d}{dt} \left( 3t^2 - 6t + 1 \right)
$$

Using the power rule of differentiation:

$$
V(t) = 6t - 6
$$

### 2. Acceleration Function

The acceleration is the derivative of the velocity function, or equivalently, the second derivative of the position function. Thus, we compute the derivative of \(V(t)\):

$$
a(t) = \frac{d}{dt} \left( 6t - 6 \right)
$$

This gives:

$$
a(t) = 6
$$

### 3. Evaluate at \( t = 2 \)

Now we will find the velocity and acceleration at \( t = 2 \).

- For velocity:

$$
V(2) = 6(2) - 6 = 12 - 6 = 6
$$

- For acceleration:

$$
a(2) = 6
$$

### Final Results:

At \( t = 2 \):
- Velocity: \( V(2) = 6 \, \text{units per time} \)
- Acceleration: \( a(2) = 6 \, \text{units per time squared} \)
