# Finding Extrema (Local Maxima and Minima) of a Function

To find the extrema of a function, we follow these steps:

### 1. First Derivative Test

- Find the first derivative of the function, \( f'(x) \), which will help us determine the critical points.
  
  Critical points occur when the first derivative is zero, or the derivative does not exist.
  
  $$ f'(x) = 0 \quad \text{or} \quad f'(x) \text{ does not exist.} $$

- Solve \( f'(x) = 0 \) to find potential critical points.

### 2. Second Derivative Test (for confirmation)

- To determine whether a critical point corresponds to a local minimum, local maximum, or a saddle point, we use the second derivative test:
  
  - If \( f''(x) > 0 \), the function has a local minimum at \( x \).
  - If \( f''(x) < 0 \), the function has a local maximum at \( x \).
  - If \( f''(x) = 0 \), the test is inconclusive, and we need further analysis.

### Example 1: Finding Extrema for \( f(x) = x^2 - 4x + 3 \)

Given the function:

$$ f(x) = x^2 - 4x + 3 $$

#### Step 1: Find the first derivative

$$ f'(x) = 2x - 4 $$

Set the first derivative equal to zero to find the critical points:

$$ 2x - 4 = 0 \quad \Rightarrow \quad x = 2 $$

#### Step 2: Find the second derivative

The second derivative of the function is:

$$ f''(x) = 2 $$

Since \( f''(x) = 2 > 0 \), this means that \( x = 2 \) is a local minimum.

#### Step 3: Conclusion

The function \( f(x) = x^2 - 4x + 3 \) has a local minimum at \( x = 2 \).

---

### Example 2: Finding Extrema for \( f(x) = -x^2 + 4x - 3 \)

Given the function:

$$ f(x) = -x^2 + 4x - 3 $$

#### Step 1: Find the first derivative

$$ f'(x) = -2x + 4 $$

Set the first derivative equal to zero to find the critical points:

$$ -2x + 4 = 0 \quad \Rightarrow \quad x = 2 $$

#### Step 2: Find the second derivative

The second derivative of the function is:

$$ f''(x) = -2 $$

Since \( f''(x) = -2 < 0 \), this means that \( x = 2 \) is a local maximum.

#### Step 3: Conclusion

The function \( f(x) = -x^2 + 4x - 3 \) has a local maximum at \( x = 2 \).

# Maximizing Profit

The profit function is given as:

$$ P(u) = -2u^2 + 50u - 300 $$

Where:
- \( P(u) \) is the profit,
- \( u \) is the number of units sold.

### 1. Find the first derivative of the profit function.

The first derivative, \( P'(u) \), gives the rate of change of the profit with respect to the number of units sold.

To differentiate \( P(u) = -2u^2 + 50u - 300 \), we use the power rule for differentiation:

$$ P'(u) = \frac{d}{du}(-2u^2 + 50u - 300) $$

Taking the derivative:

$$ P'(u) = -4u + 50 $$

### 2. Set the first derivative equal to zero to find critical points.

To find the critical points, we set \( P'(u) = 0 \):

$$ -4u + 50 = 0 $$

Solving for \( u \):

$$ -4u = -50 $$

$$ u = \frac{-50}{-4} $$

$$ u = 12.5 $$

Thus, \( u = 12.5 \) is a critical point.

### 3. Use the second derivative test to confirm if it's a maximum.

To confirm if \( u = 12.5 \) is a maximum, we need to compute the second derivative of the profit function:

$$ P''(u) = \frac{d}{du}(-4u + 50) $$

Taking the derivative:

$$ P''(u) = -4 $$

Since \( P''(u) = -4 \) is negative, the critical point at \( u = 12.5 \) is a local maximum.

# Maximizing the Area of a Rectangle

You are given 10 meters of string, and you need to use it to enclose the largest possible rectangle. We will maximize the area of the rectangle, subject to the constraint that the perimeter is 10 meters.

### Step 1: Define Variables
Let:
- \( L \) be the length of the rectangle,
- \( W \) be the width of the rectangle.

### Step 2: Set Up the Perimeter Constraint

The perimeter of a rectangle is given by the formula:

$$ P = 2L + 2W $$

Since we have 10 meters of string to enclose the rectangle, we have the constraint:

$$ 2L + 2W = 10 $$

Simplify this equation:

$$ L + W = 5 $$

Thus, the sum of the length and width is 5 meters.

### Step 3: Maximize the Area

The area \( A \) of the rectangle is given by:

$$ A = L \times W $$

Using the perimeter constraint, we can express \( W \) in terms of \( L \):

$$ W = 5 - L $$

Now substitute this expression for \( W \) into the area formula:

$$ A = L \times (5 - L) $$

Simplify the expression:

$$ A = 5L - L^2 $$

### Step 4: Find the Critical Points

To maximize the area, we need to find the critical points by taking the derivative of the area with respect to \( L \), and setting it equal to zero:

$$ \frac{dA}{dL} = 5 - 2L $$

Set the derivative equal to zero:

$$ 5 - 2L = 0 $$

Solve for \( L \):

$$ 2L = 5 $$

$$ L = \frac{5}{2} = 2.5 $$

### Step 5: Confirm it's a Maximum

To confirm that \( L = 2.5 \) gives a maximum area, we take the second derivative of the area function:

$$ \frac{d^2A}{dL^2} = -2 $$

Since the second derivative is negative, \( L = 2.5 \) corresponds to a maximum.

### Step 6: Find the Dimensions of the Rectangle

Since \( L = 2.5 \), we can substitute this value into the constraint equation \( L + W = 5 \) to find \( W \):

$$ W = 5 - 2.5 = 2.5 $$

### Conclusion

The dimensions of the rectangle that enclose the largest area are:
- Length \( L = 2.5 \) meters,
- Width \( W = 2.5 \) meters.

Thus, the largest possible rectangle is actually a **square** with side lengths of 2.5 meters.

# Finding the Extremum of the Function

We are given the function:

$$ f(x) = x^2 + 3x - 5 $$

### Step 1: Find the First Derivative

To find the extremum points, we first need to compute the first derivative of the function:

$$ f'(x) = \frac{d}{dx}(x^2 + 3x - 5) $$

Using basic differentiation rules:

$$ f'(x) = 2x + 3 $$

### Step 2: Set the First Derivative Equal to Zero

To find the critical points (the points where the extremum may occur), we set the first derivative equal to zero:

$$ 2x + 3 = 0 $$

Solve for \( x \):

$$ 2x = -3 $$

$$ x = -\frac{3}{2} $$

### Step 3: Verify if It's a Minimum or Maximum

To determine whether this critical point corresponds to a minimum or a maximum, we compute the second derivative of the function:

$$ f''(x) = \frac{d}{dx}(2x + 3) $$

Differentiating:

$$ f''(x) = 2 $$

Since the second derivative is positive, this indicates that the function has a **local minimum** at \( x = -\frac{3}{2} \).

### Step 4: Find the Minimum Value

Now, substitute \( x = -\frac{3}{2} \) into the original function to find the value of the function at this point:

$$ f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5 $$

Simplifying:

$$ f\left(-\frac{3}{2}\right) = \frac{9}{4} - \frac{9}{2} - 5 $$

$$ f\left(-\frac{3}{2}\right) = \frac{9}{4} - \frac{18}{4} - \frac{20}{4} $$

$$ f\left(-\frac{3}{2}\right) = \frac{-29}{4} $$

Thus, the minimum value of the function is:

$$ f\left(-\frac{3}{2}\right) = -\frac{29}{4} $$

### Conclusion

The function \( f(x) = x^2 + 3x - 5 \) has a local minimum at \( x = -\frac{3}{2} \), and the minimum value of the function at this point is \( -\frac{29}{4} \).
