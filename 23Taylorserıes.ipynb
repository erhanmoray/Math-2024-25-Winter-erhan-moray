# Taylor Series

The Taylor series of a function $f(x)$ around a point $x_0$ is given by:

$$
f(x) = f(x_0) + f'(x_0)(x - x_0) + \frac{f''(x_0)}{2!}(x - x_0)^2 + \frac{f^{(3)}(x_0)}{3!}(x - x_0)^3 + \dots
$$

Where:
- $f(x_0)$ is the value of the function at $x_0$,
- $f'(x_0)$ is the first derivative of the function evaluated at $x_0$,
- $f''(x_0)$ is the second derivative of the function evaluated at $x_0$,
- $f^{(n)}(x_0)$ is the $n$-th derivative of the function evaluated at $x_0$.

### Example: Taylor Series of $f(x) = \cos(x)$ around $x_0 = 0$

We will find the Taylor series expansion of $f(x) = \cos(x)$ at $x_0 = 0$.

1. **Evaluate the function at $x_0 = 0$**:
   $$ f(0) = \cos(0) = 1 $$

2. **Find the first derivative**:
   $$ f'(x) = -\sin(x) $$
   $$ f'(0) = -\sin(0) = 0 $$

3. **Find the second derivative**:
   $$ f''(x) = -\cos(x) $$
   $$ f''(0) = -\cos(0) = -1 $$

4. **Find the third derivative**:
   $$ f^{(3)}(x) = \sin(x) $$
   $$ f^{(3)}(0) = \sin(0) = 0 $$

5. **Find the fourth derivative**:
   $$ f^{(4)}(x) = \cos(x) $$
   $$ f^{(4)}(0) = \cos(0) = 1 $$

Thus, the Taylor series expansion of $\cos(x)$ around $x_0 = 0$ is:

$$
\cos(x) = 1 + 0 \cdot (x - 0) + \frac{-1}{2!}(x - 0)^2 + 0 \cdot (x - 0)^3 + \frac{1}{4!}(x - 0)^4 + \dots
$$

Simplifying:

$$
\cos(x) = 1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!} + \dots
$$

### Example: Taylor Series of $f(x) = \frac{1}{1 - x}$ around $x_0 = 0$

We will find the Taylor series expansion of $f(x) = \frac{1}{1 - x}$ at $x_0 = 0$.

1. **Evaluate the function at $x_0 = 0$**:
   $$ f(0) = \frac{1}{1 - 0} = 1 $$

2. **Find the first derivative**:
   $$ f'(x) = \frac{1}{(1 - x)^2} $$
   $$ f'(0) = \frac{1}{(1 - 0)^2} = 1 $$

3. **Find the second derivative**:
   $$ f''(x) = \frac{2}{(1 - x)^3} $$
   $$ f''(0) = \frac{2}{(1 - 0)^3} = 2 $$

4. **Find the third derivative**:
   $$ f^{(3)}(x) = \frac{6}{(1 - x)^4} $$
   $$ f^{(3)}(0) = \frac{6}{(1 - 0)^4} = 6 $$

Thus, the Taylor series expansion of $\frac{1}{1 - x}$ around $x_0 = 0$ is:

$$
\frac{1}{1 - x} = 1 + x + x^2 + x^3 + \dots
$$

### Example: Taylor Series of $f(x) = \sin(x)$ around $x_0 = \pi$

We will find the Taylor series expansion of $f(x) = \sin(x)$ at $x_0 = \pi$.

1. **Evaluate the function at $x_0 = \pi$**:
   $$ f(\pi) = \sin(\pi) = 0 $$

2. **Find the first derivative**:
   $$ f'(x) = \cos(x) $$
   $$ f'(\pi) = \cos(\pi) = -1 $$

3. **Find the second derivative**:
   $$ f''(x) = -\sin(x) $$
   $$ f''(\pi) = -\sin(\pi) = 0 $$

4. **Find the third derivative**:
   $$ f^{(3)}(x) = -\cos(x) $$
   $$ f^{(3)}(\pi) = -\cos(\pi) = 1 $$

Thus, the Taylor series expansion of $\sin(x)$ around $x_0 = \pi$ is:

$$
\sin(x) = 0 - (x - \pi) + 0 \cdot (x - \pi)^2 + \frac{1}{3!}(x - \pi)^3 + \dots
$$

Simplifying:

$$
\sin(x) = -(x - \pi) + \frac{(x - \pi)^3}{3!} + \dots
$$

---

### Final Notes

- The general form of the Taylor series for any function $f(x)$ around a point $x_0$ is:

$$
f(x) = f(x_0) + f'(x_0)(x - x_0) + \frac{f''(x_0)}{2!}(x - x_0)^2 + \frac{f^{(3)}(x_0)}{3!}(x - x_0)^3 + \dots
$$
  
- The Taylor series can be used to approximate the value of a function near a point of expansion $x_0$.


# Taylor Series Expansion

We are given the following functions and asked to find their Taylor series expansions around specified points, up to the 4th degree:

1. $f(x) = \cos(x)$ around $x = 0$
2. $h(x) = \frac{1}{1 - x}$ around $x = 0$
3. $g(x) = \sin(x)$ around $x = \pi$

Additionally, we will visualize the functions and their Taylor series expansions in Geogebra.

---

### Step 1: Taylor Series of $f(x) = \cos(x)$ around $x = 0$

The Taylor series expansion of $\cos(x)$ around $x = 0$ is:

$$
f(x) = \cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n)!} x^{2n}
$$

Up to the 4th degree, this becomes:

$$
f(x) \approx 1 - \frac{x^2}{2!} + \frac{x^4}{4!} = 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

---

### Step 2: Taylor Series of $h(x) = \frac{1}{1 - x}$ around $x = 0$

The Taylor series expansion of $\frac{1}{1 - x}$ around $x = 0$ is:

$$
h(x) = \frac{1}{1 - x} = \sum_{n=0}^{\infty} x^n
$$

Up to the 4th degree, this becomes:

$$
h(x) \approx 1 + x + x^2 + x^3 + x^4
$$

---

### Step 3: Taylor Series of $g(x) = \sin(x)$ around $x = \pi$

We first need to find the Taylor series expansion of $\sin(x)$ around $x = \pi$. Since $\sin(\pi) = 0$ and $\cos(\pi) = -1$, we have:

$$
g(x) = \sin(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(\pi)}{n!} (x - \pi)^n
$$

The derivatives of $\sin(x)$ at $x = \pi$ are:

- $f^{(0)}(\pi) = \sin(\pi) = 0$
- $f^{(1)}(\pi) = \cos(\pi) = -1$
- $f^{(2)}(\pi) = -\sin(\pi) = 0$
- $f^{(3)}(\pi) = -\cos(\pi) = 1$
- $f^{(4)}(\pi) = \sin(\pi) = 0$

Thus, the Taylor series up to the 4th degree is:

$$
g(x) \approx - (x - \pi) + \frac{1}{3!}(x - \pi)^3 = - (x - \pi) + \frac{1}{6}(x - \pi)^3
$$




### Final Summary

- The Taylor series of $f(x) = \cos(x)$ around $x = 0$ up to the 4th degree is:

  $$
  f(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}
  $$

- The Taylor series of $h(x) = \frac{1}{1 - x}$ around $x = 0$ up to the 4th degree is:

  $$
  h(x) \approx 1 + x + x^2 + x^3 + x^4
  $$

- The Taylor series of $g(x) = \sin(x)$ around $x = \pi$ up to the 4th degree is:

# Finding the Tangent Line to the Function $f(x) = e^{\sin(x)}$ at $x_0 = \pi$

We are asked to find the equation of the tangent line to the function $f(x) = e^{\sin(x)}$ at $x_0 = \pi$.

### Step 1: Compute the Derivative of $f(x)$

To find the slope of the tangent line, we first need to compute the derivative of $f(x) = e^{\sin(x)}$.

Using the chain rule:

$$
f'(x) = e^{\sin(x)} \cdot \cos(x)
$$

### Step 2: Evaluate the Derivative and Function at $x_0 = \pi$

Next, we evaluate $f(x)$ and $f'(x)$ at $x_0 = \pi$.

- For $f(\pi)$:

$$
f(\pi) = e^{\sin(\pi)} = e^0 = 1
$$

- For $f'(\pi)$:

$$
f'(\pi) = e^{\sin(\pi)} \cdot \cos(\pi) = e^0 \cdot (-1) = -1
$$

### Step 3: Write the Equation of the Tangent Line

The equation of the tangent line at $x_0 = \pi$ is given by:

$$
y = f'(x_0)(x - x_0) + f(x_0)
$$

Substituting the values of $f'(\pi)$ and $f(\pi)$:

$$
y = -1(x - \pi) + 1
$$

Simplifying:

$$
y = -x + \pi + 1
$$

Thus, the equation of the tangent line at $x_0 = \pi$ is:

$$
y = -x + \pi + 1
$$

---



### Final Notes

- The equation of the tangent line at $x_0 = \pi$ is:

  $$
  y = -x + \pi + 1
