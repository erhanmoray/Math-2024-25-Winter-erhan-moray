# Integral Notes

## 1. Introduction
An **integral** represents the area under a curve or the accumulation of quantities. It is a fundamental concept in calculus with two main types:
- **Definite Integrals**
- **Indefinite Integrals**

---

## 2. Types of Integrals

### 2.1 Indefinite Integrals
The indefinite integral of a function $f(x)$ is written as:
$$
\int f(x) \,dx = F(x) + C
$$
Where:
- $F(x)$ is the antiderivative of $f(x)$.
- $C$ is the constant of integration.

#### Basic Indefinite Integral Rules
| Function $f(x)$ | Integral $\int f(x) \,dx$ |
|-----------------|---------------------------|
| $x^n$ (where $n \neq -1$) | $\frac{x^{n+1}}{n+1} + C$ |
| $e^x$ | $e^x + C$ |
| $\sin x$ | $-\cos x + C$ |
| $\cos x$ | $\sin x + C$ |
| $\frac{1}{x}$ | $\ln |x| + C$ |

---

### 2.2 Definite Integrals
A definite integral represents the signed area under a curve between two points $a$ and $b$:
$$
\int_a^b f(x) \,dx = F(b) - F(a)
$$
Where $F(x)$ is the antiderivative of $f(x)$.

---

## 3. Fundamental Theorem of Calculus
1. If $F(x)$ is an antiderivative of $f(x)$, then:
   $$
   \int_a^b f(x) \,dx = F(b) - F(a)
   $$
2. If $F(x) = \int f(x) \,dx$, then $F'(x) = f(x)$.

---

## 4. Techniques of Integration

### 4.1 Integration by Substitution
Used when the integral can be simplified by substituting:
$$
\int f(g(x))g'(x) \,dx = \int f(u) \,du
$$

### 4.2 Integration by Parts
Based on the product rule for differentiation:
$$
\int u \,dv = uv - \int v \,du
$$
Where $u$ and $v$ are differentiable functions of $x


# Step-by-Step Integral Computations

---

1. $\int 1 \,dx$  
   **Solution:**  
   $ \int 1 \,dx = x + C $

---

2. $\int (x^2 + 2) \,dx$  
   **Step 1:** Apply linearity of integration:  
   $ \int (x^2 + 2) \,dx = \int x^2 \,dx + \int 2 \,dx $  
   **Step 2:** Integrate each term:  
   $ \int x^2 \,dx = \frac{x^3}{3}, \quad \int 2 \,dx = 2x $  
   **Result:**  
   $ \frac{x^3}{3} + 2x + C $

---

3. $\int 2\sin(x) \,dx$  
   **Step 1:** Factor out the constant:  
   $ 2 \int \sin(x) \,dx $  
   **Step 2:** Use the integral of sine:  
   $ \int \sin(x) \,dx = -\cos(x) $  
   **Result:**  
   $ -2\cos(x) + C $

---

4. $\int 3x \,dx$  
   **Step 1:** Factor out the constant:  
   $ 3 \int x \,dx $  
   **Step 2:** Integrate $x$:  
   $ \int x \,dx = \frac{x^2}{2} $  
   **Result:**  
   $ \frac{3x^2}{2} + C $

---

5. $\int \frac{1}{x^2} \,dx$  
   **Step 1:** Rewrite as $x^{-2}$:  
   $ \int x^{-2} \,dx $  
   **Step 2:** Apply the power rule:  
   $ \int x^n \,dx = \frac{x^{n+1}}{n+1} \quad \text{for} \quad n \neq -1 $  
   **Result:**  
   $ \int x^{-2} \,dx = \frac{x^{-1}}{-1} = -\frac{1}{x} + C $

---

6. $\int \left(\frac{1}{3}x^4 - 5\right) \,dx$  
   **Step 1:** Integrate each term separately:  
   $ \int \frac{1}{3}x^4 \,dx = \frac{1}{3} \cdot \frac{x^5}{5} = \frac{1}{15}x^5, \quad \int -5 \,dx = -5x $  
   **Result:**  
   $ \frac{1}{15}x^5 - 5x + C $

---

7. $\int (\sin^2 x + \cos^2 x) \,dx$  
   **Step 1:** Use the identity $\sin^2 x + \cos^2 x = 1$:  
   $ \int 1 \,dx = x + C $

---

8. $\int \left(5\sin x + 3e^x\right) \,dx$  
   **Step 1:** Integrate each term separately:  
   $ \int 5\sin x \,dx = -5\cos x, \quad \int 3e^x \,dx = 3e^x $  
   **Result:**  
   $ -5\cos x + 3e^x + C $

---

9. $\int x^3 \,dx$  
   **Step 1:** Apply the power rule:  
   $ \int x^3 \,dx = \frac{x^4}{4} $  
   **Result:**  
   $ \frac{x^4}{4} + C $

---

10. $\int 10x \,dx$  
   **Step 1:** Factor out the constant:  
   $ 10 \int x \,dx $  
   **Step 2:** Integrate $x$:  
   $ \int x \,dx = \frac{x^2}{2} $  
   **Result:**  
   $ 5x^2 + C $

---

11. $\int \cos\left(\frac{5}{2}x + 3\right) \,dx$  
   **Step 1:** Let $u = \frac{5}{2}x + 3$, so $du = \frac{5}{2}dx$ or $dx = \frac{2}{5}du$.  
   **Step 2:** Integrate:  
   $ \int \cos u \cdot \frac{2}{5} \,du = \frac{2}{5}\sin u + C $  
   Substitute $u$:  
   $ \frac{2}{5}\sin\left(\frac{5}{2}x + 3\right) + C $

---

12. $\int \frac{\cos(\ln x)}{x} \,dx$  
   **Step 1:** Let $u = \ln x$, so $du = \frac{1}{x}dx$.  
   **Step 2:** Integrate:  
   $ \int \cos u \,du = \sin u + C $  
   Substitute $u$:  
   $ \sin(\ln x) + C $

---

13. $\int x\ln x \,dx$  
   **Step 1:** Use integration by parts:  
   Let $u = \ln x$, $dv = x dx$.  
   Then, $du = \frac{1}{x} dx$, $v = \frac{x^2}{2}$.  
   **Step 2:** Apply the formula:  
   $ \int x\ln x \,dx = \frac{x^2}{2}\ln x - \int \frac{x^2}{2} \cdot \frac{1}{x} \,dx $  
   **Result:**  
   $ \frac{x^2}{2}\ln x - \frac{x^2}{4} + C $

---

14. $\int xe^x \,dx$  
   **Step 1:** Use integration by parts:  
   Let $u = x$, $dv = e^x dx$.  
   Then, $du = dx$, $v = e^x$.  
   **Step 2:** Apply the formula:  
   $ \int xe^x \,dx = xe^x - \int e^x \,dx $  
   **Result:**  
   $ xe^x - e^x + C = (x - 1)e^x + C $

# Area of the Region Bounded by Given Lines

We want to calculate the area of the region bounded by the lines:

- $x = 1$,  
- $x = 2$,  
- $y = 0$,  
- $y = x^2 + 1$.

---

### Step 1: Define the Problem
The region is bounded vertically between $x = 1$ and $x = 2$, and horizontally by $y = 0$ (the x-axis) and the curve $y = x^2 + 1$.

### Step 2: Set Up the Integral
We find the area under the curve $y = x^2 + 1$ from $x = 1$ to $x = 2$, and subtract the area under $y = 0$ (which is zero):

$$
\text{Area} = \int_1^2 (x^2 + 1) \,dx
$$

---

### Step 3: Solve the Integral
$$
\int_1^2 (x^2 + 1) \,dx = \int_1^2 x^2 \,dx + \int_1^2 1 \,dx
$$

First, compute each term:
$$
\int_1^2 x^2 \,dx = \left.\frac{x^3}{3}\right|_1^2 = \frac{2^3}{3} - \frac{1^3}{3} = \frac{8}{3} - \frac{1}{3} = \frac{7}{3}
$$

$$
\int_1^2 1 \,dx = (2 - 1) = 1
$$

Adding the results:
$$
\text{Area} = \frac{7}{3} + 1 = \frac{7}{3} + \frac{3}{3} = \frac{10}{3}
$$

# Area Under the Sine Curve over the Interval $[0, \pi]$

We want to calculate the area under the sine curve over the interval $[0, \pi]$.

The integral to compute the area is:

$$
P = \int_0^\pi \sin(x) \,dx
$$

---

### Step 1: Compute the Integral of $\sin(x)$

We need to evaluate:

$$
\int \sin(x) \,dx
$$

The integral of $\sin(x)$ is:

$$
\int \sin(x) \,dx = -\cos(x)
$$

---

### Step 2: Evaluate the Definite Integral

Now, we evaluate the definite integral from $x = 0$ to $x = \pi$:

$$
P = \left[ -\cos(x) \right]_0^\pi
$$

Substitute the limits:

$$
P = -\cos(\pi) + \cos(0)
$$

Since $\cos(\pi) = -1$ and $\cos(0) = 1$, we get:

$$
P = -(-1) + 1 = 1 + 1 = 2
$$

---

### Final Answer

The area under the sine curve over the interval $[0, \pi]$ is:

$$
P = 2
$$


# Length of the Sine Curve over the Interval $[0, \pi]$

$$
L = \int_a^b \sqrt{1 + (f'(x))^2} \,dx
$$

---

### Step 1: Find the Derivative of $f(x) = \sin(x)$

$$
f'(x) = \cos(x)
$$

Now substitute into the formula for the length of the curve:

$$
L = \int_0^\pi \sqrt{1 + \cos^2(x)} \,dx
$$

---

### Step 2: Evaluate the Integral

The integral we need to evaluate is:

$$
L = \int_0^\pi \sqrt{1 + \cos^2(x)} \,dx
$$

This integral does not have a simple elementary antiderivative, so it must be evaluated numerically. You can approximate it using numerical methods (e.g., Simpson's rule, Trapezoidal rule) or with a calculator.

---

### Final Answer

The length of the sine curve over the interval $[0, \pi]$ is approximately:

$$
L \approx 3.820
$$

This is the numerical approximation of the integral. The exact value requires numerical integration methods, as the integral cannot be expressed in terms of elementary functions.


# Distance Traveled by a Moving Particle

We are given the position function of a particle:

$$
x(t) = 3t^2 - 6t + 1
$$

We need to calculate the distance the particle travels between $t = 0$ and $t = 2$.

---

### Step 1: Find the Velocity Function

The velocity function is the derivative of the position function with respect to time:

$$
v(t) = \frac{d}{dt} \left( 3t^2 - 6t + 1 \right)
$$

Differentiating the position function:

$$
v(t) = 6t - 6
$$

---

### Step 2: Find When the Particle Changes Direction

To determine when the particle changes direction, we solve for when the velocity is zero:

$$
v(t) = 0 \quad \Rightarrow \quad 6t - 6 = 0 \quad \Rightarrow \quad t = 1
$$

This indicates that the particle changes direction at $t = 1$. Thus, we need to calculate the distance traveled in two segments: from $t = 0$ to $t = 1$ and from $t = 1$ to $t = 2$.

---

### Step 3: Calculate the Distance Traveled in Each Segment

The distance traveled is the integral of the speed function (which is the absolute value of the velocity).

#### From $t = 0$ to $t = 1$:

For $0 \leq t \leq 1$, the velocity is negative (since $v(t) = 6t - 6$), so the speed is the negative of the velocity.

$$
\text{Distance}_1 = \int_0^1 \left| 6t - 6 \right| \,dt = \int_0^1 -(6t - 6) \,dt
$$

Simplify the integral:

$$
\text{Distance}_1 = \int_0^1 (6 - 6t) \,dt
$$

Evaluate the integral:

$$
\text{Distance}_1 = \left[ 6t - 3t^2 \right]_0^1 = \left(6(1) - 3(1)^2\right) - \left(6(0) - 3(0)^2\right)
$$

$$
\text{Distance}_1 = (6 - 3) - 0 = 3
$$

#### From $t = 1$ to $t = 2$:

For $1 \leq t \leq 2$, the velocity is positive (since $v(t) = 6t - 6$), so the speed is the same as the velocity.

$$
\text{Distance}_2 = \int_1^2 (6t - 6) \,dt
$$

Evaluate the integral:

$$
\text{Distance}_2 = \left[ 3t^2 - 6t \right]_1^2 = \left(3(2)^2 - 6(2)\right) - \left(3(1)^2 - 6(1)\right)
$$

$$
\text{Distance}_2 = (12 - 12) - (3 - 6) = 0 + 3 = 3
$$

---

### Step 4: Total Distance Traveled

The total distance traveled is the sum of the distances in each segment:

$$
\text{Total Distance} = \text{Distance}_1 + \text{Distance}_2 = 3 + 3 = 6
$$

---

### Final Answer

The total distance traveled by the particle between $t = 0$ and $t = 2$ is:

$$
\text{Total Distance} = 6 \, \text{units}
$$
