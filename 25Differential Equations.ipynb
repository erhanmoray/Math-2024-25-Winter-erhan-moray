# Differential Equations

## Introduction
Differential equations are mathematical equations that relate a function to its derivatives. They are widely used in engineering, physics, economics, and biology to model dynamic systems.

---

## Types of Differential Equations

### 1. Ordinary Differential Equations (ODEs)
An ODE involves derivatives of a function with respect to a single variable.

Example of a first-order linear ODE:  
$$
\frac{dy}{dx} + P(x)y = Q(x)
$$

---

### 2. Partial Differential Equations (PDEs)
A PDE involves partial derivatives with respect to more than one variable.

Example:  
$$
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0
$$

---

## Methods of Solving ODEs

### 1. Separation of Variables
This technique is applicable when variables can be separated on opposite sides of the equation.

Example:  
$$
\frac{dy}{dx} = xy
$$

Solution:  
$$
\int \frac{dy}{y} = \int x \, dx
$$

### 2. Integrating Factor Method
Used for first-order linear ODEs of the form:  
$$
\frac{dy}{dx} + P(x)y = Q(x)
$$

The integrating factor is:  
$$
I(x) = e^{\int P(x) \, dx}
$$

---

## Example Problem
Solve the differential equation:  
$$
\frac{dy}{dx} + 3y = 6x
$$

### Step 1: Identify $P(x)$ and $Q(x)$
- $P(x) = 3$
- $Q(x) = 6x$

### Step 2: Compute the integrating factor
$$
I(x) = e^{\int 3 \, dx} = e^{3x}
$$

### Step 3: Multiply the entire equation by $I(x)$
$$
e^{3x} \frac{dy}{dx} + 3e^{3x}y = 6xe^{3x}
$$

### Step 4: Solve by integrating both sides
$$
\int d(e^{3x}y) = \int 6xe^{3x} \, dx
$$

---

## Applications
- **Physics**: Newton's law of cooling.
- **Economics**: Growth models and interest rate changes.
- **Biology**: Population dynamics.

---

## Conclusion
Differential equations help model and solve problems involving change over time or space, forming the foundation for many fields of science and engineering.



# Solving First-Order Ordinary Differential Equations

## Problem 1

Solve the differential equation:  
$$
y'(x) = y
$$

### Solution

1. The equation is separable:  
   $$ \frac{dy}{dx} = y $$
2. Separate variables:  
   $$ \frac{dy}{y} = dx $$
3. Integrate both sides:  
   $$ \int \frac{dy}{y} = \int dx $$
   $$ \ln |y| = x + C $$
4. Solving for $y$:  
   $$ y(x) = e^{x + C} = Ae^x $$  
   where $A = e^C$.

---

## Problem 2

Solve the differential equation:  
$$
y'(x) = \frac{1}{2} y(x)
$$

### Solution

1. The equation is linear:  
   $$ \frac{dy}{dx} = \frac{1}{2} y $$
2. Separate variables:  
   $$ \frac{dy}{y} = \frac{1}{2} dx $$
3. Integrate both sides:  
   $$ \int \frac{dy}{y} = \int \frac{1}{2} dx $$
   $$ \ln |y| = \frac{1}{2} x + C $$
4. Solving for $y$:  
   $$ y(x) = e^{\frac{1}{2} x + C} = Ae^{\frac{1}{2} x} $$  
   where $A = e^C$.

---

## Summary
The general solutions are:
- For $y'(x) = y$, the solution is $y(x) = Ae^x$.
- For $y'(x) = \frac{1}{2} y(x)$, the solution is $y(x) = Ae^{\frac{1}{2} x}$.



# Solving First-Order Ordinary Differential Equations Using Separation of Variables

## Problem 1

Solve the differential equation:  
$$
\frac{dy}{dx} = xy
$$

### Solution

1. Separate the variables:  
   $$ \frac{dy}{y} = x dx $$
2. Integrate both sides:  
   $$ \int \frac{dy}{y} = \int x \, dx $$
   $$ \ln |y| = \frac{x^2}{2} + C $$
3. Solving for $y$:  
   $$ y(x) = e^{\frac{x^2}{2} + C} = Ae^{\frac{x^2}{2}} $$  
   where $A = e^C$.

---

## Problem 2

Solve the differential equation:  
$$
\frac{dy}{dx} = \frac{y}{x}
$$

### Solution

1. Separate the variables:  
   $$ \frac{dy}{y} = \frac{dx}{x} $$
2. Integrate both sides:  
   $$ \int \frac{dy}{y} = \int \frac{dx}{x} $$
   $$ \ln |y| = \ln |x| + C $$
3. Solving for $y$:  
   $$ y(x) = e^{\ln |x| + C} = Ax $$  
   where $A = e^C$.

---

## Problem 3

Solve the differential equation:  
$$
\frac{dy}{dx} = yx
$$

### Solution

1. Separate the variables:  
   $$ \frac{dy}{y} = x \, dx $$
2. Integrate both sides:  
   $$ \int \frac{dy}{y} = \int x \, dx $$
   $$ \ln |y| = \frac{x^2}{2} + C $$
3. Solving for $y$:  
   $$ y(x) = e^{\frac{x^2}{2} + C} = Ae^{\frac{x^2}{2}} $$  
   where $A = e^C$.

---

## Summary
The general solutions are:
- For $\frac{dy}{dx} = xy$, the solution is $y(x) = Ae^{\frac{x^2}{2}}$.
- For $\frac{dy}{dx} = \frac{y}{x}$, the solution is $y(x) = Ax$.
- For $\frac{dy}{dx} = yx$, the solution is $y(x) = Ae^{\frac{x^2}{2}}$.


# Solving Second-Order Ordinary Differential Equations

---

## Problem 1

Solve the differential equation:  
$$
y''(x) + y'(x) = 0,
$$  
with boundary conditions $y(0) = 2$ and $y'(0) = -1$.

### Solution

1. Assume a solution of the form $y(x) = e^{rx}$.
2. Substituting $y''(x) = r^2 e^{rx}$ and $y'(x) = re^{rx}$ into the equation:  
   $$
   r^2 e^{rx} + re^{rx} = 0 \quad \Rightarrow \quad r(r + 1) = 0.
   $$
3. The roots are $r = 0$ and $r = -1$.
4. The general solution is:  
   $$
   y(x) = C_1 + C_2 e^{-x}.
   $$
5. Apply the boundary conditions:  
   - $y(0) = 2 \Rightarrow C_1 + C_2 = 2$,
   - $y'(x) = -C_2 e^{-x}$, and $y'(0) = -1 \Rightarrow -C_2 = -1 \Rightarrow C_2 = 1$.
6. Therefore, $C_1 = 2 - 1 = 1$.

**Solution**:  
$$
y(x) = 1 + e^{-x}.
$$

---

## Problem 2

Solve the differential equation:  
$$
y''(x) - y(x) = 0,
$$  
with boundary conditions $y(0) = 2$ and $y'(0) = 0$.

### Solution

1. Assume a solution of the form $y(x) = e^{rx}$.
2. Substituting $y''(x) = r^2 e^{rx}$ and $y(x) = e^{rx}$ into the equation:  
   $$
   r^2 - 1 = 0 \quad \Rightarrow \quad r = \pm 1.
   $$
3. The general solution is:  
   $$
   y(x) = C_1 e^x + C_2 e^{-x}.
   $$
4. Apply the boundary conditions:  
   - $y(0) = 2 \Rightarrow C_1 + C_2 = 2$,
   - $y'(0) = C_1 e^x - C_2 e^{-x} \big|_{x=0} = C_1 - C_2 = 0 \Rightarrow C_1 = C_2$.
5. Therefore, $C_1 = C_2 = 1$.

**Solution**:  
$$
y(x) = e^x + e^{-x}.
$$

---

## Problem 3

Solve the differential equation:  
$$
\frac{d^2 y}{dx^2} = -\omega^2 y(x).
$$

### Solution

1. Assume a solution of the form $y(x) = A \cos(\omega x) + B \sin(\omega x)$.
2. This form satisfies the equation since:  
   $$
   y''(x) = -\omega^2 A \cos(\omega x) - \omega^2 B \sin(\omega x).
   $$

**General Solution**:  
$$
y(x) = A \cos(\omega x) + B \sin(\omega x),
$$  
where $A$ and $B$ are determined by initial or boundary conditions.

---

## Summary
- For $y''(x) + y'(x) = 0$, the solution is $y(x) = 1 + e^{-x}$.
- For $y''(x) - y(x) = 0$, the solution is $y(x) = e^x + e^{-x}$.
- For $y''(x) = -\omega^2 y(x)$, the general solution is $y(x) = A \cos(\omega x) + B \sin(\omega x)$.


# Checking if the Function is a Solution to the Wave Equation

Given the function:
$$
\psi(t, x) = A \cos(\omega t + kx),
$$
we want to check if it satisfies the second-order partial differential equation (the wave equation):
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0,
$$
where \( v = \frac{\omega}{k} \), \( \omega = \frac{2\pi}{T} \), and \( k = \frac{2\pi}{\lambda} \).

---

## Step 1: Compute the Second Derivative with Respect to \( t \)

Start with the function:
$$
\psi(t, x) = A \cos(\omega t + kx).
$$

First, calculate the first derivative with respect to \( t \):
$$
\frac{\partial \psi(t, x)}{\partial t} = -A \omega \sin(\omega t + kx).
$$

Now, compute the second derivative with respect to \( t \):
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} = -A \omega^2 \cos(\omega t + kx).
$$

---

## Step 2: Compute the Second Derivative with Respect to \( x \)

Next, calculate the first derivative with respect to \( x \):
$$
\frac{\partial \psi(t, x)}{\partial x} = -A k \sin(\omega t + kx).
$$

Now, compute the second derivative with respect to \( x \):
$$
\frac{\partial^2 \psi(t, x)}{\partial x^2} = -A k^2 \cos(\omega t + kx).
$$

---

## Step 3: Substitute into the Wave Equation

Substitute the second derivatives into the wave equation:
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

From Step 1 and Step 2:
$$
-A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0.
$$

Simplifying:
$$
-A \omega^2 \cos(\omega t + kx) + A v^2 k^2 \cos(\omega t + kx) = 0.
$$

Factor out \( A \cos(\omega t + kx) \):
$$
A \cos(\omega t + kx) (\omega^2 - v^2 k^2) = 0.
$$

For this to hold true for all \( t \) and \( x \), we must have:
$$
\omega^2 = v^2 k^2.
$$

Since \( v = \frac{\omega}{k} \), we have:
$$
\omega^2 = \left( \frac{\omega}{k} \right)^2 k^2 = \omega^2.
$$

Thus, the wave equation is satisfied.

---

## Conclusion

The function \( \psi(t, x) = A \cos(\omega t + kx) \) is indeed a solution to the wave equation:
$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$
