# Determinants of Matrices

### 1. **Determinant of a 2x2 Matrix**

The determinant of a 2x2 matrix is calculated as follows:

For a matrix
$$
A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}
$$

The determinant of \( A \), denoted as \( \det(A) \), is given by:

$$
\det(A) = ad - bc
$$

**Example:**

For matrix \( A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \), the determinant is:

$$
\det(A) = 1(4) - 2(3) = 4 - 6 = -2
$$

---

### 2. **Determinant of a 3x3 Matrix**

The determinant of a 3x3 matrix is calculated using cofactor expansion along the first row. For a matrix

$$
B = \begin{pmatrix} a & b & c \\ d & e & f \\ g & h & i \end{pmatrix}
$$

The determinant of \( B \), denoted as \( \det(B) \), is given by:

$$
\det(B) = a \begin{vmatrix} e & f \\ h & i \end{vmatrix} - b \begin{vmatrix} d & f \\ g & i \end{vmatrix} + c \begin{vmatrix} d & e \\ g & h \end{vmatrix}
$$

Where each of the 2x2 determinants is calculated as follows:

$$
\begin{vmatrix} e & f \\ h & i \end{vmatrix} = ei - fh
$$

$$
\begin{vmatrix} d & f \\ g & i \end{vmatrix} = di - fg
$$

$$
\begin{vmatrix} d & e \\ g & h \end{vmatrix} = dh - eg
$$

**Example:**

For matrix \( B = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{pmatrix} \), the determinant is:

$$
\det(B) = 1 \begin{vmatrix} 5 & 6 \\ 8 & 9 \end{vmatrix} - 2 \begin{vmatrix} 4 & 6 \\ 7 & 9 \end{vmatrix} + 3 \begin{vmatrix} 4 & 5 \\ 7 & 8 \end{vmatrix}
$$

First, calculate the 2x2 determinants:

$$
\begin{vmatrix} 5 & 6 \\ 8 & 9 \end{vmatrix} = (5)(9) - (6)(8) = 45 - 48 = -3
$$

$$
\begin{vmatrix} 4 & 6 \\ 7 & 9 \end{vmatrix} = (4)(9) - (6)(7) = 36 - 42 = -6
$$

$$
\begin{vmatrix} 4 & 5 \\ 7 & 8 \end{vmatrix} = (4)(8) - (5)(7) = 32 - 35 = -3
$$

Now, substitute these values back into the formula:

$$
\det(B) = 1(-3) - 2(-6) + 3(-3) = -3 + 12 - 9 = 0
$$

So, \( \det(B) = 0 \).

---

### Conclusion

- Determinant of 2x2 matrix: \( A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix} \) is \( \det(A) = -2 \)
- Determinant of 3x3 matrix: \( B = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{pmatrix} \) is \( \det(B) = 0 \)

# Determinants of 3x3 Matrices

### 1. **Matrix D**

For the matrix
$$
D = \begin{pmatrix} 1 & 0 & 2 \\ -1 & 3 & 1 \\ 2 & 4 & -2 \end{pmatrix}
$$

The determinant of \( D \), denoted as \( \det(D) \), is calculated using cofactor expansion:

$$
\det(D) = 1 \begin{vmatrix} 3 & 1 \\ 4 & -2 \end{vmatrix} - 0 \begin{vmatrix} -1 & 1 \\ 2 & -2 \end{vmatrix} + 2 \begin{vmatrix} -1 & 3 \\ 2 & 4 \end{vmatrix}
$$

First, calculate the 2x2 determinants:

$$
\begin{vmatrix} 3 & 1 \\ 4 & -2 \end{vmatrix} = (3)(-2) - (1)(4) = -6 - 4 = -10
$$

$$
\begin{vmatrix} -1 & 3 \\ 2 & 4 \end{vmatrix} = (-1)(4) - (3)(2) = -4 - 6 = -10
$$

Now, substitute back into the formula:

$$
\det(D) = 1(-10) + 0 + 2(-10) = -10 + 0 - 20 = -30
$$

---

### 2. **Matrix E**

For the matrix
$$
E = \begin{pmatrix} 3 & 1 & -1 \\ 0 & 2 & 4 \\ 5 & 3 & 2 \end{pmatrix}
$$

The determinant of \( E \), denoted as \( \det(E) \), is calculated using cofactor expansion:

$$
\det(E) = 3 \begin{vmatrix} 2 & 4 \\ 3 & 2 \end{vmatrix} - 1 \begin{vmatrix} 0 & 4 \\ 5 & 2 \end{vmatrix} + (-1) \begin{vmatrix} 0 & 2 \\ 5 & 3 \end{vmatrix}
$$

First, calculate the 2x2 determinants:

$$
\begin{vmatrix} 2 & 4 \\ 3 & 2 \end{vmatrix} = (2)(2) - (4)(3) = 4 - 12 = -8
$$

$$
\begin{vmatrix} 0 & 4 \\ 5 & 2 \end{vmatrix} = (0)(2) - (4)(5) = 0 - 20 = -20
$$

$$
\begin{vmatrix} 0 & 2 \\ 5 & 3 \end{vmatrix} = (0)(3) - (2)(5) = 0 - 10 = -10
$$

Now, substitute back into the formula:

$$
\det(E) = 3(-8) - 1(-20) + (-1)(-10) = -24 + 20 + 10 = 6
$$

---

### 3. **Matrix F**

For the matrix
$$
F = \begin{pmatrix} 2 & -3 & 1 \\ 1 & 4 & -2 \\ 1 & 5 & 3 \end{pmatrix}
$$

The determinant of \( F \), denoted as \( \det(F) \), is calculated using cofactor expansion:

$$
\det(F) = 2 \begin{vmatrix} 4 & -2 \\ 5 & 3 \end{vmatrix} - (-3) \begin{vmatrix} 1 & -2 \\ 1 & 3 \end{vmatrix} + 1 \begin{vmatrix} 1 & 4 \\ 1 & 5 \end{vmatrix}
$$

First, calculate the 2x2 determinants:

$$
\begin{vmatrix} 4 & -2 \\ 5 & 3 \end{vmatrix} = (4)(3) - (-2)(5) = 12 + 10 = 22
$$

$$
\begin{vmatrix} 1 & -2 \\ 1 & 3 \end{vmatrix} = (1)(3) - (-2)(1) = 3 + 2 = 5
$$

$$
\begin{vmatrix} 1 & 4 \\ 1 & 5 \end{vmatrix} = (1)(5) - (4)(1) = 5 - 4 = 1
$$

Now, substitute back into the formula:

$$
\det(F) = 2(22) + 3(5) + 1(1) = 44 + 15 + 1 = 60
$$

---

### Conclusion

- Determinant of matrix \( D = \begin{pmatrix} 1 & 0 & 2 \\ -1 & 3 & 1 \\ 2 & 4 & -2 \end{pmatrix} \) is \( \det(D) = -30 \)
- Determinant of matrix \( E = \begin{pmatrix} 3 & 1 & -1 \\ 0 & 2 & 4 \\ 5 & 3 & 2 \end{pmatrix} \) is \( \det(E) = 6 \)
- Determinant of matrix \( F = \begin{pmatrix} 2 & -3 & 1 \\ 1 & 4 & -2 \\ 1 & 5 & 3 \end{pmatrix} \) is \( \det(F) = 60 \)
