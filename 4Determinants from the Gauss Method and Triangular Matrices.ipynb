# Determinants from the Gauss Method and Triangular Matrices

## Gaussian Elimination
Gaussian elimination is a method used to transform a matrix into an **upper triangular form** to simplify the determinant calculation.

### Steps
1. Apply row operations to the matrix until it becomes an upper triangular matrix (all elements below the main diagonal are zero).
   - **Row operations** that do not change the determinant:
     - Swap two rows: Multiply the determinant by $-1$.
     - Multiply a row by a scalar $k$: Multiply the determinant by $k$.
     - Add or subtract a multiple of one row to another row: No effect on the determinant.
2. The determinant of a triangular matrix is the product of its diagonal elements.

---

## Example
### Given Matrix
$$
A = \begin{bmatrix}
12 & 3 & -18 & -4 \\
1 & 2 & 3 & 4 \\
0 & 5 & 6 & 7 \\
\end{bmatrix}
$$

### Step 1: Use row operations to get an upper triangular form
1. Divide row 1 by 12 to make the leading coefficient 1.
   $$
   R_1 \to \frac{1}{12} R_1
   $$
   New $R_1 = \begin{bmatrix} 1 & \frac{1}{4} & -\frac{3}{2} & -\frac{1}{3} \end{bmatrix}$
   
2. Subtract appropriate multiples of $R_1$ from $R_2$ and $R_3$ to make the first column of $R_2$ and $R_3$ zero.

3. Continue this process until $A$ is in upper triangular form.

---

### Determinant
For an upper triangular matrix:
$$
\det(A) = a_{11} \cdot a_{22} \cdot a_{33} \cdots a_{nn}
$$
where $a_{ii}$ are the diagonal elements.

---

## Triangular Matrices
### Types
1. **Upper Triangular Matrix**: All elements below the main diagonal are zero.
   $$
   \begin{bmatrix}
   a_{11} & a_{12} & a_{13} \\
   0 & a_{22} & a_{23} \\
   0 & 0 & a_{33}
   \end{bmatrix}
   $$
   Determinant: $\det(A) = a_{11} a_{22} a_{33}$

2. **Lower Triangular Matrix**: All elements above the main diagonal are zero.
   $$
   \begin{bmatrix}
   a_{11} & 0 & 0 \\
   a_{21} & a_{22} & 0 \\
   a_{31} & a_{32} & a_{33}
   \end{bmatrix}
   $$
   Determinant: $\det(A) = a_{11} a_{22} a_{33}$
   
---

## Example
### Upper Triangular Matrix
$$
B = \begin{bmatrix}
1 & 2 & 3 \\
0 & 5 & 6 \\
0 & 0 & 9
\end{bmatrix}
$$

### Determinant of $B$
$$
\det(B) = 1 \times 5 \times 9 = 45
$$


# Row and Column Operations for Upper Triangular Form

We will use row and column operations to reduce the given matrices to an **upper triangular form** and calculate their determinants by multiplying the diagonal elements.

---

## Example 1
### Given Matrix
$$
A = \begin{bmatrix}
12 & 3 \\
-18 & -4
\end{bmatrix}
$$

### Step 1: Make the first element of the first row $a_{11} = 12$.

### Step 2: Eliminate the first element of the second row by subtracting $\frac{-18}{12}$ times the first row from the second row:
$$
R_2 \to R_2 - \frac{-18}{12} R_1
$$
This simplifies row 2.

### Step 3: Calculate the determinant
For a $2 \times 2$ matrix:
$$
\det(A) = a_{11} \cdot a_{22} - a_{12} \cdot a_{21}
$$
Substitute the values from matrix $A$:
$$
\det(A) = (12)(-4) - (3)(-18) = -48 + 54 = 6
$$

## Example 2
### Given Matrix
$$
B = \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}
$$

### Step 1: Use row operations to make the first column of row 2 and row 3 zero.

1. Subtract $4 \times R_1$ from $R_2$:
   $$
   R_2 \to R_2 - 4R_1
   $$
2. Subtract $7 \times R_1$ from $R_3$:
   $$
   R_3 \to R_3 - 7R_1
   $$

### Step 2: Apply similar operations on row 2 and row 3 to form an upper triangular matrix.

### Step 3: Calculate the determinant of the upper triangular matrix by multiplying the diagonal elements:
$$
\det(B) = a_{11} \cdot a_{22} \cdot a_{33}
$$
Substitute values after performing all row operations.

---

This method simplifies determinant calculation by transforming the matrix step-by-step into an upper triangular form.
