# Inverse of a Matrix using the Formula

## Introduction

The inverse of a square matrix \( A \) is denoted as \( A^{-1} \), and it is the matrix that, when multiplied by \( A \), results in the identity matrix \( I \). Not all matrices have an inverse. A matrix is invertible if and only if its determinant is non-zero.

The formula for the inverse of a 2x2 or 3x3 matrix can be calculated using the following steps.

### Formula for Inverse of a 2x2 Matrix

For a 2x2 matrix:
$$
A = \begin{pmatrix} a & b \\ c & d \end{pmatrix}
$$
The inverse is given by the formula:
$$
A^{-1} = \frac{1}{\det(A)} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix}
$$
where \( \det(A) = ad - bc \).

### Formula for Inverse of a 3x3 Matrix

For a 3x3 matrix:
$$
A = \begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix}
$$
The inverse is given by the formula:
$$
A^{-1} = \frac{1}{\det(A)} \begin{pmatrix}
\det(A_{11}) & -\det(A_{12}) & \det(A_{13}) \\
-\det(A_{21}) & \det(A_{22}) & -\det(A_{23}) \\
\det(A_{31}) & -\det(A_{32}) & \det(A_{33})
\end{pmatrix}
$$
where:
- \( A_{ij} \) are the 2x2 minors formed by eliminating row \( i \) and column \( j \) from the original matrix.
- \( \det(A) \) is the determinant of matrix \( A \).

## Step-by-Step Process

### Step 1: Calculate the Determinant

Before proceeding with the inverse, calculate the determinant of the matrix. If the determinant is zero, the matrix does not have an inverse.

#### Example 1: 2x2 Matrix

Given matrix \( A \):
$$
A = \begin{pmatrix}
3 & 2 \\
1 & 4
\end{pmatrix}
$$

Calculate the determinant:
$$
\det(A) = (3)(4) - (2)(1) = 12 - 2 = 10
$$

Since the determinant is non-zero, we can proceed with the inverse calculation.

#### Example 2: 3x3 Matrix

Given matrix \( A \):
$$
A = \begin{pmatrix}
2 & 1 & 3 \\
0 & 1 & 4 \\
1 & 2 & 5
\end{pmatrix}
$$

Calculate the determinant:
$$
\det(A) = 2 \begin{vmatrix} 1 & 4 \\ 2 & 5 \end{vmatrix} - 1 \begin{vmatrix} 0 & 4 \\ 1 & 5 \end{vmatrix} + 3 \begin{vmatrix} 0 & 1 \\ 1 & 2 \end{vmatrix}
$$
$$
= 2((1)(5) - (4)(2)) - 1((0)(5) - (4)(1)) + 3((0)(2) - (1)(1))
$$
$$
= 2(5 - 8) - 1(0 - 4) + 3(0 - 1)
$$
$$
= 2(-3) + 4 + 3(-1)
$$
$$
= -6 + 4 - 3 = -5
$$

Since the determinant is non-zero (\(\det(A) = -5\)), we can proceed with the inverse calculation.

### Step 2: Find the Minors and Cofactors (For 3x3 Matrices)

For the 3x3 matrix, calculate the minors and cofactors. A minor is the determinant of the submatrix obtained by removing a row and a column.

#### Minor for \( A_{11} \):
$$
A_{11} = \begin{vmatrix}
1 & 4 \\
2 & 5
\end{vmatrix} = (1)(5) - (4)(2) = 5 - 8 = -3
$$

#### Minor for \( A_{12} \):
$$
A_{12} = \begin{vmatrix}
0 & 4 \\
1 & 5
\end{vmatrix} = (0)(5) - (4)(1) = 0 - 4 = -4
$$

#### Minor for \( A_{13} \):
$$
A_{13} = \begin{vmatrix}
0 & 1 \\
1 & 2
\end{vmatrix} = (0)(2) - (1)(1) = 0 - 1 = -1
$$

### Step 3: Construct the Cofactor Matrix

Using the minors, we create the cofactor matrix by applying the appropriate signs based on the position:

$$
\text{Cofactor matrix} = \begin{pmatrix}
-3 & 4 & -1 \\
4 & -5 & 1 \\
-2 & 3 & -1
\end{pmatrix}
$$

### Step 4: Calculate the Adjugate Matrix

The adjugate matrix is the transpose of the cofactor matrix:

$$
\text{Adjugate matrix} = \begin{pmatrix}
-3 & 4 & -2 \\
4 & -5 & 3 \\
-1 & 1 & -1
\end{pmatrix}
$$

### Step 5: Multiply by the Reciprocal of the Determinant

The inverse is obtained by multiplying the adjugate matrix by \( \frac{1}{\det(A)} \). For matrix \( A \), we have \( \det(A) = -5 \), so:

$$
A^{-1} = \frac{1}{-5} \begin{pmatrix}
-3 & 4 & -2 \\
4 & -5 & 3 \\
-1 & 1 & -1
\end{pmatrix}
$$

Multiply the matrix by \( \frac{1}{-5} \):

$$
A^{-1} = \begin{pmatrix}
\frac{3}{5} & -\frac{4}{5} & \frac{2}{5} \\
-\frac{4}{5} & \frac{5}{5} & -\frac{3}{5} \\
\frac{1}{5} & -\frac{1}{5} & \frac{1}{5}
\end{pmatrix}
$$

---

## Conclusion

The inverse of a matrix can be calculated using the formula for 2x2 and 3x3 matrices. For larger matrices, the process involves finding the minors, cofactors, and adjugate matrix, then multiplying by the reciprocal of the determinant.

- **For the 2x2 matrix**: \( A^{-1} = \frac{1}{\det(A)} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix} \)
- **For the 3x3 matrix**: The process involves finding the minors, cofactors, adjugate matrix, and multiplying by \( \frac{1}{\det(A)} \).



# Finding the Inverse of Matrix A and Verification

## Given Matrix A:

$$
A = \begin{pmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{pmatrix}
$$

### Step 1: Calculate the Determinant of Matrix A

The formula for the determinant of a 3x3 matrix is:
$$
\det(A) = a(ei - fh) - b(di - fg) + c(dh - eg)
$$

For matrix \( A \), the elements are:
$$
A = \begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix}
= \begin{pmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{pmatrix}
$$

Using the formula:
$$
\det(A) = 2 \left( (1)(0) - (0)(2) \right) - 0 \left( (0)(0) - (0)(1) \right) + 1 \left( (0)(2) - (1)(1) \right)
$$

Simplifying:
$$
\det(A) = 2(0) - 0 + 1(0 - 1) = 0 - 1 = -1
$$

Since the determinant is \( \det(A) = -1 \), the matrix is invertible.

### Step 2: Calculate the Cofactor Matrix

Now we need to calculate the minors and cofactors for the 3x3 matrix. A minor is the determinant of the 2x2 submatrix obtained by removing the row and column of a specific element.

#### Minor for \( M_{11} \):
Remove the first row and first column:
$$
M_{11} = \begin{vmatrix}
1 & 0 \\
2 & 0
\end{vmatrix} = (1)(0) - (0)(2) = 0
$$

#### Minor for \( M_{12} \):
Remove the first row and second column:
$$
M_{12} = \begin{vmatrix}
0 & 0 \\
1 & 0
\end{vmatrix} = (0)(0) - (0)(1) = 0
$$

#### Minor for \( M_{13} \):
Remove the first row and third column:
$$
M_{13} = \begin{vmatrix}
0 & 1 \\
1 & 2
\end{vmatrix} = (0)(2) - (1)(1) = -1
$$

#### Minor for \( M_{21} \):
Remove the second row and first column:
$$
M_{21} = \begin{vmatrix}
0 & 1 \\
2 & 0
\end{vmatrix} = (0)(0) - (1)(2) = -2
$$

#### Minor for \( M_{22} \):
Remove the second row and second column:
$$
M_{22} = \begin{vmatrix}
2 & 1 \\
1 & 0
\end{vmatrix} = (2)(0) - (1)(1) = -1
$$

#### Minor for \( M_{23} \):
Remove the second row and third column:
$$
M_{23} = \begin{vmatrix}
2 & 0 \\
1 & 2
\end{vmatrix} = (2)(2) - (0)(1) = 4
$$

#### Minor for \( M_{31} \):
Remove the third row and first column:
$$
M_{31} = \begin{vmatrix}
0 & 1 \\
1 & 0
\end{vmatrix} = (0)(0) - (1)(1) = -1
$$

#### Minor for \( M_{32} \):
Remove the third row and second column:
$$
M_{32} = \begin{vmatrix}
2 & 1 \\
0 & 0
\end{vmatrix} = (2)(0) - (1)(0) = 0
$$

#### Minor for \( M_{33} \):
Remove the third row and third column:
$$
M_{33} = \begin{vmatrix}
2 & 0 \\
0 & 1
\end{vmatrix} = (2)(1) - (0)(0) = 2
$$

### Step 3: Construct the Cofactor Matrix

The cofactor matrix is formed by applying the sign pattern:

$$
\text{Cofactor matrix} = \begin{pmatrix}
0 & 0 & -1 \\
2 & -1 & -4 \\
-1 & 0 & 2
\end{pmatrix}
$$

### Step 4: Calculate the Adjugate Matrix

The adjugate matrix is the transpose of the cofactor matrix:

$$
\text{Adjugate matrix} = \begin{pmatrix}
0 & 2 & -1 \\
0 & -1 & 0 \\
-1 & -4 & 2
\end{pmatrix}
$$

### Step 5: Calculate the Inverse of Matrix A

To find the inverse, multiply the adjugate matrix by \( \frac{1}{\det(A)} \). Since \( \det(A) = -1 \), the inverse is:

$$
A^{-1} = \frac{1}{-1} \begin{pmatrix}
0 & 2 & -1 \\
0 & -1 & 0 \\
-1 & -4 & 2
\end{pmatrix}
$$

Thus, the inverse is:

$$
A^{-1} = \begin{pmatrix}
0 & -2 & 1 \\
0 & 1 & 0 \\
1 & 4 & -2
\end{pmatrix}
$$

---

## Step 6: Verify the Result

To verify the result, we multiply \( A \) and \( A^{-1} \) and check if we obtain the identity matrix:

$$
A \times A^{-1} = \begin{pmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{pmatrix}
\times
\begin{pmatrix}
0 & -2 & 1 \\
0 & 1 & 0 \\
1 & 4 & -2
\end{pmatrix}
$$

The result is the identity matrix:

$$
I = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix}
$$

This confirms that the inverse is correct.

---

## Conclusion

The inverse of matrix \( A \) is:

$$
A^{-1} = \begin{pmatrix}
0 & -2 & 1 \\
0 & 1 & 0 \\
1 & 4 & -2
\end{pmatrix}
$$

The verification shows that the result is correct.



# Determining the Rank of Matrix B

## Given Matrix B:

$$
B = \begin{pmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{pmatrix}
$$

### Step 1: Perform Row Operations to Convert to Row Echelon Form (REF)

We need to perform row operations to transform the matrix to row echelon form. The goal is to have leading 1's in each row, and all elements below the leading 1's to be 0's.

Start with the matrix \( B \):

$$
B = \begin{pmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{pmatrix}
$$

#### Row Operation 1: Make the first element of the first column equal to 1

To do this, divide the first row by 4:

$$
R_1 \rightarrow \frac{1}{4} R_1
$$

After this operation, the matrix becomes:

$$
B = \begin{pmatrix}
1 & -\frac{3}{4} & \frac{7}{4} \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{pmatrix}
$$

#### Row Operation 2: Make the first element of the second and third rows equal to 0

Perform the following row operations:
- Add \( R_1 \) to \( R_2 \): \( R_2 \rightarrow R_2 + R_1 \)
- Subtract \( 2 \times R_1 \) from \( R_3 \): \( R_3 \rightarrow R_3 - 2R_1 \)

After these operations, the matrix becomes:

$$
B = \begin{pmatrix}
1 & -\frac{3}{4} & \frac{7}{4} \\
0 & \frac{21}{4} & \frac{19}{4} \\
0 & \frac{21}{2} & -\frac{5}{2}
\end{pmatrix}
$$

#### Row Operation 3: Make the second element of the second row equal to 1

To do this, multiply the second row by \( \frac{4}{21} \):

$$
R_2 \rightarrow \frac{4}{21} R_2
$$

The matrix becomes:

$$
B = \begin{pmatrix}
1 & -\frac{3}{4} & \frac{7}{4} \\
0 & 1 & \frac{19}{21} \\
0 & \frac{21}{2} & -\frac{5}{2}
\end{pmatrix}
$$

#### Row Operation 4: Make the second element of the third row equal to 0

Perform the following row operation:
- Subtract \( \frac{21}{2} \times R_2 \) from \( R_3 \): \( R_3 \rightarrow R_3 - \frac{21}{2} R_2 \)

After this operation, the matrix becomes:

$$
B = \begin{pmatrix}
1 & -\frac{3}{4} & \frac{7}{4} \\
0 & 1 & \frac{19}{21} \\
0 & 0 & -\frac{77}{21}
\end{pmatrix}
$$

### Step 2: Determine the Rank of Matrix B

Now that the matrix is in row echelon form, the rank of a matrix is the number of non-zero rows. In the matrix above, all three rows are non-zero. Therefore, the rank of matrix \( B \) is 3.

### Conclusion:

The rank of matrix \( B \) is **3**.
