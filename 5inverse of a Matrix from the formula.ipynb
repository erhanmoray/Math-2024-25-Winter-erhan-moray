# Inverse of a Matrix from the Formula

The **inverse of a matrix** is a matrix that, when multiplied by the original matrix, results in the identity matrix. Not all matrices have an inverse, but for those that do, the inverse is unique.

## Formula for the Inverse of a 2x2 Matrix

For a 2x2 matrix:
$$
A = \begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

The inverse of matrix $ A $, denoted $ A^{-1} $, is given by the formula:
$$
A^{-1} = \frac{1}{\det(A)} \begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix}
$$
Where:
- $ \det(A) = ad - bc $ is the determinant of matrix $ A $
- The matrix $ A $ must have a non-zero determinant for an inverse to exist.

### Example:
For the matrix $ A = \begin{bmatrix} 2 & 3 \\ 1 & 4 \end{bmatrix} $, the inverse is:
$$
A^{-1} = \frac{1}{(2 \cdot 4 - 3 \cdot 1)} \begin{bmatrix}
4 & -3 \\
-1 & 2
\end{bmatrix}
$$
$$
A^{-1} = \frac{1}{5} \begin{bmatrix}
4 & -3 \\
-1 & 2
\end{bmatrix}
$$
So,
$$
A^{-1} = \begin{bmatrix} \frac{4}{5} & \frac{-3}{5} \\ \frac{-1}{5} & \frac{2}{5} \end{bmatrix}
$$

## Formula for the Inverse of a 3x3 Matrix

For a 3x3 matrix:
$$
A = \begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix}
$$

The inverse of a 3x3 matrix is given by the formula:
$$
A^{-1} = \frac{1}{\det(A)} \begin{bmatrix}
\text{Cofactor of } a & \text{Cofactor of } b & \text{Cofactor of } c \\
\text{Cofactor of } d & \text{Cofactor of } e & \text{Cofactor of } f \\
\text{Cofactor of } g & \text{Cofactor of } h & \text{Cofactor of } i
\end{bmatrix}^T
$$
Where:
- $ \det(A) $ is the determinant of the matrix $ A $.
- The **cofactor** is the signed determinant of the submatrix formed by deleting the row and column of each element.

### Step-by-Step Process:
1. **Find the determinant** of the matrix $ A $.
2. **Find the cofactors** for each element in the matrix.
3. **Transpose** the cofactor matrix (i.e., take the matrix of cofactors and swap rows and columns).
4. **Multiply the cofactor matrix by the inverse of the determinant**.

### Example (Outline for 3x3 Matrix):
Given matrix:
$$
A = \begin{bmatrix}
1 & 2 & 3 \\
0 & 4 & 5 \\
1 & 0 & 6
\end{bmatrix}
$$

1. **Calculate the determinant** $ \det(A) $.
2. **Find the cofactors** for each element of the matrix.
3. **Transpose** the cofactor matrix.
4. **Multiply the transposed matrix** by $ \frac{1}{\det(A)} $.

---

## Conditions for Inverse Existence

- The matrix must be **square** (i.e., same number of rows and columns).
- The matrix must have a **non-zero determinant**. If $ \det(A) = 0 $, the matrix does not have an inverse.


# Finding the Inverse of Matrix A

Given the matrix $ A $:

$$
A = \begin{bmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{bmatrix}
$$

We can find the inverse of this matrix by using the formula for the inverse of a 3x3 matrix:

$$
A^{-1} = \frac{1}{\det(A)} \text{Cofactor}(A)^T
$$

### Step 1: Calculate the Determinant of Matrix $ A $

To find the determinant of a 3x3 matrix, we use the following formula:

$$
\det(A) = a(ei - fh) - b(di - fg) + c(dh - eg)
$$

For matrix $ A $:

$$
A = \begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix}
$$

We have:

$$
a = 2, b = 0, c = 1, d = 0, e = 1, f = 0, g = 1, h = 2, i = 0
$$

Substitute these values into the determinant formula:

$$
\det(A) = 2(1 \cdot 0 - 0 \cdot 2) - 0(0 \cdot 0 - 1 \cdot 1) + 1(0 \cdot 2 - 1 \cdot 1)
$$

Simplifying:

$$
\det(A) = 2(0) - 0 + 1(-1)
$$

$$
\det(A) = 0 - 1 = -1
$$


# Determining the Rank of Matrix $ B $

Given the matrix $ B $:

$$
B = \begin{bmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{bmatrix}
$$

### Step 1: Row Operations to Get Upper Triangular Form

We will use row operations to reduce the matrix to row echelon form (upper triangular form).

Start with the original matrix:

$$
B = \begin{bmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{bmatrix}
$$

Perform the following row operations:

1. Swap $ R_1 $ and $ R_2 $ to make the pivot element of the first row as 4.

After swapping:

$$
B = \begin{bmatrix}
-1 & 6 & 3 \\
4 & -3 & 7 \\
2 & 9 & 1
\end{bmatrix}
$$

2. Multiply $ R_1 $ by $ -1 $ to make the pivot element positive.

$$
B = \begin{bmatrix}
1 & -6 & -3 \\
4 & -3 & 7 \\
2 & 9 & 1
\end{bmatrix}
$$

3. Subtract $ 4R_1 $ from $ R_2 $ to eliminate the first column element in $ R_2 $.

$$
R_2 = R_2 - 4R_1 \Rightarrow \begin{bmatrix}
1 & -6 & -3 \\
0 & 21 & 19 \\
2 & 9 & 1
\end{bmatrix}
$$

4. Subtract $ 2R_1 $ from $ R_3 $ to eliminate the first column element in $ R_3 $.

$$
R_3 = R_3 - 2R_1 \Rightarrow \begin{bmatrix}
1 & -6 & -3 \\
0 & 21 & 19 \\
0 & 21 & 7
\end{bmatrix}
$$

### Step 2: Continue with Row Operations

5. Subtract $ R_2 $ from $ R_3 $ to eliminate the second column element in $ R_3 $.

$$
R_3 = R_3 - R_2 \Rightarrow \begin{bmatrix}
1 & -6 & -3 \\
0 & 21 & 19 \\
0 & 0 & -12
\end{bmatrix}
$$

Now the matrix is in upper triangular form.

### Step 3: Count the Number of Non-Zero Rows

The rank of a matrix is the number of non-zero rows in its row echelon form. In this case, all rows are non-zero.

Thus, the rank of matrix $ B $ is:

$$
\text{Rank}(B) = 3
$$
