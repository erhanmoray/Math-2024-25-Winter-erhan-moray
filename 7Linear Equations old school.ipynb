# Linear Equations - Old School Method

Linear equations can be solved using a variety of methods. In the "old school" approach, we use methods like substitution, elimination, and matrix techniques. These are the fundamental techniques that laid the foundation for solving systems of linear equations. Let's go through them.

## 1. Substitution Method

### Steps:
1. Solve one of the equations for one variable.
2. Substitute this expression into the other equations.
3. Continue until all variables are solved.

### Example:
Solve the system of equations:
$$
x + y = 10 \\
2x - y = 3
$$

1. From the first equation, solve for $y$:
$$ y = 10 - x $$

2. Substitute $y = 10 - x$ into the second equation:
$$ 2x - (10 - x) = 3 $$

3. Simplify and solve for $x$:
$$ 2x - 10 + x = 3 \\
3x = 13 \\
x = \frac{13}{3} $$

4. Substitute $x = \frac{13}{3}$ back into $y = 10 - x$ to find $y$:
$$ y = 10 - \frac{13}{3} = \frac{30}{3} - \frac{13}{3} = \frac{17}{3} $$

Thus, the solution is:
$$ x = \frac{13}{3}, \, y = \frac{17}{3} $$

## 2. Elimination Method

### Steps:
1. Multiply one or both equations to make the coefficients of one variable the same (or opposites).
2. Add or subtract the equations to eliminate that variable.
3. Solve the resulting equation.
4. Substitute back to find the other variable.

### Example:
Solve the system of equations:
$$
x + y = 10 \\
2x - y = 3
$$

1. Add the two equations to eliminate $y$:
$$ (x + y) + (2x - y) = 10 + 3 \\
3x = 13 \\
x = \frac{13}{3} $$

2. Substitute $x = \frac{13}{3}$ into one of the original equations to find $y$:
$$ x + y = 10 \\
\frac{13}{3} + y = 10 \\
y = 10 - \frac{13}{3} = \frac{17}{3} $$

Thus, the solution is:
$$ x = \frac{13}{3}, \, y = \frac{17}{3} $$

## 3. Matrix Method

When dealing with systems of linear equations, we can represent the system in matrix form and solve using methods such as Gaussian elimination or matrix inversion.

### General Form of System of Equations:
For a system of linear equations:
$$
a_1x + b_1y + c_1z = d_1 \\
a_2x + b_2y + c_2z = d_2 \\
a_3x + b_3y + c_3z = d_3
$$

We can write the system in matrix form as:
$$
\begin{bmatrix} a_1 & b_1 & c_1 \\ a_2 & b_2 & c_2 \\ a_3 & b_3 & c_3 \end{bmatrix}
\begin{bmatrix} x \\ y \\ z \end{bmatrix}
=
\begin{bmatrix} d_1 \\ d_2 \\ d_3 \end{bmatrix}
$$

### Matrix Solution:

1. Write the augmented matrix:
$$ [A | B] = \begin{bmatrix} a_1 & b_1 & c_1 & d_1 \\ a_2 & b_2 & c_2 & d_2 \\ a_3 & b_3 & c_3 & d_3 \end{bmatrix} $$

2. Apply row operations (Gaussian elimination) to transform the matrix into row echelon form.

3. If the matrix is invertible, we can find the solution vector $[x \, y \, z]^T$ by multiplying the inverse of matrix $A$ with vector $B$:
$$ \begin{bmatrix} x \\ y \\ z \end{bmatrix} = A^{-1} \begin{bmatrix} d_1 \\ d_2 \\ d_3 \end{bmatrix} $$

### Example:
Solve the system using matrices:
$$
x + y + z = 6 \\
2x + 3y + z = 14 \\
3x + 2y + 2z = 13
$$

1. Write the system in matrix form:
$$
\begin{bmatrix}
1 & 1 & 1 \\
2 & 3 & 1 \\
3 & 2 & 2
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}
=
\begin{bmatrix}
6 \\
14 \\
13
\end{bmatrix}
$$

2. Solve for the inverse of the coefficient matrix $A$ and multiply by the constants to find the solution.

## Conclusion

These old-school methods are still widely used, especially in simpler systems of linear equations. However, for more complex systems or larger matrices, modern computational tools like matrix inversion or Gaussian elimination (using computers) are preferred.


# Solving Systems of Equations Without Using Matrices

## System 1:
Solve the system of equations:
$$
3x - 2y = 5 \\
2x + 3y = 7
$$

### Step 1: Elimination Method

1. Multiply the first equation by 3 and the second equation by 2 to align the coefficients of $y$:
   $$ 9x - 6y = 15 \quad (Multiply \, by \, 3) \\
      4x + 6y = 14 \quad (Multiply \, by \, 2) $$

2. Add the two equations to eliminate $y$:
   $$ (9x - 6y) + (4x + 6y) = 15 + 14 \\
   13x = 29 \\
   x = \frac{29}{13} $$

3. Substitute $x = \frac{29}{13}$ into one of the original equations:
   $$ 3x - 2y = 5 \\
   3\left(\frac{29}{13}\right) - 2y = 5 \\
   \frac{87}{13} - 2y = 5 $$

4. Solve for $y$:
   $$ 2y = \frac{87}{13} - 5 \\
   2y = \frac{87}{13} - \frac{65}{13} \\
   2y = \frac{22}{13} \\
   y = \frac{11}{13} $$

Thus, the solution is:
$$ x = \frac{29}{13}, \quad y = \frac{11}{13} $$

## System 2:
Solve the system of equations:
$$
2x - 3y = 10 \\
4x + 5y = 20
$$

### Step 1: Elimination Method

1. Multiply the first equation by 5 and the second equation by 3 to align the coefficients of $y$:
   $$ 10x - 15y = 50 \quad (Multiply \, by \, 5) \\
      12x + 15y = 60 \quad (Multiply \, by \, 3) $$

2. Add the two equations to eliminate $y$:
   $$ (10x - 15y) + (12x + 15y) = 50 + 60 \\
   22x = 110 \\
   x = \frac{110}{22} = 5 $$

3. Substitute $x = 5$ into one of the original equations:
   $$ 2x - 3y = 10 \\
   2(5) - 3y = 10 \\
   10 - 3y = 10 $$

4. Solve for $y$:
   $$ -3y = 10 - 10 \\
   -3y = 0 \\
   y = 0 $$

Thus, the solution is:
$$ x = 5, \quad y = 0 $$

## System 3:
Solve the system of equations:
$$
2x - y + z = 3 \\
x + 2y - z = 1 \\
3x - y + 2z = 11
$$

### Step 1: Substitution Method

1. Solve the first equation for $z$:
   $$ 2x - y + z = 3 \\
   z = 3 - 2x + y $$

2. Substitute $z = 3 - 2x + y$ into the second and third equations:
   - For the second equation:
     $$ x + 2y - (3 - 2x + y) = 1 \\
     x + 2y - 3 + 2x - y = 1 \\
     3x + y = 4 $$

   - For the third equation:
     $$ 3x - y + 2(3 - 2x + y) = 11 \\
     3x - y + 6 - 4x + 2y = 11 \\
     -x + y = 5 $$

### Step 2: Solve the system of two equations:
$$
3x + y = 4 \\
-x + y = 5
$$

1. Subtract the second equation from the first to eliminate $y$:
   $$ (3x + y) - (-x + y) = 4 - 5 \\
   4x = -1 \\
   x = -\frac{1}{4} $$

2. Substitute $x = -\frac{1}{4}$ into $3x + y = 4$:
   $$ 3\left(-\frac{1}{4}\right) + y = 4 \\
   -\frac{3}{4} + y = 4 \\
   y = 4 + \frac{3}{4} = \frac{19}{4} $$

3. Substitute $x = -\frac{1}{4}$ and $y = \frac{19}{4}$ into $z = 3 - 2x + y$:
   $$ z = 3 - 2\left(-\frac{1}{4}\right) + \frac{19}{4} \\
   z = 3 + \frac{1}{2} + \frac{19}{4} \\
   z = \frac{12}{4} + \frac{2}{4} + \frac{19}{4} = \frac{33}{4} $$

Thus, the solution is:
$$ x = -\frac{1}{4}, \quad y = \frac{19}{4}, \quad z = \frac{33}{4} $$
