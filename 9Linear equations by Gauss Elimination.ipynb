### Linear Equations by Gauss Elimination

Gauss Elimination is a method used for solving a system of linear equations by transforming the system into an upper triangular matrix. Once in upper triangular form, the system can be solved using back-substitution.

#### Steps:

1. **Write the system as an augmented matrix**.
   A system of equations like:
   $$
   \begin{aligned}
   3x + y - z &= 1 \\
   2x - 2y + 3z &= 9 \\
   x + 4y - 2z &= 3
   \end{aligned}
   $$
   can be written as the augmented matrix:
   $$
   \begin{bmatrix}
   3 & 1 & -1 & | & 1 \\
   2 & -2 & 3 & | & 9 \\
   1 & 4 & -2 & | & 3
   \end{bmatrix}
   $$

2. **Use row operations to create zeros below the pivot element**.
   The pivot element in the first row is `3` (top-left value). We will use row operations to make the elements below it (in the first column) equal to zero.
   
   * Row 2: Subtract $\frac{2}{3}$ of row 1 from row 2:
   $$
   R_2 \rightarrow R_2 - \frac{2}{3} R_1
   $$
   * Row 3: Subtract $\frac{1}{3}$ of row 1 from row 3:
   $$
   R_3 \rightarrow R_3 - \frac{1}{3} R_1
   $$

3. **Repeat for the second column**.
   Now, focus on the pivot in the second column (the element at position (2,2)) and use row operations to create zeros below it.
   * Row 3: Subtract $\frac{4}{7}$ of row 2 from row 3:
   $$
   R_3 \rightarrow R_3 - \frac{4}{7} R_2
   $$

4. **Use back-substitution**.
   Once the matrix is in upper triangular form, solve the system by back-substitution.
   
   For example, if the final matrix looks like:
   $$
   \begin{bmatrix}
   3 & 1 & -1 & | & 1 \\
   0 & 7 & 5 & | & 7 \\
   0 & 0 & 1 & | & 2
   \end{bmatrix}
   $$
   The system is:
   $$
   \begin{aligned}
   3x + y - z &= 1 \\
   7y + 5z &= 7 \\
   z &= 2
   \end{aligned}
   $$

   * Start with the last equation: \( z = 2 \)
   * Substitute \( z = 2 \) into the second equation to solve for \( y \):
   $$
   7y + 5(2) = 7 \implies 7y + 10 = 7 \implies 7y = -3 \implies y = \frac{-3}{7}
   $$
   * Substitute \( y = \frac{-3}{7} \) and \( z = 2 \) into the first equation to solve for \( x \):
   $$
   3x + \frac{-3}{7} - 2 = 1
   $$
   Simplify and solve for \( x \).

Follow the steps outlined to reduce the system to upper triangular form and use back-substitution to find the values of \( x \), \( y \), and \( z \).

The final solution will give the values of \( x \), \( y \), and \( z \).

### Conclusion:

Gauss Elimination is a systematic process to solve systems of linear equations, especially for larger systems. The key steps are:
1. Write the augmented matrix.
2. Use row operations to create zeros below pivot elements.
3. Use back-substitution to solve for the variables.
